# Elementare Datentypen
---
Python kennt vier elementare Datentypen:

| Typ | Beschreibung | Beispiele |
|-----|-------------|-----------|
| `int` | Ganze Zahlen | `42`, `-7`, `0` |
| `float` | Dezimalzahlen | `3.14`, `-0.5`, `1e10` |
| `bool` | Wahrheitswerte | `True`, `False` |
| `str` | Zeichenketten | `"Hallo"`, `'Python'` |

Mit `type()` wird der Typ einer Variable abgefragt.
Mit `isinstance()` prüft man, ob eine Variable einen bestimmten Typ hat.

`````python
# Datentypen abfragen
zahl     = 42
komma    = 3.14
wahrheit = True
text     = "Python"

print(type(zahl))      # <class 'int'>
print(type(komma))     # <class 'float'>
print(type(wahrheit))  # <class 'bool'>
print(type(text))      # <class 'str'>
`````

`````python
# Typkonvertierung (Casting)
print(int("42"))      # str  → int:   42
print(float(7))       # int  → float: 7.0
print(str(3.14))      # float → str: '3.14'
print(bool(0))        # 0     → False  (0 gilt als 'falsy')
print(bool(""))       # ""    → False  (leerer String = falsy)
print(bool("Hallo"))  # "..." → True   (nicht-leerer String = truthy)
`````

`````python
# Typ prüfen mit isinstance()
x = 42
print(isinstance(x, int))            # True
print(isinstance(x, float))          # False
print(isinstance(x, (int, float)))   # True  – einer von beiden
`````

# Datentyp-unabhängige Funktionen
---

<p><font color='darkblue' size="4">
📌 <b>Wichtig</b>

Python unterscheidet zwischen
+ **Methoden** (*gebunden* an einen Typ: `x.methode()`, z.B. int, float, str, ...) und
+ **Built-in-Funktionen** (arbeiten *typ-übergreifend*: `funktion(x)`).

</font></p>




Viele Built-in-Funktionen funktionieren mit **mehreren** Datentypen:

| Funktion | Beschreibung | Beispiel |
|----------|--------------|----------|
| `print(x)` | Ausgabe | `print('Hallo')` |
| `input(text)` | Benutzereingabe (immer `str`) | `x = input('Zahl: ')` |
| `type(x)` | Datentyp ermitteln | `type(42)` → `<class 'int'>` |
| `isinstance(x, T)` | Typ prüfen | `isinstance(3.14, float)` → `True` |
| `len(x)` | Länge / Anzahl | `len('Hallo')` → `5` |
| `id(x)` | Speicheradresse | `id(42)` → `...` |
| `dir(x)` | Alle Methoden auflisten | `dir('')` → `[...]` |
| `help(x)` | Hilfedokumentation | `help(str.upper)` |
| `sorted(x)` | Sortierte Kopie | `sorted([3,1,2])` → `[1,2,3]` |
| `reversed(x)` | Umgekehrte Reihenfolge | `list(reversed([1,2,3]))` |

`````python
# Datentyp-unabhängige Built-in-Funktionen
wert = 3.14

print(type(wert))              # <class 'float'>
print(isinstance(wert, float)) # True
print(isinstance(wert, (int, float)))  # True – einer von beiden
print(id(wert))                # Speicheradresse (variiert)
`````

`````python
# sorted() und len() – typ-übergreifend
zahlen = [4, 1, 9, 2, 7]
text   = 'Python'

print(sorted(zahlen))          # [1, 2, 4, 7, 9]
print(sorted(text))            # ['P', 'h', 'n', 'o', 't', 'y']
print(len(zahlen))             # 5
print(len(text))               # 6
`````

# Int- und Float-Methoden
---


Zahlen besitzen eingebaute Methoden; Python stellt zusätzlich Built-in-Funktionen und das `math`-Modul bereit:

| Kategorie | Methoden / Funktionen |
|-----------|----------------------|
| Built-in **Funktionen** | `abs()`, `round()`, `divmod()`, `pow()`, `min()`, `max()`, `sum()` |
| **int/float**-Methoden | `.is_integer()`, `.as_integer_ratio()`, `.hex()` |
| **math**-Modul | `sqrt()`, `floor()`, `ceil()`, `factorial()`, `log()`, `pi`, `e` |

`````python
# int/float-Methoden
x = 3.75
print(x.is_integer())          # False – kein ganzzahliger Wert
print((4.0).is_integer())      # True
print(x.as_integer_ratio())    # (15, 4) – exakter Bruch: 15/4

# Sonderwerte
print(float('inf'))            # inf  – positiv unendlich
print(float('nan'))            # nan  – Not a Number
`````

`````python
# Built-in Funktionen für Zahlen
print(abs(-9.5))               # 9.5  – Absolutwert
print(round(3.14159, 2))       # 3.14 – Runden auf 2 Nachkommastellen
print(round(2.5))              # 2    – Banker's Rounding (auf gerade Zahl)
print(divmod(17, 5))           # (3, 2) – Quotient und Rest auf einmal
print(pow(2, 10))              # 1024 – 2 hoch 10

zahlen = [4, 1, 9, 2, 7]
print(min(zahlen))             # 1
print(max(zahlen))             # 9
print(sum(zahlen))             # 23
`````

`````python
# math-Modul – Mathematische Funktionen
import math

print(math.sqrt(144))          # 12.0 – Quadratwurzel
print(math.floor(3.9))         # 3    – Abrunden
print(math.ceil(3.1))          # 4    – Aufrunden
print(math.factorial(6))       # 720  – 6! = 6*5*4*3*2*1
print(math.log(100, 10))       # 2.0  – Logarithmus zur Basis 10
print(math.log2(1024))         # 10.0 – Logarithmus zur Basis 2
print(math.pi)                 # 3.141592...
print(math.e)                  # 2.718281...
`````

# String-Methoden

Strings besitzen viele eingebaute Methoden für Textverarbeitung:

| Kategorie | Methoden |
|-----------|---------|
| Groß-/Kleinschreibung | `.upper()`, `.lower()`, `.title()`, `.capitalize()` |
| Suchen & Prüfen | `.startswith()`, `.endswith()`, `.find()`, `.count()` |
| Bereinigen | `.strip()`, `.lstrip()`, `.rstrip()` |

`````python
# Groß-/Kleinschreibung
text = "hallo python-welt"
print(text.upper())       # HALLO PYTHON-WELT
print(text.lower())       # hallo python-welt
print(text.title())       # Hallo Python-Welt
print(text.capitalize())  # Hallo python-welt
`````

`````python
# Suchen und Prüfen
text = "Python ist toll"
print(text.startswith("Python"))   # True
print(text.endswith("toll"))       # True
print(text.find("ist"))            # 7   (Position, -1 wenn nicht gefunden)
print(text.count("t"))             # 3   (Anzahl der Vorkommen)
print("toll" in text)              # True (Alternative zu find)
`````

`````python
# Bereinigen – typisch für Eingaben und Datei-Import
eingabe = "   Hallo Welt   "
print(repr(eingabe.strip()))    # 'Hallo Welt' – beidseitig
print(repr(eingabe.lstrip()))   # 'Hallo Welt   ' – nur links
print(repr(eingabe.rstrip()))   # '   Hallo Welt' – nur rechts
`````

`````python
# Teilen und Verbinden
satz    = "Apfel,Birne,Orange,Mango"
fruechte = satz.split(",")           # ['Apfel', 'Birne', 'Orange', 'Mango']
print(fruechte)

print(" – ".join(fruechte))         # Apfel – Birne – Orange – Mango

# replace: Zeichen oder Teilstrings ersetzen
code = "Ich lerne Java."
print(code.replace("Java", "Python"))  # Ich lerne Python.
`````

## Indizieren von Listen
---

**Listenindex 0**

Der Listenindex in Python beginnt bei 0 aus historischen und technischen Gründen, die mit der Speicheradressierung zusammenhängen.

Gründe für das 0-basierte Indexieren:
+ Speicheradressierung: In Arrays entspricht der Index dem Offset zur Startadresse. Bei 0-basierter Indexierung liegt das erste Element an der Startadresse, was Berechnungen effizienter macht, da kein zusätzlicher Offset benötigt wird.

+ Geschichte: Frühe Programmiersprachen wie C nutzen 0-basierte Indexierung, und Python hat dieses Konzept übernommen, um konsistent zu bleiben.

+ Mathematische Logik: In vielen mathematischen und algorithmischen Konzepten beginnt das Zählen bei 0, was die Anwendung in der Programmierung vereinfacht.

**[Start:Ende:Schrittweite]**

+ Definiert einen Bereich (auch "Slice" genannt) innerhalb einer Sequenz (Liste oder String).
+ Start: Gibt den Index an, bei dem der Bereich beginnt (**inklusive**).
+ Ende: Gibt den Index an, bei dem der Bereich endet (**exklusive**).
+ Schrittweite: Bestimmt, wie viele Elemente beim Durchlaufen der Liste übersprungen werden
+ Kein Startindex: Slice beginnt am Anfang.
+ Kein Endindex: Slice geht bis zum Ende.
+ Keine Schrittweite: Alle Elemente
+ Kein Start- und kein Endindex: Liefert die gesamte Sequenz.

**Plus/Minus**

+ Plus: Eine positive Schrittweite bedeutet, dass man von **links nach rechts** durch die Sequenz iteriert.
- Minus: Eine negative Schrittweite bedeutet, dass man von **rechts nach links** durch die Sequenz iteriert.

**Index inklusive/exklusive**

In Python sind Indizes bei der Listindizierung so konzipiert, dass der Startindex inklusiv und der Endindex exklusiv ist. Diese Konvention bietet einige Vorteile:


+ **Einfache Berechnung der Länge**: Wenn der Endindex exklusiv ist, ist die Länge des Teils einer Liste oder eines Arrays einfach als `stop - start` zu berechnen. Zum Beispiel ergibt der Slice `list[1:4]` drei Elemente, und die Länge ist `4 - 1 = 3`.

+ **Natürliches Teilen**: Das exklusive Ende macht das Aufteilen der Daten in Abschnitte einfacher und intuitiver. Wenn Sie eine Liste in zwei teilen möchten, können Sie einfach den mittleren Index als Endpunkt für den ersten Abschnitt und als Startpunkt für den zweiten Abschnitt verwenden, ohne einen Offset hinzuzufügen oder abzuziehen. Zum Beispiel, `list[:n]` und `list[n:]` teilt die Liste genau am Index `n`.

+ **Vermeidung von Überlappungen und Lücken**: Da der Endindex exklusiv ist, gibt es keine Überlappungen oder Lücken zwischen aufeinanderfolgenden Slices. Das bedeutet, dass beim Hintereinanderausführen von Slices wie `list[:n]` und `list[n:]` jedes Element der Liste genau einmal erscheint.

+ **Einfachheit bei negativen Indizes**: Die Konvention hilft auch bei der Verwendung von negativen Indizes. Zum Beispiel gibt `list[-3:]` die letzten drei Elemente der Liste zurück, und Sie müssen nicht die Gesamtlänge der Liste berechnen, um den Endindex zu bestimmen.

+ **Konsistenz mit `range` und Slicing**: Die Syntax für Slicing (`start:stop`) ist konsistent mit der `range()`-Funktion. Beispielsweise gibt `range(0, 3)` und `list[0:3]` beide den Bereich von 0 bis 2 zurück. Das erleichtert das Verständnis und die Verwendung, da beide Mechanismen auf ähnliche Weise arbeiten.

Diese Konventionen tragen dazu bei, Fehler zu vermeiden und die Lesbarkeit des Codes zu verbessern, da die Absichten des Codierenden klarer werden. Es macht die Indizierung intuitiver und die Programmierung in Python effizienter.

## Weitere String-Formate
---

**Raw Strings (r-Strings):**

Werden durch ein vorangestelltes r oder R gekennzeichnet (z. B. r"Hallo\nWelt").
Backslashes (\) werden nicht als Escape-Sequenzen interpretiert, sondern als literale Zeichen behandelt.
Nützlich für reguläre Ausdrücke und Dateipfade, wo Backslashes häufig vorkommen.

**Byte Strings (b-Strings):**

Werden durch ein vorangestelltes b oder B gekennzeichnet (z. B. b"Hallo").
Repräsentieren eine Sequenz von Bytes, nicht von Unicode-Zeichen.
Werden oft für die Verarbeitung von Binärdaten verwendet.

**Unicode Strings (u-Strings):**

Werden durch ein vorangestelltes u oder U gekennzeichnet (z. B. u"Hallo").
Repräsentieren eine Sequenz von Unicode-Zeichen.
Ab Python 3 sind alle Strings standardmäßig Unicode-Strings, daher ist das u-Präfix meist nicht mehr notwendig.  [Unicodetabelle](https://symbl.cc/de/unicode-table/)

**Triple-Quoted Strings:**

Werden durch drei einfache oder doppelte Anführungszeichen begrenzt (z. B. '''Hallo''' oder """Welt""").
Können sich über mehrere Zeilen erstrecken und Zeilenumbrüche enthalten.
Oft für Docstrings und mehrzeilige Textblöcke verwendet.


| String-Typ | Kennzeichnung | Beschreibung |
|---|---|---|
| f-String | `f"Text"` | Formatierter String, ermöglicht das Einbetten von Variablen und Ausdrücken |
| Raw String | `r"Text"` | Backslashes werden nicht als Escape-Sequenzen interpretiert |
| Byte String | `b"Text"` | Sequenz von Bytes |
| Unicode String | `u"Text"` | Sequenz von Unicode-Zeichen (Standard ab Python 3) |
| Triple-Quoted String | `'''Text'''` oder `"""Text"""` | Mehrzeiliger String |

`````python
# f-String
name = "Max"
print(f"Hallo {name}!")  # Ausgabe: Hallo Max!
`````

`````python
# Raw String
path = r"C:\Benutzer\Max\Dokumente"
print(path)  # Ausgabe: C:\Benutzer\Max\Dokumente
`````

`````python
# Byte String
data = b"\x48\x61\x6c\x6c\x6F"
print(data)  # Ausgabe: b'Hallo'
`````

`````python
# Unicode String
text = u"Grüß Gott"
print(text)  # Ausgabe: Grüß Gott
`````

`````python
# Triple-Quoted String
docstring = """
Dies ist ein Docstring.
Er kann über mehrere Zeilen gehen.
"""
print(docstring)
`````

##    Bedingungen
---

Eine Bedingung ist eine logische Aussage, die entweder erfüllt ist oder nicht. Wenn sie erfüllt ist, dann wird die Bedingung als *wahr* bewertet, sonst als *falsch*. Für die Definition einer Bedingung können Vergleich- und Logikoperatoren eingesetzt werden. Mit den Logikoperatoren werden einzelne Bedingungen formuliert. Mit den Logikoperatoren könne mehrere Bedingungen miteinander verknüpft werden.




**Beipspiele:**
<br>
<br>

| Bedingung           | Prüfung                                                           |
|---------------------|-------------------------------------------------------------------|
| A == B              | Prüfung auf Gleichheit                                            |
| A != B              | Prüfung auf Ungleicheit                                           |
| A >= B              | Prüfung auf Größer-Gleich                                           |
| A <= B              | Prüfung auf KLeiner-Gleich                                           |
| x <= A <= y              | Prüfung auf Wertebereich                                          |
| A in B              | Prüfung, ob der Wert von A in der Liste von B enthalten ist       |
| (A == B and C != D) | Prüfung, ob  die 1. **und** die 2. Bedingung erfüllt ist          |
| (A == B or C != D)  | Prüfung, ob entweder die 1. **oder** die 2. Bedingung erfüllt ist |
| (A == B ^ C != D)  | Prüfung, ob entweder die 1. **exklusiv oder** die 2. Bedingung erfüllt ist |


`````python
# Beispiele
alter_pia = 30
alter_bob = 35

alter_pia > alter_bob
`````

`````python
18 <= alter_pia <= 65
`````

`````python
# --- AND: beide Bedingungen müssen WAHR sein ---
alter       = 20
hat_ausweis = True

zugang = (alter >= 18) and (hat_ausweis == True)
print(f"Zugang erlaubt: {zugang}")          # True – beides erfüllt
`````

`````python
# --- OR: mindestens eine Bedingung muss WAHR sein ---
hat_ticket    = False
ist_mitglied  = True

einlass = (hat_ticket == True) or (ist_mitglied == True)
print(f"Einlass möglich: {einlass}")        # True – Mitglied reicht
`````

`````python
# --- NOT: Bedingung umkehren ---
regen = False

spaziergang = not regen
print(f"Spaziergang möglich: {spaziergang}")  # True – kein Regen
`````

`````python
# --- XOR: genau eine Bedingung muss WAHR sein (Caret )---
zahlt_bar   = True
zahlt_karte = False

gueltig = zahlt_bar ^ zahlt_karte
print(f"Zahlung gültig: {gueltig}")         # True – genau eine Zahlart

zahlt_bar   = True
zahlt_karte = True
gueltig = zahlt_bar ^ zahlt_karte
print(f"Zahlung gültig: {gueltig}")         # False – beide gleichzeitig ungültig
`````

`````python
# --- Kombination ---
temperatur  = 22
sonnenschein = True
wind        = False

ideal = (temperatur >= 20) and sonnenschein and (not wind)
print(f"Ideales Wetter: {ideal}")           # True – alle Bedingungen erfüllt
`````

##    Einseitige Verzweigungen
---


Bei einer einseitigen Verzweigung wird ein Code-Block nur dann ausgeführt, wenn eine bestimmte Bedingung erfüllt ist.

Grundsätzlicher Aufbau: <br>

```python
      if Bedingung:
         Anweisung 1
         Anweisung 2
         ...
      Anweisung x
      Anweisung y
```

Der Code-Block nach der if-Anweisung muss eingerückt erfolgen. Das Ende des Code-Blocks erfolgt durch Aufhebung der Einrückung

---

Bei einer zweiseitigen Verzweigung wird ein Code-Block ausgeführt, wenn eine bestimmte Bedingung erfüllt ist. Ist Sie nicht erfüllt, wird ein alternativer Code-Block ausgeführt

Grundsätzlicher Aufbau: <br>

```python

         Anweisung 2
         ...
      else:
         Anweisung 3
         Anweisung 4
         ...
      Anweisung x

```
Der Code-Block nach der if-Anweisung muss eingerückt erfolgen. Das Ende des Code-Blocks erfolgt durch Aufhebung der Einrückung

#    Schleifen, Iterationen
---



Bei einer Schleife/Iteration wird ein Code-Block mehrfach durchgeführt. Die Anzahl der Wiederholungen kann auf zwei Arten festgelegt werden:

+ Die Wiederholung erfolgt solange eine **Bedingung erfüllt** ist: Ereignisgesteuerte Schleife
+ Die Wiederholung erfolgt nach einer vorher **festgelegt Anzahl**:  Zählergesteuerte Schleife



##    Ereignisgesteuerte Schleifen
---



Bei ereignisgesteurten Schleifen/Iterationen wird ein Code-Block solange wiederholt bis eine Bedingung erfüllt ist. Ist die Bedingung von Anfang an nicht erfüllt, dann wird dieser Code-Block nicht ausgeführt/übersprungen.

Grundsätzlicher Aufbau:

```python
        while Bedingung:
            Anweisung 1
            Anweisung 2
        else:
            Anweisung 3
            Anweisung 4
        Anweisung x
        Anweisung y
```


Raketenstart

`````python
CountDown = 10

while CountDown > 0:
    print(f"{CountDown}...")
    CountDown -= 1
print("Liftoff!")
`````

## Exkurs: range()
---

Die Funktion range() in Python erzeugt eine Sequenz von Ganzzahlen (Integer-Werten).


`range(start, stop, step)`

Parameter:
+ start (optional): Der Startwert der Sequenz. Die Sequenz beginnt bei diesem Wert. Wenn kein Startwert angegeben wird, wird standardmäßig bei 0 begonnen. Der Startwert ist inklusiv, d.h., er wird in der Sequenz eingeschlossen.

+ stop (erforderlich): Der Endwert der Sequenz. Die Sequenz endet vor diesem Wert, d.h., der Endwert ist exklusiv und wird nicht eingeschlossen.

+ step (optional): Der Schrittwert, der die Differenz zwischen aufeinanderfolgenden Zahlen der Sequenz angibt. Der Standardwert ist 1. Ein negativer Schrittwert ermöglicht das Iterieren in absteigender Reihenfolge, die in einer for-Schleife oder an anderen Stellen, an denen eine Sequenz benötigt wird, verwendet werden kann. range() ist eine eingebaute Funktion und gibt ein range-Objekt zurück, das ein Iterable ist.



Die Entscheidung, dass `start` bei der Funktion range in Python inklusiv und `stop` exklusiv ist, hat mehrere praktische Gründe, die auf Effizienz und Konsistenz abzielen:
+ Konsistenz mit Listenindizes: Python verwendet nullbasierte Indizes für Listen und andere sequenzielle Datentypen. Wenn man eine Liste von Elementen durchläuft, ist es üblich, range(len(liste)) zu verwenden, um alle Indizes der Liste zu erhalten. Da die Obergrenze exklusiv ist, wird der Index des letzten Elements korrekt berücksichtigt, ohne dass eine Anpassung erforderlich ist.
+ Mathematische Notation: Die Notation [start:end) (inklusive Start, exklusive Ende) ist in der Mathematik weit verbreitet und wird oft bei Intervallen verwendet. Diese Notation macht es einfach, Bereiche zu kombinieren oder zu teilen, ohne dass sich Bereiche überschneiden oder Lücken entstehen3.
+ Vermeidung von Off-by-One-Fehlern: Durch die exklusive Obergrenze wird das Risiko von Off-by-One-Fehlern reduziert, da man sich keine Gedanken darüber machen muss, den Endpunkt explizit zu reduzieren oder anzupassen.
+ Flexibilität bei der Bereichserstellung: Mit einer exklusiven Obergrenze kann man leicht Teilbereiche erstellen oder Bereiche aneinanderreihen. Zum Beispiel kann man zwei Bereiche range(a, b) und range(b, c) nahtlos aneinanderfügen, ohne dass eine Anpassung notwendig ist.

Diese Designentscheidung erleichtert viele Programmieraufgaben und sorgt für eine konsistente Handhabung von Bereichen in Python.

##    Vorzeitiges Beenden
----


Schleifen durchlaufen einen Code-Block, bis der die Bedingung falsch  oder der Zähler abgelaufen. Manchmal soll die aktuelle Iteration oder sogar die gesamte Schleife beenden.  <br>  <br>
In Python können break- und continue-Anweisungen den Ablauf einer normalen Schleife/Iteration ändern.
<br>
<br>


`````python
# Beispiel: Gib i aus, verlasse die Schleife, wenn i 5 ist:
#
i = 1
while True:
    print(i)
    if i == 5:
      break
    i += 1
`````

`````python
# Beisiel: Drucken eine Fruchtliste aus, verlasse die Schleife bei der Frucht Kirsche
#
obst = ['🍉', '🍈', '🍎', '🍏', '🍒', '🍓']
for frucht in obst:
    print(frucht)
    if frucht == '🍒':
        break
`````

`````python
# Beispiel: Drucken eine Fruchtliste aus, verlasse die Schleife für die "roter Apfel"
#
obst = ['🍉', '🍈', '🍎', '🍏', '🍒', '🍓']
for frucht in obst:
    if frucht == '🍎':
        continue
    print(frucht)
`````

##    Dictionary

Dictionary's (Dict)  werden verwendet, um Datenwerte in Schlüssel-Wert-Paaren zu speichern. Ein Dictionary ist eine Sammlung, die geordnet ist, veränderbar ist und keine Duplikate zulässt. Ein Dictionary ist veränderbar, was bedeutet, dass Elemente verändert, hingezufügt oder entfernt werden kann, nachdem das Dictionary erstellt wurde.

##    Tuple
---


Tuple werden verwendet, um mehrere Elemente in einer einzigen Variablen zu speichern. Tuple ist eine von 4 integrierten Datentypen in Python, die zum Speichern von Datensammlungen verwendet werden, die anderen 3 sind List, Set und Dictionary, alle mit unterschiedlichen Qualitäten und Verwendungen.

Ein Tuple ist eine Sammlung, die **geordnet** und **unveränderlich** ist. D.h., dass Elemente nicht geändert, hinzugefügt oder entfernt werden können, nachdem das Tuple erstellt wurde. Geordnet bedeutet, dass die Elemente eine definierte Reihenfolge haben und diese Reihenfolge sich nicht ändert.

Tuples besitzen wenige Methoden – sie sind unveränderlich; Python stellt Built-in-Funktionen bereit:

| Kategorie | Methoden / Funktionen |
|-----------|----------------------|
| Erstellen & Konvertieren | `tuple()`, `()` |
| Länge & Aggregation | `len()`, `min()`, `max()` |
| Suchen & Zählen | `.index()`, `.count()` |
| Prüfen & Iterieren | `value in t`, `for elem in t` |





Tuple werden mit runden Klammern geschrieben.

`````python
# Beispiel: Drucken Sie jede Frucht in einer Fruchtliste aus
frucht = ("Apfel", "Banane", "Kirsche", "Birne")
for x in frucht:
  print(x)

print(type(frucht))
`````

##    Set
---


Sets werden verwendet, um mehrere Elemente in einer einzigen Variablen zu speichern. Set ist einer von 4 integrierten Datentypen in Python, die zum Speichern von Datensammlungen verwendet werden, die anderen 3 sind List, und Dictionary, alle mit unterschiedlichen Qualitäten und Verwendungen.

Ein Set ist eine Sammlung, die sowohl **ungeordnet** als auch  **unveränderlich** ist. Ungeordnet bedeutet, dass die Artikel in einem Satz keine definierte Reihenfolge haben. Sets dürfen nicht zwei Elemente mit demselben Wert haben. Sets sind unveränderlich, was bedeutet, dass die Elemente nicht geändert werden können, nachdem der Set erstellt wurde, aber man kann Elemente entfernen und neue Elemente hinzufügen.

Sets besitzen eingebaute Methoden; Python stellt zusätzlich Mengenoperationen bereit:

| Kategorie | Methoden / Funktionen |
|-----------|----------------------|
| Erstellen | `set()`, `{}` |
| Hinzufügen & Entfernen | `.add()`, `.remove()`, `.discard()`, `.pop()`, `.clear()` |
| Mengenoperationen | `.union()` (`|`), `.intersection()` (`&`), `.difference()` (`-`), `.symmetric_difference()` (`^`) |
| Aktualisieren | `.update()`, `.intersection_update()`, `.difference_update()` |
| Prüfen & Iterieren | `elem in s`, `len(s)`, `for elem in s` |





Sets werden mit geschweiften Klammern geschrieben.

**Besonderheit**: <br>
Auf Sets können die Operatoren der Mengenlehre angewandt werden (Vereinigungsmenge, Schnittmenge, Differenzmenge, symetrische Differenzmenge).

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Code für Grafik</font> </br></p>
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Definition der Mengen
A = {"Apfel", "Banane", "Kirsche", "Birne"}
B = {"Feige", "Aprikose", "Kirsche", "Birne", "Kiwi", "Orange"}

# Funktion um alle Kreise des Venn-Diagramms mit einem Rahmen zu versehen
def add_all_patch_borders_to_venn(venn_diagram, ax):
    for region in venn_diagram.id2idx.keys():
        try:
            path = venn_diagram.get_patch_by_id(region).get_path()
            patch = plt.PathPatch(path, edgecolor='black', lw=1)
            ax.add_patch(patch)
        except:
            pass

# Erstellung eines Plots mit 4 Subplots nebeneinander
fig, axs = plt.subplots(1, 4, figsize=(20, 5))

# Vereinigungsmenge
v = venn2(subsets=(len(A - B), len(B - A), len(A & B)), set_labels=('A', 'B'), ax=axs[0])
add_all_patch_borders_to_venn(v, axs[0])
axs[0].set_title('Vereinigungsmenge (A ∪ B)')

# Schnittmenge
v = venn2(subsets=(len(A - B), len(B - A), len(A & B)), set_labels=('A', 'B'), ax=axs[1])
add_all_patch_borders_to_venn(v, axs[1])
v.get_label_by_id('10').set_text('')
v.get_label_by_id('01').set_text('')
axs[1].set_title('Schnittmenge (A ∩ B)')

# Differenzmenge
v = venn2(subsets=(len(A - B), len(B - A), len(A & B)), set_labels=('A', 'B'), ax=axs[2])
add_all_patch_borders_to_venn(v, axs[2])
v.get_label_by_id('01').set_text('')
v.get_label_by_id('11').set_text('')
axs[2].set_title('Differenzmenge (A - B)')

# Symmetrische Differenzmenge
v = venn2(subsets=(len(A - B), len(B - A), len(A & B)), set_labels=('A', 'B'), ax=axs[3])
add_all_patch_borders_to_venn(v, axs[3])
v.get_label_by_id('11').set_text('')
axs[3].set_title('Symmetrische Differenzmenge (A Δ B)')

# Anzeige des Plots
plt.tight_layout()
plt.show()
`````

## Vergleich List, Dict, Tuple, Set
---


| **Eigenschaft**         | **List**                           | **Dict**                                    | **Tuple**                    | **Set**                                    |
| ----------------------- | ----------------------------------- | ------------------------------------------- | ---------------------------- | ------------------------------------------ |
| **Ordnung**             | Geordnet (ab Python 3.7 garantiert) | Geordnet (ab Python 3.7 garantiert)         | Geordnet                     | Ungeordnet, keine Garantie der Reihenfolge |
| **Duplikate**           | Erlaubt                             | Schlüssel: nicht erlaubt<br>Werte: erlaubt  | Erlaubt                      | Nicht erlaubt                              |
| **Änderbarkeit**        | Änderbar (mutabel)                  | Änderbar (mutabel)                          | Nicht änderbar (unmutabel)   | Änderbar, aber einzelne Elemente unmutabel |
| **Indexierung**         | Unterstützt (per Index)             | Über Schlüssel                              | Unterstützt (per Index)      | Nicht unterstützt                          |
| **Suchzeit**            | linear                              | sehr schnell (Hash-Zugriff)                 | linear                       | sehr schnell (Hash-Zugriff)                |
| **Einfügezeit am Ende** | konstant                            | konstant (ab Python 3.7 Einfügereihenfolge) | N/A (unmutabel)              | konstant                                   |
| **Syntax**              | `[1, 2, 3]`                         | `{'a': 1, 'b': 2}`                          | `(1, 2, 3)`                  | `{1, 2, 3}`                                |
| **Verwendung**          | Allgemeine Datenstruktur            | Schlüssel-Wert-Paare, Mapping               | Feste, unveränderliche Daten | Eindeutige Werte, Mengenoperationen        |




🔧 Hash-Zugriff:

Jedes Element (bei set) oder jeder Schlüssel (bei dict) wird durch eine Hashfunktion in eine Zahl (den Hashwert) umgewandelt.

Dieser Hashwert bestimmt, wo im Speicher das Element oder der Wert liegt.

Beim Zugriff oder der Suche wird nicht die ganze Struktur durchsucht, sondern nur der Hashwert berechnet → direkter Zugriff.

# Comprehensions
---

Comprehensions in Python bieten eine kompakte und expressive Syntax, um neue Kollektionen basierend auf existierenden Sequenzen oder Iterables zu erstellen. Sie erlauben es, Elemente zu filtern, zu transformieren und auf effiziente Weise in einer neuen Kollektion zusammenzufassen. ## Funktionen mit variablen Argumenten

```*args``` steht für "arguments" und ermöglicht es, eine beliebige Anzahl von Positional-Argumenten (Argumente, die nicht durch ein Schlüsselwort benannt sind) an eine Funktion zu übergeben. In der Funktion werden diese Argumente als  zugänglich gemacht. Das Sternchen (*) vor dem Namen args ist entscheidend, da es Python anweist, alle zusätzlichen Positional-Argumente in einem  zu sammeln.

`````python
def summe(*args):
    return sum(args)

print(summe(1, 2, 3, 4))
`````

`````python
def my_function(*args):
    ln = len(args)
    y = args[-1]
    print(f"Der Funktion wurden {ln} Argumente übergeben, das letzte Argument ist: {y}")

my_function("Apfel", "Birne", 99, "Karotte")
my_function("Apfel", "Birne", 99, "Karotte", 55)
`````

```**kwargs``` steht für "keyword arguments" und ermöglicht die Übergabe einer beliebigen Anzahl von Schlüsselwortargumenten (Argumente, die durch ein Schlüsselwort benannt sind) an eine Funktion. In der Funktion werden diese Argumente als Wörterbuch zugänglich gemacht, wobei die Schlüsselwörter die Schlüssel und die zugehörigen Werte die Werte des Wörterbuchs sind. Das doppelte Sternchen (**) vor dem Namen kwargs ist entscheidend, da es Python anweist, alle zusätzlichen Schlüsselwort-Argumente in einem Wörterbuch zu sammeln.

`````python
def begruessung(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

begruessung(Name="Max", Alter=30)  # Gibt "Name: Max" und "Alter: 30" aus
`````

4! = 1 * 2 * 3 * 4

# Higher-Order-Functions (HOFs)
---

Higher-Order Functions sind Funktionen, die andere Funktionen als Argumente annehmen oder Funktionen als Ergebnis zurückgeben. map und filter sind klassische Beispiele dafür, da beide eine Funktion als eines ihrer Argumente nehmen. Beide Funktionen geben einen Iteratoren zurück. Ein Iterator in Python ist ein Objekt, das eine Sequenz von Werten nacheinander durchlaufen kann. Werte Werte werden nach und nach bereitgestellt. Es erfolgt eine sog. *Lazy Evaluation*.

**Lazy Evaluation**, oder **träges Auswerten**, bedeutet, dass der Wert eines Ausdrucks nur berechnet wird, **wenn er benötigt wird**. Dies kann die Effizienz verbessern, insbesondere bei der Verarbeitung großer Datensätze oder bei der Erstellung von potenziell unendlichen Datenstrukturen. ```map``` und ```filter``` in Python geben einen Iteratoren zurück, die diesem Prinzip folgen, da Elemente einzeln und nur auf Anfrage bearbeitet werden.

Die Ergebnisse eines Iterators lassen sich z.B. mit list(), zählergesteuerte Schleife, next() oder Verwendung von '*' Operator anzeigen.

| Eigenschaft                  | Eager Evaluation                                   | Lazy Evaluation                           |
|------------------------------|---------------------------------------------------|-------------------------------------------|
| Wann werden die Elemente generiert? | Sofort (alle auf einmal)                           | On demand (bei Bedarf)                     |
| Speicherverbrauch             | Höher (alle Elemente werden im Voraus gespeichert) | Gering (nur das aktuell benötigte Element) |
| Beispiele in Python           | Listen                                    | range(), map(), filter()                   |

##    map Funktion

##    filter Funktion
---

Die `filter` Funktion verarbeitet eine Eingabe-Liste und filtert Werte nach einer Wahrheitsbedingung heraus.




`filter(boolean-function, liste)`


`````python
# Beispiel für eine benutzerdefinierte Funktion, die gerade Zahlen filtert
def is_even(number):
    return number % 2 == 0

numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(is_even, numbers))
print(even_numbers)  # Ausgabe: [2, 4, 6]
`````

`````python
# Beispiel mit einer Lambda-Funktion, die ungerade Zahlen filtert
odd_numbers = list(filter(lambda x: x % 2 != 0, numbers))
print(odd_numbers)  # Ausgabe: [1, 3, 5]
`````

`````python
# Beispiel mit None als Funktion, um alle "wahren" Werte zu filtern
mixed_list = [0, 1, '', 'text', None, [], [1], False, True]
truthy_values = list(filter(None, mixed_list))
print(truthy_values)  # Ausgabe: [1, 'text', [1], True]
`````

`````python
# Beispiel für die Verwendung von `filter` mit einer benutzerdefinierten Funktion, die ein Kriterium prüft
def is_positive(number):
    return number > 0

positive_numbers = list(filter(is_positive, [-2, -1, 0, 1, 2]))
print(positive_numbers)  # Ausgabe: [1, 2]
`````

##    Klassen und Instanzen
---

+ **Klasse**: ist der *`Bauplan`* oder die Vorlage für Objekte. Sie definiert die Struktur (Attribute) und das Verhalten (Methoden), über die alle daraus erstellten Instanzen verfügen werden.

+ **Instanz**: ist eine *`konkrete Ausprägung`* der Klasse, die die in der Klasse definierten Funktionalitäten ausführen können.

+ **@dataclass**: Ein Dekorator (seit Python 3.7), der für Klassen *`automatisch Standardmethoden`* erstellt und so Boilerplate-Code (Code-Abschnitte, die an vielen Stellen fast unverändert wiederholt werden müssen) spart.

+ **Decorator**: ist eine Art Hülle um eine Funktion oder Klasse. Er sorgt dafür, dass zusätzliches Verhalten hinzugefügt wird - ohne den ursprünglichen Code zu ändern.

+ **field()**: Funktion aus `dataclasses`, mit der einzelne Felder konfiguriert werden, z.B. `field(default=100, init=False)` für Felder ohne Init-Parameter, oder `field(default_factory=list)` für veränderliche Defaults.


+ **ClassVar**: ist eine Typ-Annotation aus `typing`, die ein Attribut als Klassenattribut kennzeichnet. `@dataclass` überspringt `ClassVar`-Felder bei der automatischen Generierung von Attributen.

## Defaultwerte bei Attributen
---

Attribute können mit einem **Standardwert** (Default) versehen werden. Beim Erstellen einer Instanz kann der Wert dann weggelassen, überschrieben oder explizit angegeben werden:

| Attribut | Pflichtfeld | Mit Default |
|----------|-------------|-------------|
| `hersteller: str` | ✅ muss übergeben werden | — |
| `modell: str` | ✅ muss übergeben werden | — |
| `leistung: int` | ✅ muss übergeben werden | — |
| `baujahr: int = 2024` | — | ✅ optional, Default wird verwendet |
| `farbe: str = "Schwarz"` | — | ✅ optional, Default wird verwendet |
| `elektro: bool = False` | — | ✅ optional, Default wird verwendet |

> ⚠️ **Reihenfolge**: Pflichtfelder (ohne Default) müssen **vor** Feldern mit Default stehen.

In Python ist `self` eine Konvention, die in Attributen und Methoden einer Klasse verwendet wird. Es dient als erster Parameter und repräsentiert die **aktuelle Instanz der Klasse**, auf der die Methode aufgerufen wird. Eine Klasse definiert den allgemeinen Bauplan, von der beliebig viele Instanzen erzeugt werden können. Damit das Programm weiß, auf welcher Instanz es gerade arbeitet, wird `self` verwendet.

> `self` ist wie das Wort 'ich' in einem Tagebucheintrag: Es sagt Python, dass du gerade über dich selbst (die  aktuelle Instanz) sprichst."


<details>

<summary><strong>

</strong></summary


**Hätte man ohne `self` auskommen können?**   
Technisch wäre es möglich gewesen, OOP in Python ohne die explizite Verwendung von self zu gestalten, wie es in manchen anderen Sprachen der Fall ist.

Allerdings würde das:

+ Weniger Flexibilität bieten: Durch die explizite Verwendung von self kann eine Methode auch von anderen Instanzen oder **ohne Instanzbezug** aufgerufen werden, was zusätzliche Möglichkeiten schafft.
+ Die Philosophie von Python verletzen: Python legt großen Wert auf expliziten und leicht verständlichen Code. Ein verstecktes oder implizites self würde diese Philosophie untergraben.

**Fazit:**
Obwohl es theoretisch möglich gewesen wäre, OOP in Python ohne self zu gestalten, entspricht die explizite Verwendung von self der Philosophie von Python und trägt zur Klarheit und Flexibilität des Codes bei.

`````python
# Beispiel: Klasse Auto um Methode erweitern
from dataclasses import dataclass

@dataclass
class Auto:
    # --- Instanzattribute ---
    hersteller: str
    modell: str
    leistung: int
    energie: float = 100.0
    x_pos: float   = 0.0
    y_pos: float   = 0.0

    # --- Instanzmethode ---
    def fahren(self, a, b):
        self.x_pos += a
        self.y_pos += b
        self.energie -= (a + b) * 0.25
        print("\nMethode Fahren")
        print(f"neue Position: {self.x_pos} {self.y_pos} Energie: {self.energie:.2f}")

    def tanken(self):
        print("\nMethode Tanken")
        print("vorher: Energie:", self.energie, end=" ")
        self.energie = 100
        print(f"nachher: Energie: {self.energie:.2f}")
`````

## Instanzen-Interaktionen
---

In der objektorientierten Programmierung (OOP) interagieren Instanzen miteinander durch Methodenaufrufe. Eine Instanz einer Klasse kann Methoden einer anderen Instanz aufrufen, um bestimmte Aktionen durchzuführen oder Informationen zu erhalten.

Um in Python eine Instanz einer Klasse an eine Methode einer anderen Klasse zu übergeben, kann man einfach die Instanz als ein Argument der Methode behandeln.
Die Methode kann dann auf die Attribute und Methoden der übergebenen Instanz zugreifen und sie bei Bedarf ändern.

<p><font color='black' size="5">
Version 1: Beteiligte und Aktionen (eine Klasse)
</font></p>

`````python
from dataclasses import dataclass

@dataclass
class Ritter:
    # --- Instanzattribute ---
    name: str
    lebenspunkte: int = 100

    # --- Instanzmethode ---
    def schwertangriff(self, gegner):
        if gegner.lebenspunkte >= 25:
            gegner.lebenspunkte -= 25
            print(f"Der Ritter {self.name} hat dem {gegner.name} einen Schaden (-25) zugefügt!")
        else:
            print(f"Der Gegner {gegner.name} ist tod, er hat nicht mehr genug Lebenspunkte!")
`````

`````python
# Instanzen erstellen
kunibert = Ritter("Kunibert")
eisenherz = Ritter("Eisenherz")
`````

`````python
# Kunibert greift den Eisenherz an und verursacht Schaden
kunibert.schwertangriff(eisenherz)
eisenherz
`````

`````python
# Eisenherz greift den Kunibert an und verursacht Schaden
eisenherz.schwertangriff(kunibert)
kunibert
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Instanzenliste</font> </br></p>

from dataclasses import dataclass
from typing import ClassVar

@dataclass
class Ritter:
    # --- Klassenattribut ---
    _alle_instanzen: ClassVar[list] = []  # ClassVar: @dataclass überspringt dieses Feld (kein __init__-Parameter)

    # --- Instanzattribute ---
    name: str
    waffe: str

    # --- Initialisierung (Instanzmethode, automatisch aufgerufen) ---
    def __post_init__(self):
        self.__class__._alle_instanzen.append(self)
        print(f"Instanz von {self.name} wurde in Klasse {self.__class__.__name__} erstellt.")

    # --- Instanzmethode ---
    def loeschen(self):
        self.__class__._alle_instanzen.remove(self)
        print(f"Instanz von {self.name} wurde gelöscht.")

    # --- Klassenmethode ---
    @classmethod
    def get_instanzen(cls):
        return list(cls._alle_instanzen)
`````

`````python
# Acht Instanzen erstellen
adrian       = Ritter("Adrian von Edelfürst",               "Schwert")
bartholomeus = Ritter("Bartholomeus Drachenkuss",            "Lanze")
benno        = Ritter("Benno der Mutige von Falkenberg",     "Bogen")
wolfherz     = Ritter("Lady Wolfherz von Schattenwald",      "Morgenstern")
tristan      = Ritter("Tristan von Rettemich",               "Streitaxt")
alexander    = Ritter("Alexander Räubertum von Löwenstein",  "Hellebarde")
anton        = Ritter("Anton zu Schafspelz",                 "Armbrust")
bella        = Ritter("Bella von Bärenstark",                "Dolch")
`````

`````python
# Instanzen abfragen
print(f"Anzahl der Instanzen: {len(Ritter.get_instanzen())}")
print()

for item in Ritter.get_instanzen():
    print(item)
`````

#    Bibliotheken
---


**Bibliotheken:**

Im allgemeinen Sprachgebrauch wird in Python oft von "Bibliothek" gesprochen, wenn es um importierte Module oder Pakete geht.

**Modul:**

Ein Modul ist eine einzelne Datei, die Python-Code enthält.
Es kann Funktionen, Klassen und Variablen sowie ausführbaren Code enthalten.
Module sind der Grundmechanismus in Python, um Code in logische Einheiten zu organisieren und den Namensraum zu trennen.
Ein Modul kann mit dem Schlüsselwort import in einen anderen Python-Code importiert werden.


**Paket:**

Ein Paket ist eine Sammlung von Python-Modulen, die in einem Verzeichnis organisiert sind.
Das Verzeichnis muss eine __init__.py-Datei enthalten (die auch leer sein kann), um von Python als Paket erkannt zu werden.
Pakete ermöglichen eine hierarchische Strukturierung des Modul-Namensraums.
Ein Modul in einem Paket wird als "Untermodul" bezeichnet und kann mit dem Punkt-Operator importiert werden, z.B. import paketname.modulname.

##    Module & Pakete
---


Module bestehen, ebenso wie Python-Skript, aus Quellcode-Dateien mit der Endung .py. Prinzipiell kann somit jedes Python-Skript als eigenständiges Python-Modul angesehen werden, es existieren allerdings auch Module, die nur nur aus Hilfsfunktionen bestehen.

In der Praxis werden Module in erster Linie verwendet, um den Quellcode eines umfangreicheren Software-Projekts leichter zu organisieren und übersichtlich zu halten. Eine grobe Regel besagt, dass ein Modul nur so umfangreich sein sollte, dass sich seine Inhalte und sein Verwendungszweck mit ein bis zwei Sätzen zusammenfassen lassen sollten.

Jedes Modul stellt einen eigenen Namensraum für Variablen dar, so dass gleichnamige Funktionen, die in unterschiedlichen Modulen definiert sind, keine Konflikte verursachen.

Modulnamen werden in Python allgemein in Kleinbuchstaben geschrieben. Um in das aktuelle Programm ein externes Modul, also eine andere .py-Datei, zu integrieren, ist folgende Syntax üblich:

````
import modulname

````

Nach dem Importieren eines Moduls können die im Modul definierten Klassen, Funktionen und Variablen mit folgender Syntax aufgerufen werden:

+ modulname.Klasse
+ modulname.Variable
+ modulname.Funktion

Lange Modulnamen können mit einem `alias` abgekürzt werden:

````
import modulname as alias

````

Direkter Import von Funktionen oder Klasssen:

````
from modulname import funktion
````

`````python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

url = 'https://en.wikipedia.org/wiki/Main_Page'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'}

request = Request(url, headers=headers)
with urlopen(request) as response:
    soup = BeautifulSoup(response, 'html.parser')
    i = 0
    for anchor in soup.find_all('a'):
        print(anchor.get('href', '/'))
        i += 1
        if i == 25:
            break
`````

##    csv/txt Dateien (Python)
---

Umfangreicher Ein-/Ausgaben müssen über Dateien realisiert werden. Die Funktion open() liefert eine Referenz auf eine Datei. Zudem muss die Zugriffsart definiert erden, z.B. "w" für schreibenden und "r" für lesenden Zugriff. Mit dem Mode "a", können neue Daten an eine bestehende Datei angehängt werden. Nach Benutzung der Datei muss dieses mit der Funktion close() geschlossen werden.

Die Hauptmodi und ihre Bedeutungen:

+ r: Öffnet die Datei zum Lesen. Dies ist der Standardmodus, wenn kein Modus angegeben ist. Die Datei muss existieren, sonst wird eine Ausnahme ausgelöst.

+ w: Öffnet die Datei zum Schreiben. Wenn die Datei bereits existiert, wird ihr Inhalt gelöscht, bevor mit dem Schreiben begonnen wird. Wenn die Datei nicht existiert, wird sie erstellt.

+ a: Öffnet die Datei zum Anhängen. Wenn die Datei existiert, wird der neue Inhalt an das Ende der Datei angehängt, ohne den bestehenden Inhalt zu löschen. Wenn die Datei nicht existiert, wird sie erstellt.

Mit `with`: Der `with`-Befehl in Python wird verwendet, um das automatische Öffnen und Schließen von Dateien durchzuführen.

`````python
#
# ------ Öffne die Datei im Modus 'r+' (Lesen und Schreiben), Datei wird nach Ausführung "with" (Kontextmanager, siehe unten) geschlossen
#
with open(dateiname, 'r+') as datei_1:
    # Lies den Inhalt der Datei
    inhalt_der_datei = datei_1.read()
    print('Aktueller Inhalt:\n', "-" * 40, "\n", inhalt_der_datei)

    # Schreibe weitere Datensätze in die Datei
    for i in range(10):
        datei_1.write(f"{generate_fake_person()}\n")

    # Parameter: 0 (Anfang der Datei), 1 (aktuelle Position) und 2 (Ende der Datei)
    # Optional: Gehe zum Anfang der Datei, um den neuen Inhalt zu lesen
    datei_1.seek(0)
    NeuerInhalt = datei_1.read()
    print('Neuer Inhalt:\n', "-"*40, "\n", NeuerInhalt)
`````

#    Globale/lokale Variablen
---




Variablen, die außerhalb einer Funktion erstellt werden (wie in den Beispielen unter), werden als globale Variablen bezeichnet. Globale Variablen können im Code verwendet werden, sowohl innerhalb als auch außerhalb von Funktionen. Wird eine Variable mit demselben Namen innerhalb einer Funktion erstellen, ist diese Variable lokal und kann nur innerhalb der Funktion verwendet werden. Die globale Variable mit dem gleichen Namen bleibt unverändert, global und mit dem ursprünglichen Wert.

Um eine globale Variable innerhalb einer Funktion zu erstellen, kann man das Schlüsselwort "global" verwenden.

**Grundregeln**:




| | Globale Variable | Lokale Variable |
|---|---|---|
| **Lebt in** | gesamtem Programm | nur innerhalb der Funktion |
| **Lesen** in Funktion | ✅ erlaubt | ✅ erlaubt |
| **Schreiben** in Funktion | nur mit `global` | direkt möglich |

`````python
# Variante 1: Je Fehlerfall werden Anweisungen definiert
#
try:
    x = int(input("Bitte erste Zahlen eingeben: "))
    y = int(input("Bitte zweite Zahlen eingeben: "))
    z = x/y
    print(z)
    print("Programm ohne Fehler beendet")
except:
    print("Es gab einen Fehler")
`````

`````python
# Variante 2: Je Fehlerfall werden Anweisungen definiert
#
try:
    x = int(input("Bitte erste Zahlen eingeben: "))
    y = int(input("Bitte zweite Zahlen eingeben: "))
    z = x/y
    print(z)
    print("Programm ohne Fehler beendet")
except ZeroDivisionError:
    print("Fehler bei Division durch Null, Programm wird beendet")
except ValueError:
    print("Es sollte eine Zahl eingegeben werden")
except:
    print("Es gab einen anderen Fehler")
`````

`````python
# Variante 3: Für mehrere Fehlerfälle wird ein Code-Block verwendet
#
def programm():
    x = int(input("Bitte erste Zahlen eingeben: "))
    y = int(input("Bitte zweite Zahlen eingeben: "))
    z = x/y
    print(z)
    print("Programm ohne Fehler beendet")
`````

`````python
try:
    programm()
except (ZeroDivisionError, ValueError) as error_descr:
    print("Es gab den Fehler", error_descr)
except:
    print("Es gab einen weiteren Fehler")
`````

`````python
# Variante 4: Fehlerfälle werden nicht unterschieden (one-fits-all)
#
try:
    programm()
except Exception as error_descr:
    print("Mit Fehler: ", error_descr, "beendet")
`````

`````python
# Variante 5: Fehlerverfolgung
#
import traceback
import sys

try:
    programm()
except Exception:
    # exc_info() liefert ein Tupel mit (Typ, Wert, Traceback-Objekt)
    exc_type, exc_value, exc_traceback = sys.exc_info()

    # Traceback-Informationen extrahieren
    filename, line_number, function_name, text = traceback.extract_tb(exc_traceback)[-1]

    print(f"Ein Fehler ist in der Datei '{filename}' aufgetreten.")
    print(f"Fehler in Zeile: {line_number}")
    print(f"In der Funktion: '{function_name}'")
    print(f"Ausgeführter Code: '{text}'")
    print(f"Fehlertyp: {exc_type.__name__}")
    print(f"Fehlermeldung: {exc_value}")
`````

`````python
# Variante 6: Fehlerverfolgung
#
import traceback

def fehlerhafte_funktion():
    # Diese Zeile (Zeile 5) wird den Fehler verursachen
    ergebnis = 100 / 0
    return ergebnis

def aufrufende_funktion():
    print("Die aufrufende Funktion startet...")
    # Hier wird die Funktion aufgerufen, in der der Fehler auftritt
    fehlerhafte_funktion()
    print("Dieser Text wird nie erreicht.")
`````

`````python
# Hauptprogramm
try:
    aufrufende_funktion()
except Exception:
    print("\nEin Fehler wurde abgefangen! Hier sind die Details:")
    # traceback.format_exc() zeigt den gesamten Weg zum Fehler
    error_details = traceback.format_exc()
    print(error_details)
`````

## Veränderlich vs Unveränderlich
---


In Python gibt es zwei Arten von Variablen: mutable (veränderliche) und immutable (unveränderliche).


+ Mutable (veränderlich): Objekte, deren Inhalt geändert werden kann, ohne dass eine neue Referenz (Speicheradresse) erstellt wird.
+ Immutable (unveränderlich): Objekte, deren Inhalt nach ihrer Erstellung nicht mehr geändert werden kann. Jede Änderung erzeugt ein neues Objekt (ggf. mit gleichem Variablennamen) mit einer neuen Referenz (Speicheradresse).

| Datentyp        | Immutable/Mutable   | Beispiel       |
|-----------------|---------------------|----------------|
| `int`           | Immutable           | `x = 1`        |
| `float`         | Immutable           | `x = 1.0`      |
| `str`           | Immutable           | `x = "Hello"`  |
| `tuple`         | Immutable           | `x = (1, 2)`   |
| `list`          | Mutable             | `x = [1, 2]`   |
| `set`           | Mutable             | `x = {1, 2}`   |
| `dict`          | Mutable             | `x = {'a': 1}` |



**Warum ist das wichtig?**

+ Speicherverwaltung: Bei unveränderlichen Objekten kann Python den Speicher effizienter nutzen, indem es dasselbe Objekt an mehreren Stellen verwendet, da sicher ist, dass es nicht verändert wird.
+ Fehlervermeidung: Wenn man weiß, welche Objekte veränderlich sind, kann man ungewollte Änderungen in Funktionen vermeiden, die Parameter modifizieren.
+ Performance: Bei veränderlichen Objekten kann der Speicher an Ort und Stelle geändert werden, was in manchen Situationen effizienter ist.

`````python
# --- Beispiel immutable
a = 3
b = a  # b verweist auf Objekt a mit Inhalt 3

print("a:", a, id(a))
print("b:", b, id(b))
`````

`````python
a = 4  # a verweist nun auf eine neues Objekt mit neuem Inhalt 4, b verweist weiterhin auf 3

print("a:", a, id(a))
print("b:", b, id(b))
`````

`````python
# --- Beispiel mutable
a = [3]
b = a        # b verweist jetzt auch auf dieselbe Liste wie a

print("a:", a, id(a))
print("b:", b, id(b))
`````

`````python
a[0] = 4    # Eine Änderung an der Liste a wirkt auch bei b; die Verweise ändern sich nicht

print("a:", a, id(a))
print("b:", b, id(b))
`````

## High-Order-Functions (ext)
---

Eine  High-Order-Function (Höherwertige Funktion) in Python ist eine Funktion, die entweder eine andere Funktion als Argument akzeptiert, eine Funktion als Rückgabewert liefert oder beides tut.

### Generator







In der Informatik ist ein Generator eine Routine, mit der das Iterationsverhalten einer Schleife gesteuert werden kann. Alle Generatoren sind auch Iteratoren. Man kann sich einen Generator also vorstellen wie eine Funktion, die mit jedem Aufruf nur das nächste Element einer Liste zurückgibt. Kann der Generator keinen weiteren Rückgabewert liefern, wird ein StopIteration-Error ausgelöst.

Der Vorteil von Generatoren gegenüber Listen liegt darin, dass auch bei sehr großen Datenmengen kein großer Speicherbedarf nötig ist, da immer nur das jeweils aktuell zurückgegebene Objekt im Speicher existiert. Beispielsweise handelt es sich auch bei File-Objekten um Generatoren, die beim Aufruf von ``next(object)`` jeweils die nächste Zeile der geöffneten ausgeben. Auf diese Weise kann der Inhalt einer (beliebig) großen Datei beispielsweise mittels einer for-Schleife abgearbeitet werden, ohne dass der Inhalt der Datei auf einmal eingelesen und als Liste gespeichert werden muss.

Quelle: https://www.grund-wissen.de/informatik/python/klassen-und-objektorientierung.html

**Beispiel: Liste mit den Quadraten der Werte 1 - 9**

`````python
#
# zählergesteuerte Schleife
#
mylist = []
for i in range(1,10):
    mylist.append(i**2)

print("vollständige Liste: ", mylist)
`````

`````python
#
# Beispiel: Generator-Ausdruck
#
mygen  = (i**2 for i in range(1,10))
print ("Generator: ", next(mygen))
print ("Generator: ", next(mygen))
print ("Generator: ", next(mygen))
`````

`````python
#
# Beispiel: Generator-Funktion
#
def my_generator(n):
    for i in range(n):
        yield i * 2      # yield ist ein Schlüsselwort in Python, mit dem eine Funktion zu einem Generator
`````

`````python
# Generator mit einer for-Schleife aufrufen
for value in my_generator(5):
    print(value)
`````

`````python
gen = my_generator(5)
print(next(gen))  # Gibt den ersten Wert aus
print(next(gen))  # Gibt den zweiten Wert aus
print(next(gen))  # Gibt den dritten Wert aus
# usw.
`````

### Decorator



Ein Dekorator ist im Wesentlichen eine Funktion, die eine andere Funktion als Argument nimmt und eine neue Funktion zurückgibt. Diese neue Funktion enthält die **ursprüngliche** Funktionalität sowie die **hinzugefügte** Funktionalität des Dekorators. Die grundlegende Syntax eines Dekorators in Python verwendet das `@`-Symbol, gefolgt vom Namen des Dekorators, direkt über der zu dekorierenden Funktion.

**Vorteile von Dekoratoren:**

+ Reduzierung von Boilerplate-Code: Dekoratoren helfen dabei, wiederkehrende Muster und redundanten Code zu vermeiden, indem sie häufig benötigte Funktionalitäten kapseln und wiederverwendbar machen.
+ Trennung von Anliegen: Sie ermöglichen eine klare Trennung von Kernfunktionalität und zusätzlichen Aspekten wie Logging, Caching oder Authentifizierung.
+ Verbesserte Lesbarkeit und Wartbarkeit: Durch die Verwendung von Dekoratoren wird der Code übersichtlicher und leichter zu warten, da die zusätzliche Funktionalität klar und explizit an einer Stelle definiert ist.

`````python
#
# Beispiel: Dekorator-Funktion, die die Ausführungszeit einer Funktion misst
#
import time

def timing_decorator(func):
    # Wrapper-Funktion, die die eigentliche Funktion umschließt
    def wrapper(*args, **kwargs):
        start_time = time.time()        # Startzeit aufzeichnen
        result = func(*args, **kwargs)  # Die dekorierte Funktion mit allen Argumenten aufrufen
        end_time = time.time()          # Endzeit aufzeichnen
        # Ausgabe der Ausführungszeit der Funktion
        print(f"Funktion {func.__name__} ausgeführt in: {end_time - start_time:.7f} Sekunden.")
        return result                   # Ergebnis der Funktion zurückgeben

    return wrapper                      # Wrapper-Funktion zurückgeben
`````

`````python
# Anwendung des Dekorators auf die Addiere-Funktion
@timing_decorator
def Addiere(*args):
    return sum(args)

# Die dekorierte Addiere-Funktion aufrufen und das Ergebnis ausgeben
print(Addiere(1, 2, 3, 4, 5, 6, 7))
`````

`````python
# Anwendung des Dekorators auf die Fibonacci-Funktion
# mathematischen Zahlenreihe, bei der jede Zahl die Summe der beiden vorherigen ist.
#
@timing_decorator
def fibonacci(n):
    # Initialisiere die ersten beiden Terme
    a, b = 0, 1
    # Initialisiere die Liste mit dem ersten Term
    sequence = [a]
    # Führe die Iteration n-1 mal durch, da der erste Term bereits in der Liste ist
    for i in range(n-1):
        # Aktualisiere a und b auf die nächsten zwei Terme
        a, b = b, a + b
        # Füge den neuen Term der Liste hinzu
        sequence.append(a)
    return sequence

# Die dekorierte Fibonacci-Funktion aufrufen und das Ergebnis ausgeben
print(fibonacci(100)[:10], " ...")
`````

### Context Manger


Ein Context Manager in Python ist ein Objekt, das dazu dient, einen Kontext für die Ausführung eines Codeblocks zu definieren und sicherzustellen, dass Ressourcen ordnungsgemäß aufgebaut und wieder freigegeben werden. Die häufigste Verwendung von Context Managern ist die with-Anweisung, die oft für das sichere Öffnen und Schließen von Dateien verwendet wird



Die Python ````with```` Anweisung erstellt einen Laufzeitkontext , mit dem man eine Gruppe von Anweisungen unter der Kontrolle eines Kontextmanagers ausführen können.

Grundsätzlicher Aufbau:
````
    with expression as target_var:
         do_something(target_var)

````

`````python
#
# Beispiel 3: Klassenbasierte Implementierung
#
class RessourceManager():
    # --- Instanzmethode ---
    def __enter__(self):
        # Ressource aufbauen
        print("Ressource wird aufgebaut.")
        self.ressource = "Beispielressource"
        return self.ressource  # Dieser Wert wird an den 'as' Teil der 'with'-Anweisung übergeben

    def __exit__(self, exc_type, exc_value, traceback):
        # Ressource freigeben
        print("Ressource wird freigegeben.")
        # Hier könnten Sie Code zum Freigeben oder Schließen der Ressource hinzufügen
        # Zum Beispiel: self.ressource.close() wenn es sich um eine schließbare Ressource handelt

# Den Context Manager verwenden
with RessourceManager() as ressource:
    print(f"Verwende {ressource}.")
`````

**Beispiel: Funktion**

Die Fibonacci-Folge ist die unendliche Folge natürlicher Zahlen, die (ursprünglich) mit zweimal der Zahl 1 beginnt oder (häufig, in moderner Schreibweise) zusätzlich mit einer führenden Zahl 0 versehen ist. Im Anschluss ergibt jeweils die Summe zweier aufeinanderfolgender Zahlen die unmittelbar danach folgende Zahl.

`````python
# Beispiel: iterative Funktion
def fib_i(n):
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return a

# Beispiel: rekursive Funktion
def fib_r(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib_r(n-1) + fib_r(n-2)

# Beispiel: rekursive Funktion mit Memo-Dict
memo = {0:0, 1:1}
def fib_m(n):

    if not n in memo:
        memo[n] = fib_m(n-1) + fib_m(n-2)
    return memo[n]
`````

`````python
import time

n = 30
fib_i_start = time.time(); fib_i(n); fib_i_dauer = time.time() - fib_i_start
fib_r_start = time.time(); fib_r(n); fib_r_dauer = time.time() - fib_r_start
fib_m_start = time.time(); fib_m(n); fib_m_dauer = time.time() - fib_m_start

print(f"iterativ : {fib_i_dauer:.6f} Sek")
print(f"rekursiv : {fib_r_dauer:.6f} Sek")
print(f"rekursiv+: {fib_m_dauer:.6f} Sek")
`````

## Concurrent Programming
---



Concurrent Computing ist eine Form des Computings, bei der mehrere Berechnungen gleichzeitigausgeführt werden - während überlappender Zeiträume - anstatt sequenziell -wobei eine abgeschlossen wird, bevor die nächste beginnt.

### Sequentiell



Einfach zu verstehen und zu implementieren, aber möglicherweise nicht effizient für Aufgaben, die parallelisiert werden können.

`````python
import time

def rechenintensive_aufgabe(name):
    print(f"{name} startet")
    time.sleep(1)  # Simuliere eine rechenintensive Aufgabe
    print(f"{name} ist fertig")

start_time = time.time()

rechenintensive_aufgabe("Aufgabe 1")
rechenintensive_aufgabe("Aufgabe 2")

print(f"Sequenzielle Ausführungsdauer: {(time.time() - start_time):.5f} Sekunden")
`````

### Threading



Gut für **I/O-intensive Aufgaben**. In Python jedoch durch den Global Interpreter Lock (GIL) in Bezug auf CPU-intensive Aufgaben eingeschränkt.


`````python
import threading
import time

def rechenintensive_aufgabe(name):
    print(f"{name} startet")
    time.sleep(1)
    print(f"{name} ist fertig")

start_time = time.time()

thread1 = threading.Thread(target=rechenintensive_aufgabe, args=("Aufgabe 1",))
thread2 = threading.Thread(target=rechenintensive_aufgabe, args=("Aufgabe 2",))

thread1.start()
thread2.start()

thread1.join()
thread2.join()

print(f"Threading Ausführungsdauer: {(time.time() - start_time):.5f} Sekunden")
`````

### Multiprocessing



Umgeht den GIL und eignet sich gut für **CPU-intensive Aufgaben**, hat aber einen höheren Overhead für Prozesserstellung und -kommunikation.


`````python
from multiprocessing import Process
import time

def rechenintensive_aufgabe(name):
    print(f"{name} startet")
    time.sleep(1)  # Simuliert eine rechenintensive Aufgabe
    print(f"{name} ist fertig")

start_time = time.time()

# Erstelle Prozesse für parallele Ausführung
prozesse = [
    Process(target=rechenintensive_aufgabe, args=("Aufgabe 1",)),
    Process(target=rechenintensive_aufgabe, args=("Aufgabe 2",))
]

# Starte alle Prozesse
for p in prozesse:
    p.start()

# Warte, bis alle Prozesse abgeschlossen sind
for p in prozesse:
    p.join()

print(f"Parallelverarbeitung Ausführungsdauer: {time.time() - start_time} Sekunden")
`````

### Parallelprocessing


 Vereinfacht die Parallelisierung von Aufgaben, besonders nützlich in wissenschaftlichen Berechnungen und beim maschinellen Lernen, wo Aufgaben oft unabhängig sind.

`````python
from joblib import Parallel, delayed
import time

def rechenintensive_aufgabe(name):
    print(f"{name} startet")
    time.sleep(1)
    print(f"{name} ist fertig")

start_time = time.time()

# Verwende joblib, um die Funktionen parallel auszuführen
Parallel(n_jobs=2)(delayed(rechenintensive_aufgabe)(f"Aufgabe {i+1}") for i in range(2))

print(f"Parallele Ausführungsdauer mit joblib: {(time.time() - start_time):.5f} Sekunden")
`````

## Datenklasse vs normale Klasse
---

Das dataclasses-Modul (eingeführt in Python 3.7, 2018) ist ein Standardmodul, das das Schreiben von Klassen vereinfacht, die hauptsächlich Daten speichern — also genau die Art von Klassen, die sonst viel Boilerplate-Code enthalten.

Der Auslöser war ein konkreter Bedarf in der Python-Community: Datenklassen tauchen in fast jedem Projekt auf, und die manuelle Implementierung war fehleranfällig und redundant. PEP 557 (2017) formalisierte das Muster, das viele Entwickler bereits mit Drittanbieter-Bibliotheken nachgebaut hatten.
`@dataclass` war also keine neue Idee, sondern die offizielle Antwort auf ein bekanntes, weit verbreitetes Problem – und übernahm dabei die bewährten Konzepte direkt in die Standardbibliothek.

| Variante | Syntax | Automatische Methoden | Veränderbar | Geeignet für |
|---|---|---|---|---|
| **Normale Klasse** | `class Auto:` | keine | ✅ | komplexe Logik, volle Kontrolle |
| **Datenklasse** | `@dataclass` | `__init__`, `__repr__`, `__eq__` | ✅ | Klassen, die hauptsächlich Daten speichern |
| **Frozen Datenklasse** | `@dataclass(frozen=True)` | + `__hash__` | ❌ | unveränderliche Datenobjekte |
| **NamedTuple** | `class Auto(NamedTuple):` | `__init__`, `__repr__`, `__eq__` | ❌ | einfache, unveränderliche Datensätze |
| **TypedDict** | `class Auto(TypedDict):` | keine | ✅ | Typ-Annotationen für Dictionaries |
| **Enum** | `class Farbe(Enum):` | `__repr__`, `__eq__` | ❌ | feste Wertemenge (z.B. Zustände, Farben) |
| **Abstract Class** | `class Auto(ABC):` | keine | ✅ | Vorlagen/Interfaces, nicht direkt instanziierbar |


**Zusammengefasst:**

Datenklassen (`@dataclass`) sind die **häufigste und modernste Wahl** für Datenobjekte in Python und Standard dank Flexibilität und weniger Boilerplate.

## frozen=True – Unveränderliche Instanzen

Mit `@dataclass(frozen=True)` werden Instanzen unveränderlich (immutable): Attributwerte können nach der Erstellung nicht mehr geändert werden. Python wirft dann einen `FrozenInstanceError`. Geeignet für Werte-Objekte wie Koordinaten oder Konfigurationen.

`````python
from dataclasses import dataclass

@dataclass(frozen=True)
class Punkt:
    # --- Instanzattribute ---
    x: float
    y: float

p = Punkt(1.0, 2.0)
print(p)          # Punkt(x=1.0, y=2.0)

# p.x = 5.0       # → FrozenInstanceError: cannot assign to field 'x'
`````

## order=True – Automatische Vergleichsoperatoren

Mit `@dataclass(order=True)` werden `<`, `>`, `<=`, `>=` automatisch generiert. Der Vergleich erfolgt feldweise in der Reihenfolge der Felder. Damit lassen sich Instanzen direkt sortieren.

`````python
from dataclasses import dataclass

@dataclass(order=True)
class Produkt:
    # --- Instanzattribute ---
    preis: float
    name: str

p1 = Produkt(19.99, "Buch")
p2 = Produkt( 9.99, "Stift")
p3 = Produkt(49.99, "Headset")

print(sorted([p1, p2, p3]))   # sortiert nach preis (erstes Feld)
print(p1 > p2)                # True
`````

## Trigger auf Attribute

> Mit `@dataclass` und `@property` lassen sich Attribute mit kontrolliertem Zugriff (Getter/Setter) definieren. `@property` ergänzt `@dataclass` – es ersetzt keinen Mechanismus, sondern fügt Validierungslogik beim Lesen/Schreiben hinzu.

`````python
from dataclasses import dataclass, field

@dataclass
class Troll:
    # --- Instanzattribute ---
    lebenspunkte: int                              # Startwert
    _lp: int = field(init=False, repr=False)       # internes Attribut

    # --- Initialisierung (Instanzmethode, automatisch aufgerufen) ---
    def __post_init__(self):
        self._lp = self.lebenspunkte

    # --- Instanzmethode ---
    def check_lebenspunkte(self):
        if self._lp < 0:
            print(f"Lebenspunkte: {self._lp}! Der Troll ist tot.")

    @property                       # Holen des Wertes (Getter)
    def punkte(self):
        print(f"1. Getter: {self._lp}")
        return self._lp

    @punkte.setter                  # Setzen des Wertes (Setter)
    def punkte(self, wert):
        print(f"2. Setter: {wert}")
        self._lp = wert
        self.check_lebenspunkte()
`````

Ein ```Dekorator``` ist eine Zusatz, der einem Attribut oder einer Funktion neue Fähigkeiten oder Verhaltensweisen hinzufügt.

Der ```@property```-Dekorator erlaubt es, den Wert eines Attributs zu bekommen, um den Wert zu überprüfen oder zu modifizieren.

Der ```@lebenspunkte.setter```-Dekorator erlaubt es, dass eine Methode aufgerufen wird, der es erlaubt, den Wert eines Attributs zu setzen und/oder dabei eine zusätzliche Logik ausführt, wie z.B. den Wert zu validieren oder zu modifizieren, bevor er tatsächlich gesetzt wird.

Der ```@property```-Dekorator macht lebenspunkte zu einer berechneten Eigenschaft, die beim Lesen den Wert von _lebenspunkte zurückgibt. Der ```@lebenspunkte.setter```-Dekorator ermöglicht es dann, eine Methode zu definieren, die aufgerufen wird, wenn der Wert von lebenspunkte geändert wird. Diese Methode erlaubt es, zusätzliche Logik auszuführen, wie in diesem Fall die Überprüfung der Lebenspunkte.

`````python
Grindel = Troll(25)
`````

`````python
# Trigger auslösen
Grindel.punkte -= 20
`````

## Abstrakte Klassen
---


Abstrakte Klassen dienen dazu, unvollständige Klassen zu definieren, die als Basis für andere Klassen dienen, die bestimmte Methoden implementieren müssen.

In Python werden abstrakte Klassen mit dem Modul abc (Abstract Base Classes) implementiert. Dies ermöglicht es, Klassen zu definieren, die nicht direkt instanziiert werden können, und Methoden, die in abgeleiteten Klassen implementiert werden müssen.

Eine abstrakte Klasse Auto definiert die Struktur für abgeleitete Klassen, indem sie sicherstellt, dass bestimmte Methoden in jeder abgeleiteten Klasse implementiert werden.

`````python
from abc import ABC, abstractmethod
from dataclasses import dataclass

@dataclass
class Auto(ABC):
    # --- Instanzattribute ---
    marke: str

    @abstractmethod
    # --- Instanzmethode ---
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

@dataclass
class ElektroAuto(Auto):
    # --- Instanzmethode ---
    def start(self):
        print(f"{self.marke} Elektroauto startet leise.")

    def stop(self):
        print(f"{self.marke} Elektroauto hält an.")

@dataclass
class VerbrennungsAuto(Auto):
    # --- Instanzmethode ---
    def start(self):
        print(f"{self.marke} Verbrennungsauto startet mit Motorgeräusch.")

    def stop(self):
        print(f"{self.marke} Verbrennungsauto hält an.")
`````

`````python
meinElektroAuto = ElektroAuto("Tesla")
meinElektroAuto.start()
meinElektroAuto.stop()
`````

`````python
meinVerbrennungsAuto = VerbrennungsAuto("Volkswagen")
meinVerbrennungsAuto.start()
meinVerbrennungsAuto.stop()
`````

## Komposition und Aggregation
---

Beides sind Konzepte, die verwendet werden, um komplexe Typen durch Kombination von Objekten einfacherer Typen zu bilden.


### Komposition
---

Bei der Komposition ist ein Objekt Eigentümer eines anderen Objekts. Wenn das Eigentümerobjekt zerstört wird, wird auch das Eigentumsobjekt zerstört. Im Automobilkontext könnte das Auto aus mehreren Komponenten bestehen, wie z.B. einem Motor. Wenn das Auto zerstört wird, existiert der Motor nicht mehr unabhängig.

`````python
from dataclasses import dataclass

@dataclass
class Motor:
    # --- Instanzattribute ---
    typ: str

    # --- Instanzmethode ---
    def zeige_details(self):
        print(f"Motor Typ: {self.typ}")

@dataclass
class Auto:
    # --- Instanzattribute ---
    marke: str
    motor: Motor   # Komposition: Motor gehört zum Auto

    # --- Instanzmethode ---
    def zeige_details(self):
        print(f"Auto Marke: {self.marke}")
        self.motor.zeige_details()
`````

`````python
meinAuto = Auto("Tesla", Motor("Elektro"))
`````

`````python
# Ausgabe
meinAuto.zeige_details()
`````

### Aggregation
---


Bei der Aggregation hat ein Objekt ein oder mehrere andere Objekte, aber diese Objekte können ohne das Eigentümerobjekt existieren. Zum Beispiel könnte ein Auto mehrere Fahrer haben, aber die Fahrer existieren unabhängig vom Auto.

`````python
from dataclasses import dataclass, field

@dataclass
class Fahrer:
    # --- Instanzattribute ---
    name: str

    # --- Instanzmethode ---
    def zeige_details(self):
        print(f"Fahrer Name: {self.name}")

@dataclass
class Auto:
    # --- Instanzattribute ---
    marke: str
    fahrer: list = field(default_factory=list)   # Aggregation: Fahrer existieren unabhängig

    # --- Instanzmethode ---
    def fahrer_hinzufuegen(self, fahrer):
        self.fahrer.append(fahrer)

    def zeige_details(self):
        print(f"Auto Marke: {self.marke}")
        for f in self.fahrer:
            f.zeige_details()
`````

`````python
# Beispiel Nutzung
fahrer1 = Fahrer("Max")
fahrer2 = Fahrer("Anna")
meinAuto = Auto("Volkswagen")
meinAuto.fahrer_hinzufuegen(fahrer1)
meinAuto.fahrer_hinzufuegen(fahrer2)
`````

`````python
# Ausgabe
meinAuto.zeige_details()
`````

Magic commands sind spezielle Befehle, die über die normale Python-Syntax hinausgehen und für eine Vielzahl von Aufgaben nützlich sind, von der Arbeitsflusssteuerung bis hin zur Performance-Analyse. Diese Befehle sind in zwei Hauptkategorien unterteilt: "Line Magic" und "Cell Magic".

+ Line Magic-Befehle werden mit einem einzelnen Prozentzeichen (%) eingeleitet und wirken sich nur auf die Zeile aus, in der sie stehen. Sie können als Ausdruck verwendet werden, und ihr Rückgabewert kann einer Variablen zugewiesen werden. Ein Beispiel für einen Line Magic-Befehl ist %timeit, der die Ausführungszeit eines einzelnen Python-Ausdrucks oder einer Anweisung misst und dabei mehrere Durchläufe zur Genauigkeitssteigerung durchführt


+ Cell Magic-Befehle hingegen werden mit zwei Prozentzeichen (%%) eingeleitet und beziehen sich auf die gesamte Zelle, in der sie stehen. Sie können mehrere Zeilen Code umfassen und sogar Eingaben verarbeiten, die kein gültiger Python-Code sind. Ein Beispiel für einen Cell Magic-Befehl ist %%timeit, der ähnlich wie sein Line Magic-Pendant funktioniert, aber die Ausführungszeit des gesamten Zellinhalts misst.

| Symbol | Description   |
|--------|---------------|
| `!`    | Shell Command |
| `%`    | Line Magic    |
| `%%`   | Cell Magic    |

`````python
from babel.dates import format_datetime
import pytz
from datetime import datetime

# Aktuelles Datum und Uhrzeit in der "Europe/Berlin"-Zeitzone
berlin_tz = pytz.timezone('Europe/Berlin')
dt_berlin = datetime.now(berlin_tz)

# Aktuelles Datum und Uhrzeit in der "America/New_York"-Zeitzone
ny_tz = pytz.timezone('America/New_York')
dt_ny = datetime.now(ny_tz)

# Formatierung mit automatischer Zeitzonen-Berücksichtigung
print(format_datetime(dt_berlin, format='full', locale='de_DE'))

print(format_datetime(dt_ny, format='full', locale='en_US'))
`````

Arrow ist eine populäre Python-Bibliothek, die das Arbeiten mit Datums- und Zeitstempeln erleichtert. Sie bietet viele Formatierungsoptionen, ähnlich wie das eingebaute datetime-Modul, bietet jedoch zusätzliche Funktionalität und ist benutzerfreundlicher.

Hier sind einige der gängigsten Zeitformatierungs-Token, die Sie mit Arrow verwenden können:

- YYYY: Vierstelliges Jahr (z.B. "2023")
- YY: Zweistelliges Jahr (z.B. "23")
- MM: Monat als Zahl mit führender Null (z.B. "07" für Juli)
- MMM: Abgekürzter Monatsname (z.B. "Jul")
- MMMM: Vollständiger Monatsname (z.B. "July")
- DD: Tag des Monats mit führender Null
- ddd: Abgekürzter Wochentag (z.B. "Sun")
- dddd: Vollständiger Wochentag (z.B. "Sunday")
- HH: Stunden im 24-Stunden-Format mit führender- Null
- hh: Stunden im 12-Stunden-Format mit führender Null
- mm: Minuten mit führender Null
- ss: Sekunden mit führender Null
- A: AM/PM
- ZZZ: Zeitzonenname (z.B. "UTC", "EST")
- ZZ: Zeitzone im Format "+HH:MM" oder "-HH:MM"
