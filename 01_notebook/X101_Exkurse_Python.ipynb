#  IPython

IPython erweitert die Standard-Python-Shell um zahlreiche Funktionen, die die interaktive Arbeit mit Python effizienter und benutzerfreundlicher machen. Diese zusätzlichen Funktionen sind besonders nützlich für wissenschaftliches Rechnen, Datenanalyse und explorative Programmierung.


[Doku](https://ipython.readthedocs.io/en/stable/index.html)

**Mit Perplexity.ai erstellt:**

IPython bietet im Vergleich zur Standard-Python-Shell eine Vielzahl zusätzlicher Funktionen, die die interaktive Arbeit mit Python erheblich verbessern. Hier sind einige der wichtigsten zusätzlichen Funktionen von IPython:

1. Magische Befehle (Magic Commands):
IPython führt spezielle Befehle ein, die mit einem Prozentzeichen (`%`) beginnen und eine Vielzahl von Aufgaben erleichtern. Beispiele sind `%timeit` zur Messung der Ausführungszeit von Code-Snippets und `%run`, um Python-Skripte auszuführen[1][2][3].

2. Erweiterte Tab-Vervollständigung:
IPython bietet eine erweiterte Tab-Vervollständigung, die nicht nur Variablennamen, sondern auch Methoden und Attribute von Objekten vervollständigt. Dies erleichtert das Schreiben von Code und das Erkunden von Bibliotheken erheblich[2][6].

3. Objekt-Introspektion:
Mit IPython können Benutzer detaillierte Informationen über Objekte abrufen, einschließlich Docstrings, Funktionsdefinitionen und Quellcode. Dies wird durch einfache Befehle wie `?` und `??` ermöglicht[2][7].

4. System-Shell-Zugriff:
IPython ermöglicht den direkten Zugriff auf System-Shell-Befehle innerhalb der IPython-Sitzung. Befehle wie `!ls` oder `!pwd` können verwendet werden, um Systembefehle auszuführen, ohne die IPython-Shell zu verlassen[5][7].

5. Befehlshistorie und Wiederverwendung:
IPython speichert die Befehlshistorie über Sitzungen hinweg, was es einfach macht, frühere Befehle zu durchsuchen und erneut auszuführen. Befehle wie `%history` und `%recall` sind dafür besonders nützlich[1][5].

6. Integration mit Jupyter Notebooks:
IPython ist eng mit Jupyter Notebooks integriert, einer browserbasierten Umgebung, die es ermöglicht, Code, Visualisierungen und narrative Texte in einem einzigen Dokument zu kombinieren. Dies ist besonders nützlich für Datenanalyse und wissenschaftliche Kommunikation[1][3][4].

7. Unterstützung für paralleles Rechnen:
IPython bietet eine Architektur für interaktives paralleles Rechnen, die es ermöglicht, Aufgaben auf mehrere Prozessoren zu verteilen. Dies ist besonders nützlich für rechenintensive Anwendungen[2].

8. Erweiterte Fehlerbehandlung und Debugging:
IPython bietet erweiterte Funktionen zur Fehlerbehandlung und Debugging, einschließlich farbiger Tracebacks und einfacher Integration mit Debugging-Tools wie `pdb`[8].

9. Flexibles GUI-Event-Loop-Management:
IPython unterstützt die interaktive Steuerung von GUI-Toolkits wie GTK, Qt und WX, was die Entwicklung und das Testen von grafischen Anwendungen erleichtert[8].

**Fazit:**  
IPython erweitert die Standard-Python-Shell um zahlreiche Funktionen, die die interaktive Arbeit mit Python effizienter und benutzerfreundlicher machen. Diese zusätzlichen Funktionen sind besonders nützlich für wissenschaftliches Rechnen, Datenanalyse und explorative Programmierung.

## Veränderlich vs Unveränderlich
---


In Python gibt es zwei Arten von Variablen: mutable (veränderliche) und immutable (unveränderliche).


+ Mutable (veränderlich): Objekte, deren Inhalt geändert werden kann, ohne dass eine neue Referenz (Speicheradresse) erstellt wird.
+ Immutable (unveränderlich): Objekte, deren Inhalt nach ihrer Erstellung nicht mehr geändert werden kann. Jede Änderung erzeugt ein neues Objekt (ggf. mit gleichem Variablennamen) mit einer neuen Referenz (Speicheradresse).

| Datentyp        | Immutable/Mutable   | Beispiel       |
|-----------------|---------------------|----------------|
| `int`           | Immutable           | `x = 1`        |
| `float`         | Immutable           | `x = 1.0`      |
| `str`           | Immutable           | `x = "Hello"`  |
| `tuple`         | Immutable           | `x = (1, 2)`   |
| `list`          | Mutable             | `x = [1, 2]`   |
| `set`           | Mutable             | `x = {1, 2}`   |
| `dict`          | Mutable             | `x = {'a': 1}` |



**Warum ist das wichtig?**

+ Speicherverwaltung: Bei unveränderlichen Objekten kann Python den Speicher effizienter nutzen, indem es dasselbe Objekt an mehreren Stellen verwendet, da sicher ist, dass es nicht verändert wird.
+ Fehlervermeidung: Wenn man weiß, welche Objekte veränderlich sind, kann man ungewollte Änderungen in Funktionen vermeiden, die Parameter modifizieren.
+ Performance: Bei veränderlichen Objekten kann der Speicher an Ort und Stelle geändert werden, was in manchen Situationen effizienter ist.

`````python
# --- Beispiel immutable
a = 3
b = a  # b verweist auf Objekt a mit Inhalt 3

print("a:", a, id(a))
print("b:", b, id(b))
`````

`````python
a = 4  # a verweist nun auf eine neues Objekt mit neuem Inhalt 4, b verweist weiterhin auf 3

print("a:", a, id(a))
print("b:", b, id(b))
`````

`````python
# --- Beispiel mutable
a = [3]
b = a        # b verweist jetzt auch auf dieselbe Liste wie a

print("a:", a, id(a))
print("b:", b, id(b))
`````

`````python
a[0] = 4    # Eine Änderung an der Liste a wirkt auch bei b; die Verweise ändern sich nicht

print("a:", a, id(a))
print("b:", b, id(b))
`````

Eine  High-Order-Function (Höherwertige Funktion) in Python ist eine Funktion, die entweder eine andere Funktion als Argument akzeptiert, eine Funktion als Rückgabewert liefert oder beides tut.

### Generator







In der Informatik ist ein Generator eine Routine, mit der das Iterationsverhalten einer Schleife gesteuert werden kann. Alle Generatoren sind auch Iteratoren. Man kann sich einen Generator also vorstellen wie eine Funktion, die mit jedem Aufruf nur das nächste Element einer Liste zurückgibt. Kann der Generator keinen weiteren Rückgabewert liefern, wird ein StopIteration-Error ausgelöst.

Der Vorteil von Generatoren gegenüber Listen liegt darin, dass auch bei sehr großen Datenmengen kein großer Speicherbedarf nötig ist, da immer nur das jeweils aktuell zurückgegebene Objekt im Speicher existiert. Beispielsweise handelt es sich auch bei File-Objekten um Generatoren, die beim Aufruf von ``next(object)`` jeweils die nächste Zeile der geöffneten ausgeben. Auf diese Weise kann der Inhalt einer (beliebig) großen Datei beispielsweise mittels einer for-Schleife abgearbeitet werden, ohne dass der Inhalt der Datei auf einmal eingelesen und als Liste gespeichert werden muss.

Quelle: https://www.grund-wissen.de/informatik/python/klassen-und-objektorientierung.html

**Beispiel: Liste mit den Quadraten der Werte 1 - 9**

`````python
#
# zählergesteuerte Schleife
#
mylist = []
for i in range(1,10):
    mylist.append(i**2)

print("vollständige Liste: ", mylist)
`````

`````python
#
# Beispiel: Generator-Ausdruck
#
mygen  = (i**2 for i in range(1,10))
print ("Generator: ", next(mygen))
print ("Generator: ", next(mygen))
print ("Generator: ", next(mygen))
`````

`````python
#
# Beispiel: Generator-Funktion
#
def my_generator(n):
    for i in range(n):
        yield i * 2      # yield ist ein Schlüsselwort in Python, mit dem eine Funktion zu einem Generator
`````

`````python
# Generator mit einer for-Schleife aufrufen
for value in my_generator(5):
    print(value)
`````

`````python
gen = my_generator(5)
print(next(gen))  # Gibt den ersten Wert aus
print(next(gen))  # Gibt den zweiten Wert aus
print(next(gen))  # Gibt den dritten Wert aus
# usw.
`````

### Decorator



Ein Dekorator ist im Wesentlichen eine Funktion, die eine andere Funktion als Argument nimmt und eine neue Funktion zurückgibt. Diese neue Funktion enthält die **ursprüngliche** Funktionalität sowie die **hinzugefügte** Funktionalität des Dekorators. Die grundlegende Syntax eines Dekorators in Python verwendet das `@`-Symbol, gefolgt vom Namen des Dekorators, direkt über der zu dekorierenden Funktion.

**Vorteile von Dekoratoren:**

+ Reduzierung von Boilerplate-Code: Dekoratoren helfen dabei, wiederkehrende Muster und redundanten Code zu vermeiden, indem sie häufig benötigte Funktionalitäten kapseln und wiederverwendbar machen.
+ Trennung von Anliegen: Sie ermöglichen eine klare Trennung von Kernfunktionalität und zusätzlichen Aspekten wie Logging, Caching oder Authentifizierung.
+ Verbesserte Lesbarkeit und Wartbarkeit: Durch die Verwendung von Dekoratoren wird der Code übersichtlicher und leichter zu warten, da die zusätzliche Funktionalität klar und explizit an einer Stelle definiert ist.

`````python
#
# Beispiel: Dekorator-Funktion, die die Ausführungszeit einer Funktion misst
#
import time

def timing_decorator(func):
    # Wrapper-Funktion, die die eigentliche Funktion umschließt
    def wrapper(*args, **kwargs):
        start_time = time.time()        # Startzeit aufzeichnen
        result = func(*args, **kwargs)  # Die dekorierte Funktion mit allen Argumenten aufrufen
        end_time = time.time()          # Endzeit aufzeichnen
        # Ausgabe der Ausführungszeit der Funktion
        print(f"Funktion {func.__name__} ausgeführt in: {end_time - start_time:.7f} Sekunden.")
        return result                   # Ergebnis der Funktion zurückgeben

    return wrapper                      # Wrapper-Funktion zurückgeben
`````

`````python
# Anwendung des Dekorators auf die Addiere-Funktion
@timing_decorator
def Addiere(*args):
    return sum(args)

# Die dekorierte Addiere-Funktion aufrufen und das Ergebnis ausgeben
print(Addiere(1, 2, 3, 4, 5, 6, 7))
`````

`````python
# Anwendung des Dekorators auf die Fibonacci-Funktion
# mathematischen Zahlenreihe, bei der jede Zahl die Summe der beiden vorherigen ist.
#
@timing_decorator
def fibonacci(n):
    # Initialisiere die ersten beiden Terme
    a, b = 0, 1
    # Initialisiere die Liste mit dem ersten Term
    sequence = [a]
    # Führe die Iteration n-1 mal durch, da der erste Term bereits in der Liste ist
    for i in range(n-1):
        # Aktualisiere a und b auf die nächsten zwei Terme
        a, b = b, a + b
        # Füge den neuen Term der Liste hinzu
        sequence.append(a)
    return sequence

# Die dekorierte Fibonacci-Funktion aufrufen und das Ergebnis ausgeben
print(fibonacci(100)[:10], " ...")
`````

### Context Manger


Ein Context Manager in Python ist ein Objekt, das dazu dient, einen Kontext für die Ausführung eines Codeblocks zu definieren und sicherzustellen, dass Ressourcen ordnungsgemäß aufgebaut und wieder freigegeben werden. Die häufigste Verwendung von Context Managern ist die with-Anweisung, die oft für das sichere Öffnen und Schließen von Dateien verwendet wird



Die Python ````with```` Anweisung erstellt einen Laufzeitkontext , mit dem man eine Gruppe von Anweisungen unter der Kontrolle eines Kontextmanagers ausführen können.

Grundsätzlicher Aufbau:
````
    with expression as target_var:
         do_something(target_var)

````

`````python
#
# Beispiel 1: Dateizugriff, öffnen/schliessen
#
with open("/content/drive/MyDrive/Python_Intro/02daten/datei_in.txt", mode="r") as file:
    inhalt = file.read()
    print(inhalt)
`````

`````python
#
# Beispiel 2: Durchsuchen von Verzeichnissen
#
import os

with os.scandir(".") as entries:
     for entry in entries:
         print(entry.name, "->", entry.stat().st_size, "bytes")
`````

`````python
#
# Beispiel 3: Klassenbasierte Implementierung
#
class RessourceManager():
    # --- Instanzmethode ---
    def __enter__(self):
        # Ressource aufbauen
        print("Ressource wird aufgebaut.")
        self.ressource = "Beispielressource"
        return self.ressource  # Dieser Wert wird an den 'as' Teil der 'with'-Anweisung übergeben

    def __exit__(self, exc_type, exc_value, traceback):
        # Ressource freigeben
        print("Ressource wird freigegeben.")
        # Hier könnten Sie Code zum Freigeben oder Schließen der Ressource hinzufügen
        # Zum Beispiel: self.ressource.close() wenn es sich um eine schließbare Ressource handelt

# Den Context Manager verwenden
with RessourceManager() as ressource:
    print(f"Verwende {ressource}.")
`````

`````python
#
# Beispiel 4: Funktionsbasierte Implementierung mit dem contextlib-Modul:
#
from contextlib import contextmanager

@contextmanager
def simple_context_manager():
    print("Ressourcen werden aufgebaut.")
    try:
        yield "Hallo aus dem Context Manager!"
    finally:
        print("Ressourcen werden freigegeben.")

# Den Context Manager verwenden
with simple_context_manager() as value:
    print(value)
`````

## Data Structures & Algorithms
---


**Beispiel: Funktion**

Die Fibonacci-Folge ist die unendliche Folge natürlicher Zahlen, die (ursprünglich) mit zweimal der Zahl 1 beginnt oder (häufig, in moderner Schreibweise) zusätzlich mit einer führenden Zahl 0 versehen ist. Im Anschluss ergibt jeweils die Summe zweier aufeinanderfolgender Zahlen die unmittelbar danach folgende Zahl.

`````python
# Beispiel: iterative Funktion
def fib_i(n):
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return a

# Beispiel: rekursive Funktion
def fib_r(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib_r(n-1) + fib_r(n-2)

# Beispiel: rekursive Funktion mit Memo-Dict
memo = {0:0, 1:1}
def fib_m(n):

    if not n in memo:
        memo[n] = fib_m(n-1) + fib_m(n-2)
    return memo[n]
`````

`````python
import time

n = 30
fib_i_start = time.time(); fib_i(n); fib_i_dauer = time.time() - fib_i_start
fib_r_start = time.time(); fib_r(n); fib_r_dauer = time.time() - fib_r_start
fib_m_start = time.time(); fib_m(n); fib_m_dauer = time.time() - fib_m_start

print(f"iterativ : {fib_i_dauer:.6f} Sek")
print(f"rekursiv : {fib_r_dauer:.6f} Sek")
print(f"rekursiv+: {fib_m_dauer:.6f} Sek")
`````

## Concurrent Programming
---



Concurrent Computing ist eine Form des Computings, bei der mehrere Berechnungen gleichzeitigausgeführt werden - während überlappender Zeiträume - anstatt sequenziell -wobei eine abgeschlossen wird, bevor die nächste beginnt.

Quelle: Wikipedia

### Sequentiell



Einfach zu verstehen und zu implementieren, aber möglicherweise nicht effizient für Aufgaben, die parallelisiert werden können.

`````python
import time

def rechenintensive_aufgabe(name):
    print(f"{name} startet")
    time.sleep(1)  # Simuliere eine rechenintensive Aufgabe
    print(f"{name} ist fertig")

start_time = time.time()

rechenintensive_aufgabe("Aufgabe 1")
rechenintensive_aufgabe("Aufgabe 2")

print(f"Sequenzielle Ausführungsdauer: {(time.time() - start_time):.5f} Sekunden")
`````

### Threading



Gut für **I/O-intensive Aufgaben**. In Python jedoch durch den Global Interpreter Lock (GIL) in Bezug auf CPU-intensive Aufgaben eingeschränkt.


`````python
import threading
import time

def rechenintensive_aufgabe(name):
    print(f"{name} startet")
    time.sleep(1)
    print(f"{name} ist fertig")

start_time = time.time()

thread1 = threading.Thread(target=rechenintensive_aufgabe, args=("Aufgabe 1",))
thread2 = threading.Thread(target=rechenintensive_aufgabe, args=("Aufgabe 2",))

thread1.start()
thread2.start()

thread1.join()
thread2.join()

print(f"Threading Ausführungsdauer: {(time.time() - start_time):.5f} Sekunden")
`````

### Multiprocessing



Umgeht den GIL und eignet sich gut für **CPU-intensive Aufgaben**, hat aber einen höheren Overhead für Prozesserstellung und -kommunikation.


`````python
from multiprocessing import Process
import time

def rechenintensive_aufgabe(name):
    print(f"{name} startet")
    time.sleep(1)  # Simuliert eine rechenintensive Aufgabe
    print(f"{name} ist fertig")

start_time = time.time()

# Erstelle Prozesse für parallele Ausführung
prozesse = [
    Process(target=rechenintensive_aufgabe, args=("Aufgabe 1",)),
    Process(target=rechenintensive_aufgabe, args=("Aufgabe 2",))
]

# Starte alle Prozesse
for p in prozesse:
    p.start()

# Warte, bis alle Prozesse abgeschlossen sind
for p in prozesse:
    p.join()

print(f"Parallelverarbeitung Ausführungsdauer: {time.time() - start_time} Sekunden")
`````

### Parallelprocessing


 Vereinfacht die Parallelisierung von Aufgaben, besonders nützlich in wissenschaftlichen Berechnungen und beim maschinellen Lernen, wo Aufgaben oft unabhängig sind.

`````python
from joblib import Parallel, delayed
import time

def rechenintensive_aufgabe(name):
    print(f"{name} startet")
    time.sleep(1)
    print(f"{name} ist fertig")

start_time = time.time()

# Verwende joblib, um die Funktionen parallel auszuführen
Parallel(n_jobs=2)(delayed(rechenintensive_aufgabe)(f"Aufgabe {i+1}") for i in range(2))

print(f"Parallele Ausführungsdauer mit joblib: {(time.time() - start_time):.5f} Sekunden")
`````

## Datenklasse vs normale Klasse
---

Das dataclasses-Modul (eingeführt in Python 3.7, 2018) ist ein Standardmodul, das das Schreiben von Klassen vereinfacht, die hauptsächlich Daten speichern — also genau die Art von Klassen, die sonst viel Boilerplate-Code enthalten.

Der Auslöser war ein konkreter Bedarf in der Python-Community: Datenklassen tauchen in fast jedem Projekt auf, und die manuelle Implementierung war fehleranfällig und redundant. PEP 557 (2017) formalisierte das Muster, das viele Entwickler bereits mit Drittanbieter-Bibliotheken nachgebaut hatten.
`@dataclass` war also keine neue Idee, sondern die offizielle Antwort auf ein bekanntes, weit verbreitetes Problem – und übernahm dabei die bewährten Konzepte direkt in die Standardbibliothek.

| Variante | Syntax | Automatische Methoden | Veränderbar | Geeignet für |
|---|---|---|---|---|
| **Normale Klasse** | `class Auto:` | keine | ✅ | komplexe Logik, volle Kontrolle |
| **Datenklasse** | `@dataclass` | `__init__`, `__repr__`, `__eq__` | ✅ | Klassen, die hauptsächlich Daten speichern |
| **Frozen Datenklasse** | `@dataclass(frozen=True)` | + `__hash__` | ❌ | unveränderliche Datenobjekte |
| **NamedTuple** | `class Auto(NamedTuple):` | `__init__`, `__repr__`, `__eq__` | ❌ | einfache, unveränderliche Datensätze |
| **TypedDict** | `class Auto(TypedDict):` | keine | ✅ | Typ-Annotationen für Dictionaries |
| **Enum** | `class Farbe(Enum):` | `__repr__`, `__eq__` | ❌ | feste Wertemenge (z.B. Zustände, Farben) |
| **Abstract Class** | `class Auto(ABC):` | keine | ✅ | Vorlagen/Interfaces, nicht direkt instanziierbar |


**Zusammengefasst:**

Datenklassen (`@dataclass`) sind die **häufigste und modernste Wahl** für Datenobjekte in Python und Standard dank Flexibilität und weniger Boilerplate.

## frozen=True – Unveränderliche Instanzen

Mit `@dataclass(frozen=True)` werden Instanzen unveränderlich (immutable): Attributwerte können nach der Erstellung nicht mehr geändert werden. Python wirft dann einen `FrozenInstanceError`. Geeignet für Werte-Objekte wie Koordinaten oder Konfigurationen.

`````python
from dataclasses import dataclass

@dataclass(frozen=True)
class Punkt:
    # --- Instanzattribute ---
    x: float
    y: float

p = Punkt(1.0, 2.0)
print(p)          # Punkt(x=1.0, y=2.0)

# p.x = 5.0       # → FrozenInstanceError: cannot assign to field 'x'
`````

## order=True – Automatische Vergleichsoperatoren

Mit `@dataclass(order=True)` werden `<`, `>`, `<=`, `>=` automatisch generiert. Der Vergleich erfolgt feldweise in der Reihenfolge der Felder. Damit lassen sich Instanzen direkt sortieren.

`````python
from dataclasses import dataclass

@dataclass(order=True)
class Produkt:
    # --- Instanzattribute ---
    preis: float
    name: str

p1 = Produkt(19.99, "Buch")
p2 = Produkt( 9.99, "Stift")
p3 = Produkt(49.99, "Headset")

print(sorted([p1, p2, p3]))   # sortiert nach preis (erstes Feld)
print(p1 > p2)                # True
`````

## Trigger auf Attribute

> Mit `@dataclass` und `@property` lassen sich Attribute mit kontrolliertem Zugriff (Getter/Setter) definieren. `@property` ergänzt `@dataclass` – es ersetzt keinen Mechanismus, sondern fügt Validierungslogik beim Lesen/Schreiben hinzu.

`````python
from dataclasses import dataclass, field

@dataclass
class Troll:
    # --- Instanzattribute ---
    lebenspunkte: int                              # Startwert
    _lp: int = field(init=False, repr=False)       # internes Attribut

    # --- Initialisierung (Instanzmethode, automatisch aufgerufen) ---
    def __post_init__(self):
        self._lp = self.lebenspunkte

    # --- Instanzmethode ---
    def check_lebenspunkte(self):
        if self._lp < 0:
            print(f"Lebenspunkte: {self._lp}! Der Troll ist tot.")

    @property                       # Holen des Wertes (Getter)
    def punkte(self):
        print(f"1. Getter: {self._lp}")
        return self._lp

    @punkte.setter                  # Setzen des Wertes (Setter)
    def punkte(self, wert):
        print(f"2. Setter: {wert}")
        self._lp = wert
        self.check_lebenspunkte()
`````

Ein ```Dekorator``` ist eine Zusatz, der einem Attribut oder einer Funktion neue Fähigkeiten oder Verhaltensweisen hinzufügt.

Der ```@property```-Dekorator erlaubt es, den Wert eines Attributs zu bekommen, um den Wert zu überprüfen oder zu modifizieren.

Der ```@lebenspunkte.setter```-Dekorator erlaubt es, dass eine Methode aufgerufen wird, der es erlaubt, den Wert eines Attributs zu setzen und/oder dabei eine zusätzliche Logik ausführt, wie z.B. den Wert zu validieren oder zu modifizieren, bevor er tatsächlich gesetzt wird.

Der ```@property```-Dekorator macht lebenspunkte zu einer berechneten Eigenschaft, die beim Lesen den Wert von _lebenspunkte zurückgibt. Der ```@lebenspunkte.setter```-Dekorator ermöglicht es dann, eine Methode zu definieren, die aufgerufen wird, wenn der Wert von lebenspunkte geändert wird. Diese Methode erlaubt es, zusätzliche Logik auszuführen, wie in diesem Fall die Überprüfung der Lebenspunkte.

`````python
Grindel = Troll(25)
`````

`````python
# Trigger auslösen
Grindel.punkte -= 20
`````

## Abstrakte Klassen
---


Abstrakte Klassen dienen dazu, unvollständige Klassen zu definieren, die als Basis für andere Klassen dienen, die bestimmte Methoden implementieren müssen.

In Python werden abstrakte Klassen mit dem Modul abc (Abstract Base Classes) implementiert. Dies ermöglicht es, Klassen zu definieren, die nicht direkt instanziiert werden können, und Methoden, die in abgeleiteten Klassen implementiert werden müssen.

Eine abstrakte Klasse Auto definiert die Struktur für abgeleitete Klassen, indem sie sicherstellt, dass bestimmte Methoden in jeder abgeleiteten Klasse implementiert werden.

`````python
from abc import ABC, abstractmethod
from dataclasses import dataclass

@dataclass
class Auto(ABC):
    # --- Instanzattribute ---
    marke: str

    @abstractmethod
    # --- Instanzmethode ---
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

@dataclass
class ElektroAuto(Auto):
    # --- Instanzmethode ---
    def start(self):
        print(f"{self.marke} Elektroauto startet leise.")

    def stop(self):
        print(f"{self.marke} Elektroauto hält an.")

@dataclass
class VerbrennungsAuto(Auto):
    # --- Instanzmethode ---
    def start(self):
        print(f"{self.marke} Verbrennungsauto startet mit Motorgeräusch.")

    def stop(self):
        print(f"{self.marke} Verbrennungsauto hält an.")
`````

`````python
meinElektroAuto = ElektroAuto("Tesla")
meinElektroAuto.start()
meinElektroAuto.stop()
`````

`````python
meinVerbrennungsAuto = VerbrennungsAuto("Volkswagen")
meinVerbrennungsAuto.start()
meinVerbrennungsAuto.stop()
`````

## Komposition und Aggregation
---

Beides sind Konzepte, die verwendet werden, um komplexe Typen durch Kombination von Objekten einfacherer Typen zu bilden.


### Komposition
---

Bei der Komposition ist ein Objekt Eigentümer eines anderen Objekts. Wenn das Eigentümerobjekt zerstört wird, wird auch das Eigentumsobjekt zerstört. Im Automobilkontext könnte das Auto aus mehreren Komponenten bestehen, wie z.B. einem Motor. Wenn das Auto zerstört wird, existiert der Motor nicht mehr unabhängig.

`````python
from dataclasses import dataclass

@dataclass
class Motor:
    # --- Instanzattribute ---
    typ: str

    # --- Instanzmethode ---
    def zeige_details(self):
        print(f"Motor Typ: {self.typ}")

@dataclass
class Auto:
    # --- Instanzattribute ---
    marke: str
    motor: Motor   # Komposition: Motor gehört zum Auto

    # --- Instanzmethode ---
    def zeige_details(self):
        print(f"Auto Marke: {self.marke}")
        self.motor.zeige_details()
`````

`````python
meinAuto = Auto("Tesla", Motor("Elektro"))
`````

`````python
# Ausgabe
meinAuto.zeige_details()
`````

### Aggregation
---


Bei der Aggregation hat ein Objekt ein oder mehrere andere Objekte, aber diese Objekte können ohne das Eigentümerobjekt existieren. Zum Beispiel könnte ein Auto mehrere Fahrer haben, aber die Fahrer existieren unabhängig vom Auto.

`````python
from dataclasses import dataclass, field

@dataclass
class Fahrer:
    # --- Instanzattribute ---
    name: str

    # --- Instanzmethode ---
    def zeige_details(self):
        print(f"Fahrer Name: {self.name}")

@dataclass
class Auto:
    # --- Instanzattribute ---
    marke: str
    fahrer: list = field(default_factory=list)   # Aggregation: Fahrer existieren unabhängig

    # --- Instanzmethode ---
    def fahrer_hinzufuegen(self, fahrer):
        self.fahrer.append(fahrer)

    def zeige_details(self):
        print(f"Auto Marke: {self.marke}")
        for f in self.fahrer:
            f.zeige_details()
`````

`````python
# Beispiel Nutzung
fahrer1 = Fahrer("Max")
fahrer2 = Fahrer("Anna")
meinAuto = Auto("Volkswagen")
meinAuto.fahrer_hinzufuegen(fahrer1)
meinAuto.fahrer_hinzufuegen(fahrer2)
`````

`````python
# Ausgabe
meinAuto.zeige_details()
`````

`````python


from dataclasses import dataclass
from typing import ClassVar

@dataclass
class Ritter:
    # --- Klassenattribut ---
    _alle_instanzen: ClassVar[list] = []  # ClassVar: @dataclass überspringt dieses Feld (kein __init__-Parameter)

    # --- Instanzattribute ---
    name: str
    waffe: str

    # --- Initialisierung (Instanzmethode, automatisch aufgerufen) ---
    def __post_init__(self):
        self.__class__._alle_instanzen.append(self)
        print(f"Instanz von {self.name} wurde in Klasse {self.__class__.__name__} erstellt.")

    # --- Instanzmethode ---
    def loeschen(self):
        self.__class__._alle_instanzen.remove(self)
        print(f"Instanz von {self.name} wurde gelöscht.")

    # --- Klassenmethode ---
    @classmethod
    def get_instanzen(cls):
        return list(cls._alle_instanzen)
`````

##  Klassenattribute & Klassenmethoden
---



+ **ClassVar**: ist eine Typ-Annotation aus `typing`, die ein Attribut als Klassenattribut kennzeichnet. `@dataclass` überspringt `ClassVar`-Felder bei der automatischen Generierung von Attributen.

+ **@classmethod**: Ist ein Dekorator für Methoden, die an die Klasse statt an eine Instanz gebunden sind. Die Methode erhält als ersten Parameter die Klasse (`cls`) und kann damit auf Klassenattribute zugreifen oder alternative Konstruktoren implementieren.


`````python
from dataclasses import dataclass
from typing import ClassVar


@dataclass
class Auto:
    # --- Klassenattribut ---
    _alle_instanzen: ClassVar[list] = []  # ClassVar: @dataclass überspringt dieses Feld (kein __init__-Parameter)

    # --- Instanzattribute ---
    marke: str
    modell: str

    # --- Initialisierung (Instanzmethode, automatisch aufgerufen) ---
    def __post_init__(self):
        self.__class__._alle_instanzen.append(self)

    # --- Instanzmethode ---
    def loeschen(self):
        self.__class__._alle_instanzen.remove(self)

    # --- Klassenmethode ---
    @classmethod
    def get_instanzen(cls):
        return list(cls._alle_instanzen)
`````

Eine **API** (Application Programming Interface) ist eine **Schnittstelle**, über die Programme **Daten von anderen Systemen abrufen oder senden** – meist über das Internet.

Stell dir eine API wie einen **Kellner im Restaurant** vor:

Die meisten modernen Web-APIs – wie die Länder-API – basieren auf dem Prinzip **REST**:

> **REST = Representational State Transfer**

REST ist ein **einfacher Architekturstil** für APIs, der auf **HTTP** basiert, also genau dem Protokoll, das auch Webseiten nutzen.

Typische Eigenschaften von REST-APIs:
- Jede **Ressource** (z. B. ein Land) hat eine eigene **URL**
- Die Kommunikation erfolgt über **HTTP-Methoden** wie `GET`, `POST`, `DELETE`
- Die Antwort kommt meist im **JSON-Format**
- Die Aufrufe sind **stateless** (jede Anfrage ist unabhängig)

#     Shell & Magic Commands
---

Shell-Befehle sind Anweisungen, die in der Kommandozeile (CLI) eines Betriebssystems verwendet werden, um mit dem System zu interagieren und Aufgaben auszuführen.

Magic commands sind spezielle Befehle, die über die normale Python-Syntax hinausgehen und für eine Vielzahl von Aufgaben nützlich sind, von der Arbeitsflusssteuerung bis hin zur Performance-Analyse. Diese Befehle sind in zwei Hauptkategorien unterteilt: "Line Magic" und "Cell Magic".

+ Line Magic-Befehle werden mit einem einzelnen Prozentzeichen (%) eingeleitet und wirken sich nur auf die Zeile aus, in der sie stehen. Sie können als Ausdruck verwendet werden, und ihr Rückgabewert kann einer Variablen zugewiesen werden. Ein Beispiel für einen Line Magic-Befehl ist %timeit, der die Ausführungszeit eines einzelnen Python-Ausdrucks oder einer Anweisung misst und dabei mehrere Durchläufe zur Genauigkeitssteigerung durchführt


+ Cell Magic-Befehle hingegen werden mit zwei Prozentzeichen (%%) eingeleitet und beziehen sich auf die gesamte Zelle, in der sie stehen. Sie können mehrere Zeilen Code umfassen und sogar Eingaben verarbeiten, die kein gültiger Python-Code sind. Ein Beispiel für einen Cell Magic-Befehl ist %%timeit, der ähnlich wie sein Line Magic-Pendant funktioniert, aber die Ausführungszeit des gesamten Zellinhalts misst.

| Symbol | Description   |
|--------|---------------|
| `!`    | Shell Command |
| `%`    | Line Magic    |
| `%%`   | Cell Magic    |

#  Bibliothekssystem 🪂
---

In dieser Aufgabe entwickeln Sie ein einfaches Bibliothekssystem, in dem Mitglieder Bücher ausleihen und zurückgeben können. Dabei sollen zwei Klassen erstellt werden: `Buch` und `Mitglied`. Die beiden Klassen interagieren miteinander, indem Mitglieder Bücher ausleihen und zurückgeben, was den Status der Bücher beeinflusst.

**Anforderungen:**
+ Klasse `Buch`:
Erstellen Sie eine Klasse `Buch`, die folgende Attribute besitzt:
  + `titel`: Der Titel des Buches.
  + `autor`: Der Autor des Buches.
  + `ist_ausgeliehen`: Ein boolscher Wert (False), der angibt, ob das Buch ausgeliehen ist.

+ Klasse `Mitglied`:
Erstellen Sie eine Klasse `Mitglied`, die folgende Attribute besitzt:
  + `name`: Der Name des Mitglieds.
  + `ausgeliehene_buecher`: Eine Liste der ausgeliehenen Bücher (leer am Anfang).

Methoden für `Mitglied`:
  + `buch_ausleihen(buch)`: Verleiht ein Buch an das Mitglied.
  + `buch_zurueckgeben(buch)`: Gibt ein Buch zurück.

> 💡 **Tipp:** Die Klasse `Buch` eignet sich gut für `@dataclass` (reine Datenhaltung). Für `Mitglied` mit einer Liste als Attribut verwenden Sie `field(default_factory=list)`.

Schwierigkeit: ⭐⭐⭐

`````python
#
#
#
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>

from dataclasses import dataclass, field

@dataclass
class Buch:
    # --- Instanzattribute ---
    titel: str
    autor: str
    ist_ausgeliehen: bool = False   #  dataclass erlaubt direkte Default-Werte

@dataclass
class Mitglied:
    # --- Instanzattribute ---
    name: str
    ausgeliehene_buecher: list = field(default_factory=list)  # veränderliches Default

    # --- Instanzmethode ---
    def buch_ausleihen(self, buch):
        if not buch.ist_ausgeliehen:
            buch.ist_ausgeliehen = True
            self.ausgeliehene_buecher.append(buch)
            print(f"{self.name} hat '{buch.titel}' ausgeliehen.")
        else:
            print(f"'{buch.titel}' ist bereits ausgeliehen.")

    def buch_zurueckgeben(self, buch):
        if buch in self.ausgeliehene_buecher:
            buch.ist_ausgeliehen = False
            self.ausgeliehene_buecher.remove(buch)
            print(f"{self.name} hat '{buch.titel}' zurückgegeben.")
        else:
            print(f"{self.name} hat '{buch.titel}' nicht ausgeliehen.")

# Beispiel
buch1 = Buch("Der kleine Prinz", "Antoine de Saint-Exupéry")
mitglied1 = Mitglied("Anna")

mitglied1.buch_ausleihen(buch1)
mitglied1.buch_zurueckgeben(buch1)
`````

**Erweiterung:**

+ Maximale Anzahl an Büchern, die ein Mitglied ausleihen kann: Ein Mitglied kann nur eine bestimmte Anzahl von Büchern gleichzeitig ausleihen.
+ Anzeige aller verfügbaren Bücher in der Bibliothek: Die Bibliothek zeigt eine Liste der Bücher an, die noch nicht ausgeliehen wurden.

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
from dataclasses import dataclass, field

@dataclass
class Buch:
    # --- Instanzattribute ---
    titel: str
    autor: str
    ist_ausgeliehen: bool = False

@dataclass
class Mitglied:
    # --- Instanzattribute ---
    name: str
    maximal_ausleihen: int = 3
    ausgeliehene_buecher: list = field(default_factory=list)

    # --- Instanzmethode ---
    def buch_ausleihen(self, buch):
        if len(self.ausgeliehene_buecher) < self.maximal_ausleihen:
            if not buch.ist_ausgeliehen:
                buch.ist_ausgeliehen = True
                self.ausgeliehene_buecher.append(buch)
                print(f"{self.name} hat '{buch.titel}' ausgeliehen.")
            else:
                print(f"'{buch.titel}' ist bereits ausgeliehen.")
        else:
            print(f"{self.name} kann keine weiteren Bücher ausleihen, da das Limit erreicht ist.")

    def buch_zurueckgeben(self, buch):
        if buch in self.ausgeliehene_buecher:
            buch.ist_ausgeliehen = False
            self.ausgeliehene_buecher.remove(buch)
            print(f"{self.name} hat '{buch.titel}' zurückgegeben.")
        else:
            print(f"{self.name} hat '{buch.titel}' nicht ausgeliehen.")

    def anzahl_ausgeliehene_buecher(self):
        return len(self.ausgeliehene_buecher)

@dataclass
class Bibliothek:
    # --- Instanzattribute ---
    name: str
    buecher: list = field(default_factory=list)

    # --- Instanzmethode ---
    def buch_hinzufuegen(self, buch):
        self.buecher.append(buch)
        print(f"'{buch.titel}' wurde zur Bibliothek hinzugefügt.")

    def verfuegbare_buecher_anzeigen(self):
        verfuegbare = [b for b in self.buecher if not b.ist_ausgeliehen]
        if verfuegbare:
            print("Verfügbare Bücher:")
            for b in verfuegbare:
                print(f"- {b.titel} von {b.autor}")
        else:
            print("Keine verfügbaren Bücher.")

# Beispiel
buch1 = Buch("Der kleine Prinz", "Antoine de Saint-Exupéry")
buch2 = Buch("1984", "George Orwell")
buch3 = Buch("Der Prozess", "Franz Kafka")
buch4 = Buch("Die Verwandlung", "Franz Kafka")

anna = Mitglied("Anna")
bibliothek = Bibliothek("Stadtbibliothek")

bibliothek.buch_hinzufuegen(buch1)
bibliothek.buch_hinzufuegen(buch2)
bibliothek.buch_hinzufuegen(buch3)
bibliothek.buch_hinzufuegen(buch4)

bibliothek.verfuegbare_buecher_anzeigen()

anna.buch_ausleihen(buch1)
anna.buch_ausleihen(buch2)
anna.buch_ausleihen(buch3)
anna.buch_ausleihen(buch4)   # Limit erreicht

anna.buch_zurueckgeben(buch1)
anna.buch_ausleihen(buch4)

bibliothek.verfuegbare_buecher_anzeigen()
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
def ist_palindrom(s: str) -> bool:
    # Entfernen von Satzzeichen, Leerzeichen und Umwandlung in Kleinbuchstaben
    gereinigter_string = ''.join(e for e in s if e.isalnum()).lower()
    # Vergleich des Strings mit seiner Umkehrung
    return gereinigter_string == gereinigter_string[::-1]

# Test der Funktion
test_string = input("Palindrom? Wort oder Satz angeben: ")
if ist_palindrom(test_string):
    print(f"'{test_string}' ist ein Palindrom.")
else:
    print(f"'{test_string}' ist kein Palindrom.")
`````

#  Blockchain
---

Schreiben Sie ein Programm das einen Blockchain erstellt. Ein Blockchain ist eine Kette von Datenblöcken, bei denen jeder Block einen Verweis auf den vorherigen Block enthält (meist in Form eines Hashes).

Es soll diese Transaktionen fälschungssicher speichern, also so, dass sie nachträglich möglichst schwer zu manipulieren sind.

Der "Block" enthält jedweils die nachfolgenden Daten:

+ Index
+ alter hash (des vorigen Blocks)
+ Zeitstempel
+ Daten
+ neuer hash

Verwende die Hash-Funktion `sha256`.



```
import hashlib
.encode('utf-8')  # Methode konvertiert String in einen Byte-Array, wird von hashlib erwartet
hashlib.sha256()  # Funktion erzeugt ein Hash-Objekt
.hexdigest()      # Methode konvertiert das Hash-Objekts in einen hexadezimalen String
```

3. Fügen Sie den fertigen Block an die Blockchain an.
4. Erzeugen Sie den zweiten bis x. Block.

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
# Hinweis: Blockchain-Implementierungen nutzen oft klassische Klassen mit __init__,
# da die Hash-Berechnung in __post_init__ oder __hash__ von @dataclass(frozen=True)
# kollidieren würde. Hier ein Beispiel mit traditioneller Syntax:
import time
from datetime import datetime
# --------------------------------------------------------
class Block:
    # --- Initialisierung (Instanzmethode, automatisch aufgerufen) ---
    def __init__(self, index, previous_hash, timestamp, data, hash):
        self.index = index
        self.previous_hash = previous_hash
        self.timestamp = timestamp
        self.data = data
        self.hash = hash

    # --- Darstellung (Instanzmethode, automatisch aufgerufen) ---
    def __str__(self):
        return (f"Index: {self.index}\n"
                f"Previous Hash: {self.previous_hash}\n"
                f"Timestamp: {self.timestamp}\n"
                f"Data: {self.data}\n"
                f"Hash: {self.hash}\n")
# --------------------------------------------------------
# Ermittle Hash-Objekt
def calculate_hash(index, previous_hash, timestamp, data):
    return hashlib.sha256(f'{index}{previous_hash}{timestamp}{data}'.encode('utf-8')).hexdigest()
# Erstelle Genesis-Block
def create_genesis_block():
    timestamp = datetime.now().strftime('%Y-%m-%d-%H-%M-%S-%f')
    return Block(0, '0', timestamp, 'Genesis Block', calculate_hash(0, '0', timestamp, 'Genesis Block'))
# Erstelle neuen Folgeblock
def create_new_block(prev_block, data):
    index = prev_block.index + 1
    timestamp = datetime.now().strftime('%Y-%m-%d-%H-%M-%S-%f')
    hash = calculate_hash(index, prev_block.hash, timestamp, data)
    return Block(index, prev_block.hash, timestamp, data, hash)
# --------------------------------------------------------
# Erstelle die Blockchain und füge den Genesis-Block hinzu
blockchain = [create_genesis_block()]
previous_block = blockchain[0]
# --------------------------------------------------------
# Füge 10 weitere Blöcke zur Blockchain hinzu
num_of_blocks_to_add = 10
# --------------------------------------------------------
for i in range(1, num_of_blocks_to_add + 1):
    new_data = f"Block #{i} data"
    new_block = create_new_block(previous_block, new_data)
    blockchain.append(new_block)
    previous_block = new_block
    print(f"Block #{new_block.index} has been added to the blockchain!")
    print(f"Hash: {new_block.hash}\n")
# --------------------------------------------------------
# Beispiele
print(blockchain[0])  # Genesis-Block
print(blockchain[5])  # 5. Block
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
#
# Version 1 - exklusive doc-strings
# Erstellt mit ChatGPT
#
# reading the file
with open(file_name) as fp:
    contents=fp.readlines()

# initialize two counter to check mismatch between "(" and ")"
open_bracket_counter=0
close_bracket_counter=0

# whenever an element deleted from the list length of the list will be decreased
decreasing_counter=0

for number in range(len(contents)):

    # checking if the line contains "#" or not
    if "#" in contents[number-decreasing_counter]:

        # delete the line if startswith "#"
        if contents[number-decreasing_counter].startswith("#"):
            contents.remove(contents[number-decreasing_counter])
            decreasing_counter+=1

        # delete the character after the "#"
        else:
            newline=""
            for character in contents[number-decreasing_counter]:
                if character=="(":
                    open_bracket_counter+=1
                    newline+=character
                elif character==")":
                    close_bracket_counter+=1
                    newline+=character
                elif character=="#" and open_bracket_counter==close_bracket_counter:
                    break
                else:
                    newline+=character
            contents.remove(contents[number-decreasing_counter])
            contents.insert(number-decreasing_counter,newline)

new_file_name = "new_" + file_name
# writing into a new file
with open(new_file_name,"w") as fp:
    fp.writelines(contents)
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
#
# Version 2 - inklusive doc-strings
# Erstellt mit ChatGPT
#
import io, tokenize, re
def remove_comments_and_docstrings(source):
    io_obj = io.StringIO(source)
    out = ""
    prev_toktype = tokenize.INDENT
    last_lineno = -1
    last_col = 0
    for tok in tokenize.generate_tokens(io_obj.readline):
        token_type = tok[0]
        token_string = tok[1]
        start_line, start_col = tok[2]
        end_line, end_col = tok[3]
        ltext = tok[4]
        if start_line > last_lineno:
            last_col = 0
        if start_col > last_col:
            out += (" " * (start_col - last_col))
        if token_type == tokenize.COMMENT:
            pass
        elif token_type == tokenize.STRING:
            if prev_toktype != tokenize.INDENT:
                if prev_toktype != tokenize.NEWLINE:
                    if start_col > 0:
                        out += token_string
        else:
            out += token_string
        prev_toktype = token_type
        last_col = end_col
        last_lineno = end_line
    out = '\n'.join(l for l in out.splitlines() if l.strip())
    return out


# reading the file
with open(file_name) as f_in:
  new_file_name = "new_" + file_name
  with open(new_file_name,"w") as f_out:
    f_out.writelines(remove_comments_and_docstrings(f_in.read()))
`````

`````python
#
`````
`````python
# Code durch ChatGPT erstellt
#
def einkommensteuer_berechnen(zu_versteuerndes_einkommen, tarif):
    if tarif == "splitting":
        zu_versteuerndes_einkommen /= 2

    # Berechnung der Einkommensteuer gemäß den Grenzen und Formeln
    if zu_versteuerndes_einkommen <= 10347:
        steuer = 0
    elif zu_versteuerndes_einkommen <= 62978:
        z = (zu_versteuerndes_einkommen - 10347) / 10000
        steuer = (995.21 * z + 1400) * z
    elif zu_versteuerndes_einkommen <= 300919:
        y = (zu_versteuerndes_einkommen - 62978) / 10000
        steuer = (216.16 * y + 2397) * y + 950
    else:
        steuer = 0.45 * zu_versteuerndes_einkommen - 17386

    if tarif == "splitting":
        steuer *= 2  # Steuer wird verdoppelt, da das Einkommen halbiert wurde

    return round(steuer, 2)  # Rundet die Steuer auf zwei Dezimalstellen
`````

## Reine Funktion
---


Eine reine Funktion (engl. pure function) ist ein Konzept aus der funktionalen Programmierung, das auch in anderen Programmierparadigmen verwendet werden kann.

+ Keine Seiteneffekte: Eine reine Funktion beeinflusst keine globalen Zustände oder externe Variablen. Ihre einzige Aufgabe besteht darin, basierend auf den Eingabewerten ein Ergebnis zurückzugeben.
+ Keine Ein-/Ausgaben: Eine reine Funktion darf keine Ein- oder Ausgaben (z. B. Konsolenausgaben oder das Schreiben/Lesen von Dateien) haben, da dies als Seiteneffekte gilt.
+ Determinismus: Für dieselben Eingabewerte liefert eine reine Funktion immer dasselbe Ergebnis, unabhängig davon, wie oft oder wo sie aufgerufen wird.

## Higher-Order-Functions (HOFs)

**Higher-Order-Functions** nehmen Funktionen als Argumente oder geben Funktionen zurück.
In Python: `map()`, `filter()`, `sorted()` mit `key=`.

| Funktion | Beschreibung |
|----------|--------------|
| `map(fn, iterable)` | wendet `fn` auf jedes Element an → Iterator |
| `filter(fn, iterable)` | behält Elemente, für die `fn` True zurückgibt → Iterator |
| `sorted(iterable, key=fn)` | sortiert mit benutzerdefinierter Sortierfunktion |

**map() — Funktion auf alle Elemente anwenden**

`````python
zahlen = [1, 2, 3, 4, 5]

# map: Funktion auf jedes Element anwenden
quadrate = list(map(lambda x: x**2, zahlen))
print(quadrate)   # → [1, 4, 9, 16, 25]

# Äquivalent mit List-Comprehension (moderner)
quadrate2 = [x**2 for x in zahlen]
print(quadrate2)  # → [1, 4, 9, 16, 25]
`````

**filter() — Elemente nach Bedingung filtern**

`````python
zahlen = [1, 2, 3, 4, 5, 6, 7, 8]

# filter: nur gerade Zahlen behalten
gerade = list(filter(lambda x: x % 2 == 0, zahlen))
print(gerade)   # → [2, 4, 6, 8]

# Äquivalent mit List-Comprehension (moderner)
gerade2 = [x for x in zahlen if x % 2 == 0]
print(gerade2)  # → [2, 4, 6, 8]
`````

**sorted() mit key-Funktion**

`````python
woerter = ['Banane', 'Apfel', 'Kirsche', 'Dattel']

# Nach Länge sortieren
nach_laenge = sorted(woerter, key=len)
print(nach_laenge)   # → ['Apfel', 'Dattel', 'Banane', 'Kirsche']

# Nach letztem Buchstaben sortieren
nach_letztem = sorted(woerter, key=lambda w: w[-1])
print(nach_letztem)
`````

## Iteratoren

Ein **Iterator** ist ein Objekt, das mit `next()` Element für Element durchläuft.
`iter()` erzeugt einen Iterator aus einem iterierbaren Objekt.

Die meisten Python-Iteratoren sind **lazy** (berechnen Werte erst bei Bedarf) — effizient bei großen Datenmengen.

`````python
# iter() und next()
zahlen = [10, 20, 30]
it = iter(zahlen)

print(next(it))  # → 10
print(next(it))  # → 20
print(next(it))  # → 30
# next(it) würde jetzt StopIteration auslösen
`````

`````python
# range() ist ein Iterator-ähnliches Objekt (lazy)
r = range(1_000_000)   # kein Speicher für 1 Mio Zahlen verbraucht
print(r[999_999])     # → 999999

# map/filter geben Iteratoren zurück
it = map(lambda x: x*2, [1, 2, 3])
print(next(it))  # → 2
`````

##    Rekursive Funktionen
---


Python akzeptiert auch Funktionsrekursion, was bedeutet, dass eine definierte Funktion sich selbst aufrufen kann.

Rekursion ist ein gängiges mathematisches und programmtechnisches Konzept. Das bedeutet, dass sich eine Funktion selbst aufruft. Dies hat den Vorteil, dass Sie Daten durchschleifen können, um zu einem Ergebnis zu gelangen.

Der Entwickler sollte mit der Rekursion sehr vorsichtig sein, da es ziemlich einfach sein kann, eine Funktion zu schreiben, die nie beendet wird oder die übermäßig viel Speicher oder Prozessorleistung verbraucht. Wenn sie jedoch richtig geschrieben ist, kann Rekursion ein sehr effizienter und mathematisch eleganter Ansatz für die Programmierung sein.



**Beispiel:** <br>

Die Fakultät ist in der Mathematik eine Funktion, die einer natürlichen Zahl das Produkt aller natürlichen Zahlen (ohne Null) kleiner und gleich dieser Zahl zuordnet. Sie wird durch ein dem Argument nachgestelltes Ausrufezeichen („!“) abgekürzt.

4! = 1 * 2 * 3 * 4

Quelle: Wikipedia


`````python
# Beispiel: Fakultät berechnen - iterativ
def iterative_fakultaet(n):
    result = 1
    for i in range(2,n+1):
        result *= i
    return result

iterative_fakultaet(5)
`````

`````python
# Beispiel: Fakultät berechnen - rekursiv
def fakultaet(n):
    if n == 1:
        return 1
    else:
        return n * fakultaet(n-1)

fakultaet(5)
`````

# Comprehensions
---

Comprehensions in Python bieten eine kompakte und expressive Syntax, um neue Kollektionen basierend auf existierenden Sequenzen oder Iterables zu erstellen. Sie erlauben es, Elemente zu filtern, zu transformieren und auf effiziente Weise in einer neuen Kollektion zusammenzufassen. Dieser Mechanismus kann für verschiedene Datentypen angewandt werden, einschließlich Listen (List Comprehensions), Mengen (Set Comprehensions), Wörterbücher (Dictionary Comprehensions) und Generatoren (Generator Expressions).

##    List Comprehensions
----

Verkürzte Möglichkeit Listen zu erstellen. Mit diesem eleganten Ansatz kann man die `for` Schleife in nur einer einzigen Codezeile programmieren.

**Genereller Aufbau:**  

list = [<font color='blue'>expression</font> <font color='red'> for item in iterable</font> <font color='green'> (if conditional)</font> ]




+ <font color='blue'>expression:</font> Variable, Formel, Funktion, Bedingungen
+ <font color='red'>for item in iterabel:</font> Iteration über ein Listelement
+ <font color='green'>if conditional: </font>Bedingung für Auswertung expression

`````python
# Bedingungen  `if-else` in expression
nums = [1, 2, 3, 4, 5]
new_nums = [x if x % 2 == 0 else -x for x in nums]

# Bedingungen `if-else` in expression
result = ["klein" if x < 3 else "groß" for x in range(5)]

# Bedingungen `if-else` in expression
liste = [3,4,5,'a',6]
result = [x**2 if isinstance(x, int) else x*3 for x in liste]         # isinstance: prüft, ob ein Objekt zu einer bestimmten Klasse oder einem bestimmten Datentyp gehört
`````
