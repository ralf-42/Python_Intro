+ Die durch 4 ganzzahlig teilbaren Jahre sind, abgesehen von den folgenden Ausnahmen, Schaltjahre.
+ Skularjahre, also die Jahre, die ein Jahrhundert abschlieen (z. B. 1800, 1900, 2100 und 2200), sind, abgesehen von der folgenden Ausnahme, keine Schaltjahre.
+ Die durch 400 ganzzahlig teilbaren Skularjahre, zum Beispiel das Jahr 2000, sind jedoch Schaltjahre.

Die Logik zur Berechnung eines Schaltjahres (Gregorianischer Kalender) finden Sie auf folgendem Link: https://de.wikipedia.org/wiki/Schaltjahr


Schwierigkeit: 

`````python
#
# Bitte hier Ihren Code eingeben
#
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lsungsvorschlag</font> </br></p>
jahr = int(input("Jahr: "))
schaltjahr = "kein"
if jahr % 4 == 0:
    schaltjahr = "ein"
    if jahr % 100 == 0:
        schaltjahr = "kein"
    if jahr % 400 == 0:
        schaltjahr = "ein"

print(f"Das Jahr {jahr} ist {schaltjahr} Schaltjahr")
`````

Erstelle ein Programm, das die Gltigkeit eines Passworts basierend auf mehreren Kriterien berprft. Das Programm soll eine Zeichenkette (das Passwort) als Eingabe erhalten und berprfen, ob das Passwort die folgenden Bedingungen erfllt:

+ Es muss mindestens 8 Zeichen lang sein.
+ Es muss mindestens einen Grobuchstaben enthalten.
+ Es muss mindestens einen Kleinbuchstaben enthalten.
+ Es muss mindestens eine Zahl enthalten.
+ Es muss mindestens ein Sonderzeichen aus dieser Menge enthalten: @#$%&


Wenn das Passwort alle Bedingungen erfllt, soll das Programm ausgeben: "Passwort ist gltig." Wenn das Passwort eine oder mehrere Bedingungen nicht erfllt, soll das Programm ausgeben: "Passwort ist ungltig."



Schwierigkeit: 

`````python
#
# Bitte hier Ihren Code eingeben
#
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lsungsvorschlag</font> </br></p>

# Starte die ereignisgesteuerte Schleife
while True:
    # Fordere den Benutzer auf, ein Passwort einzugeben
    BenutzerPasswort = input("Bitte geben Sie ein Passwort ein: ")

    # Initialisiere die Kriterien fr ein gltiges Passwort
    MinLaenge = 8
    HatGrossbuchstabe = False
    HatKleinbuchstabe = False
    HatZahl = False
    HatSonderzeichen = False
    Sonderzeichen = "_-"

    # berprfe die Lnge des Passworts
    if len(BenutzerPasswort) >= MinLaenge:
        # berprfe jedes Zeichen im Passwort
        for zeichen in BenutzerPasswort:
            if zeichen.isupper():
                HatGrossbuchstabe = True
            if zeichen.islower():
                HatKleinbuchstabe = True
            if zeichen.isdigit():
                HatZahl = True
            if zeichen in Sonderzeichen:
                HatSonderzeichen = True

    # berprfe, ob alle Kriterien erfllt sind
    if HatGrossbuchstabe and HatKleinbuchstabe and HatZahl and HatSonderzeichen:
        print("Passwort ist gltig.")
        break  # Beende die Schleife, wenn das Passwort gltig ist
    else:
        # Informiere den Benutzer, dass das Passwort ungltig ist und fordere zur erneuten Eingabe auf
        print("Passwort ist ungltig. Bitte versuchen Sie es erneut.")
`````

## 12.5 |  Vigenre-Code

Eine sehr viel sicherere Verschlsselungsmethode als der Csarcode ist der Vigenere-Code (s. wikipedia). Dabei wird ganz analog zum Csarcode verschlsselt, aber es werden statt eines festen ks verschiedene ks verwendet, die aus einem Schlsswelwort hervorgehen. Haben wir als Schlsselwort z.B.
`````python
#@title
#@markdown   <p><font size="4" color='blue'>  Lsungshilfe - bei Bedarf lesen</font> </br></p>
#
# itertools.cycle ist eine Funktion aus dem itertools-Modul in Python.
# Sie gibt einen unendlichen Iterator zurck, der ber die gegebenen Eingabewerte
# iteriert. Wenn das Ende der Eingabe erreicht ist, fngt sie wieder von vorne an.
# Das bedeutet, dass cycle die Eingabewerte endlos wiederholt.
# -----
from itertools import cycle
zahlen = cycle([1, 2, 3])
for i in range(10):
    print(next(zahlen))
# -----
import string
string.printable
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lsungsvorschlag</font> </br></p>
import string
from itertools import cycle

def vigenere_encode(text, key):
    return vigenere_cypher(text, key)

def vigenere_decode(text, key):
    return vigenere_cypher(text, key, mode="decode")

def vigenere_cypher(text, key, mode="encode"):
    result = []
    key_cycle = cycle(key)
    base = (string.printable) * 2

    for t in text:
        if t in base:
            k = next(key_cycle)
            t_index = base.index(t)
            k_index = base.index(k)

            if mode == "decode":
                new_index = (t_index - k_index)
            else:
                new_index = (t_index + k_index)

            result.append(base[new_index])
        else:
            result.append(t)
            next(key_cycle)

    return ''.join(result)

if __name__ == "__main__":
    text = input("Bitte geben Sie den Text ein: ")
    key = input("Bitte geben Sie den Schlssel ein: ")
    mode = input("Whlen Sie den Modus (encode/decode): ")

    if not key:
        print("Der Schlssel darf nicht leer sein!")
    else:
        match mode:
            case "encode":
                processed_text = vigenere_encode(text, key)
                print(f"Verschlsselter Text: {processed_text}")
            case "decode":
                processed_text = vigenere_decode(text, key)
                print(f"Entschlsselter Text: {processed_text}")
            case _:
                print("Ungltiger Modus!")
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lsungsvorschlag</font> </br></p>
import random

def computer_choice():
    return random.choice(['Schere', 'Stein', 'Papier'])

def player_choice():
    print("Whle eine Option:")
    print("1. Schere")
    print("2. Stein")
    print("3. Papier")
    choice = input("Deine Wahl (1/2/3): ")
    while choice not in ['1', '2', '3']:
        print("Ungltige Eingabe. Bitte versuche es erneut.")
        choice = input("Deine Wahl (1/2/3): ")
    return ['Schere', 'Stein', 'Papier'][int(choice) - 1]

def round_winner(player, computer):
    if player == computer:
        return 'Unentschieden'
    elif (player == 'Schere' and computer == 'Papier') or \
         (player == 'Papier' and computer == 'Stein') or \
         (player == 'Stein' and computer == 'Schere'):
        return 'Spieler'
    else:
        return 'Computer'

def overall_winner(player_score, computer_score):
    if player_score > computer_score:
        return "Du bist der Gesamtsieger!"
    elif computer_score > player_score:
        return "Computer ist der Gesamtsieger!"
    else:
        return "Das Spiel endet Unentschieden!"

def main():
    print("Schere, Stein, Papier Spiel!")

    player_score = 0
    computer_score = 0

    for i in range(3):
        print(f"\nRunde {i + 1} von 3:")

        player = player_choice()
        computer = computer_choice()
        print(f"\nDu hast {player} gewhlt.")
        print(f"Computer hat {computer} gewhlt.")

        winner = round_winner(player, computer)
        print(winner, "gewinnt diese Runde!")

        if winner == 'Spieler':
            player_score += 1
        elif winner == 'Computer':
            computer_score += 1

    print("\n" + overall_winner(player_score, computer_score))

if __name__ == "__main__":
    main()
`````

## 14.4 |  Rollenspiel
---

In dieser bungsaufgabe soll ein Rollenspiel zu den zwei Klassen Magier und Amazone/Schwertkmpfer erstellt werden.

**Vorschlag, gerne ndern:**

Ziel: Entwickeln Sie ein einfaches Rollenspiel, in dem ein Magier gegen eine Amazone/einen Schwertkmpfer kmpft.

**Magier:**

Attribute:
+ name (Name des Magiers)
+ lebenspunkte (startet mit 100)
+ mana (magische Energie, startet mit 50)

Methoden:
+ zaubern(gegner): Wenn genug Mana vorhanden ist, versucht der Magier, den Gegner mit einem Zauber zu treffen (80 % Trefferchance, -20 Lebenspunkte). Mana wird verbraucht (-10). Sonst normaler Angriff.
+ normaler_angriff(gegner): Wenn kein Mana mehr vorhanden, 60 % Trefferchance (-5 Lebenspunkte).
+ info(): Zeigt Name, Lebenspunkte und Mana an.

**Amazone:**

Attribute:
+ name (Name der Amazone)
+ lebenspunkte (startet mit 120)
+ pfeile (startet mit 5)

Methoden:
+ schiessen(gegner): Wenn Pfeile vorhanden, 70 % Trefferchance (-15 Lebenspunkte). Sonst normaler Angriff.
+ normaler_angriff(gegner): 65 % Trefferchance (-7 Lebenspunkte).
+ info(): Zeigt Name, Lebenspunkte und Pfeile an.

>  **Tipp:** Verwenden Sie `@dataclass` mit `field(default=..., init=False)` fr Attribute wie `lebenspunkte` und `mana`, die nicht beim Erstellen der Instanz bergeben werden sollen.

Schwierigkeit: 

`````python
#
# Hier bitte Ihren Code eingeben
#
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lsungsvorschlag Klasse Magier</font> </br></p>
import random
from dataclasses import dataclass, field

@dataclass
class Magier:
    # --- Instanzattribute ---
    name: str
    lebenspunkte: int = field(default=100, init=False)
    mana: int         = field(default=50,  init=False)

    # --- Instanzmethode ---
    def zaubern(self, gegner):
        if self.mana >= 10:
            if random.random() < 0.8:
                gegner.lebenspunkte -= 20
                print(f"{self.name}  zaubert einen Spruch  auf {gegner.name} und verursacht 20 Schaden!")
            else:
                print(f"{self.name}'s Zauber hat verfehlt!")
            self.mana -= 10
        else:
            self.normaler_angriff(gegner)

    def normaler_angriff(self, gegner):
        if random.random() < 0.6:
            gegner.lebenspunkte -= 5
            print(f"{self.name} greift {gegner.name} normal an und verursacht 5 Schaden!")
        else:
            print(f"{self.name}'s Angriff hat verfehlt!")

    def info(self):
        print(f"{self.name} - Lebenspunkte: {self.lebenspunkte}, Mana: {self.mana}")

magier = Magier("Merlin")
magier.info()
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lsungsvorschlag Klasse Amazone</font> </br></p>
from dataclasses import dataclass, field

@dataclass
class Amazone:
    # --- Instanzattribute ---
    name: str
    lebenspunkte: int = field(default=120, init=False)
    pfeile: int       = field(default=5,   init=False)

    # --- Instanzmethode ---
    def schiessen(self, gegner):
        if self.pfeile > 0:
            if random.random() < 0.7:
                gegner.lebenspunkte -= 15
                print(f"{self.name}  schiet einen Pfeil  auf {gegner.name} und verursacht 15 Schaden!")
            else:
                print(f"{self.name}'s Pfeil hat verfehlt!")
            self.pfeile -= 1
        else:
            self.normaler_angriff(gegner)

    def normaler_angriff(self, gegner):
        if random.random() < 0.65:
            gegner.lebenspunkte -= 7
            print(f"{self.name} greift {gegner.name} normal an und verursacht 7 Schaden!")
        else:
            print(f"{self.name}'s Angriff hat verfehlt!")

    def info(self):
        print(f"{self.name} - Lebenspunkte: {self.lebenspunkte}, Pfeile: {self.pfeile}")

amazone = Amazone("Diana")
amazone.info()
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lsungsvorschlag Rollenspiel Kampf</font> </br></p>
def Kampf(magier, amazone):
    while magier.lebenspunkte > 0 and amazone.lebenspunkte > 0:
        magier.zaubern(amazone)
        if amazone.lebenspunkte <= 0:
            print(f"{amazone.name} wurde besiegt!")
            return

        amazone.schiessen(magier)
        if magier.lebenspunkte <= 0:
            print(f"{magier.name} wurde besiegt!")
            return

Kampf(magier, amazone)
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lsungsideen</font> </br></p>
from dataclasses import dataclass, field

@dataclass
class Kurs:
    # --- Instanzattribute ---
    name: str
    max_teilnehmer: int
    teilnehmer_liste: list = field(default_factory=list)

    # --- Instanzmethode ---
    def teilnehmer_hinzufuegen(self, mitglied):
        if len(self.teilnehmer_liste) < self.max_teilnehmer:
            self.teilnehmer_liste.append(mitglied)
            return True
        return False

    def teilnehmer_entfernen(self, mitglied):
        if mitglied in self.teilnehmer_liste:
            self.teilnehmer_liste.remove(mitglied)

@dataclass
class Mitglied:
    # --- Instanzattribute ---
    name: str
    mitgliedsnummer: int
    gebuchte_kurse: list = field(default_factory=list)

    # --- Instanzmethode ---
    def kurs_buchen(self, kurs):
        if kurs in self.gebuchte_kurse:
            print(f"{self.name} ist bereits fr den Kurs '{kurs.name}' angemeldet.")
        else:
            if kurs.teilnehmer_hinzufuegen(self):
                self.gebuchte_kurse.append(kurs)
                print(f"{self.name} hat sich erfolgreich fr den Kurs '{kurs.name}' angemeldet.")
            else:
                print(f"Der Kurs '{kurs.name}' ist bereits voll.")

    def kurs_stornieren(self, kurs):
        if kurs in self.gebuchte_kurse:
            kurs.teilnehmer_entfernen(self)
            self.gebuchte_kurse.remove(kurs)
            print(f"{self.name} hat die Buchung fr den Kurs '{kurs.name}' storniert.")
        else:
            print(f"{self.name} ist nicht fr den Kurs '{kurs.name}' angemeldet.")

@dataclass
class Fitnessstudio:
    # --- Instanzattribute ---
    name: str
    mitglieder_liste: list = field(default_factory=list)
    kurs_liste: list       = field(default_factory=list)

    # --- Instanzmethode ---
    def mitglied_hinzufuegen(self, mitglied):
        self.mitglieder_liste.append(mitglied)
        print(f"Mitglied {mitglied.name} wurde hinzugefgt.")

    def kurs_hinzufuegen(self, kurs):
        self.kurs_liste.append(kurs)
        print(f"Kurs '{kurs.name}' wurde hinzugefgt.")

    def verfuegbare_kurse_anzeigen(self):
        print("Verfgbare Kurse mit freien Pltzen:")
        for kurs in self.kurs_liste:
            freie_plaetze = kurs.max_teilnehmer - len(kurs.teilnehmer_liste)
            if freie_plaetze > 0:
                print(f"- {kurs.name}: {freie_plaetze} freie Pltze")

# Beispiel
anna = Mitglied("Anna", 1)
ben  = Mitglied("Ben",  2)

yoga   = Kurs("Yoga",   2)
zumba  = Kurs("Zumba",  3)

fitnessstudio = Fitnessstudio("Fit & Gesund")

fitnessstudio.mitglied_hinzufuegen(anna)
fitnessstudio.mitglied_hinzufuegen(ben)
fitnessstudio.kurs_hinzufuegen(yoga)
fitnessstudio.kurs_hinzufuegen(zumba)

fitnessstudio.verfuegbare_kurse_anzeigen()

anna.kurs_buchen(yoga)
ben.kurs_buchen(yoga)

fitnessstudio.verfuegbare_kurse_anzeigen()

anna.kurs_stornieren(yoga)
fitnessstudio.verfuegbare_kurse_anzeigen()
`````

`````python
#
# Code durch ChatGPT erstellt
#
def einkommensteuer_berechnen(zu_versteuerndes_einkommen, tarif):
    if tarif == "splitting":
        zu_versteuerndes_einkommen /= 2

    # Berechnung der Einkommensteuer gem den Grenzen und Formeln
    if zu_versteuerndes_einkommen <= 10347:
        steuer = 0
    elif zu_versteuerndes_einkommen <= 62978:
        z = (zu_versteuerndes_einkommen - 10347) / 10000
        steuer = (995.21 * z + 1400) * z
    elif zu_versteuerndes_einkommen <= 300919:
        y = (zu_versteuerndes_einkommen - 62978) / 10000
        steuer = (216.16 * y + 2397) * y + 950
    else:
        steuer = 0.45 * zu_versteuerndes_einkommen - 17386

    if tarif == "splitting":
        steuer *= 2  # Steuer wird verdoppelt, da das Einkommen halbiert wurde

    return round(steuer, 2)  # Rundet die Steuer auf zwei Dezimalstellen
`````
