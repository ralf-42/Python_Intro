```python
import gradio as gr
from googletrans import Translator, LANGUAGES
```

```python
# Initialisiere den Übersetzer
translator = Translator()

def translate_text(text, source_lang, target_lang):
    # Übersetze den Text
    translated = translator.translate(text, src=source_lang, dest=target_lang)
    return translated.text
```

```python
# Erstelle die Gradio-Benutzeroberfläche
with gr.Blocks() as app:
    gr.Markdown("## Einfacher Sprachübersetzer")

    with gr.Row():
        text_input = gr.Textbox(label="Gib den Text ein, den du übersetzen möchtest")
        source_lang = gr.Dropdown(choices=list(LANGUAGES.values()), label="Quellsprache")
        target_lang = gr.Dropdown(choices=list(LANGUAGES.values()), label="Zielsprache")

    output_text = gr.Textbox(label="Übersetzter Text")

    gr.Button("Übersetzen").click(translate_text, inputs=[text_input, source_lang, target_lang], outputs=output_text)
```

```python
# Starte die Gradio-App
app.launch()
```
