Schreiben Sie ein Programm das folgenden Funktionen abdeckt:

+ Einlesen einer Jahreszahl
+ Ermittlung, ob das eingegebene Jahr ein Schaltjahr ist oder nicht.
+ Die durch 4 ganzzahlig teilbaren Jahre sind, abgesehen von den folgenden Ausnahmen, Schaltjahre.
+ Säkularjahre, also die Jahre, die ein Jahrhundert abschließen (z. B. 1800, 1900, 2100 und 2200), sind, abgesehen von der folgenden Ausnahme, keine Schaltjahre.
+ Die durch 400 ganzzahlig teilbaren Säkularjahre, zum Beispiel das Jahr 2000, sind jedoch Schaltjahre.

Die Logik zur Berechnung eines Schaltjahres (Gregorianischer Kalender) finden Sie auf folgendem Link: https://de.wikipedia.org/wiki/Schaltjahr


Schwierigkeit: ⭐⭐⭐

`````python
#
# Bitte hier Ihren Code eingeben
#
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
jahr = int(input("Jahr: "))
schaltjahr = "kein"
if jahr % 4 == 0:
    schaltjahr = "ein"
    if jahr % 100 == 0:
        schaltjahr = "kein"
    if jahr % 400 == 0:
        schaltjahr = "ein"

print(f"Das Jahr {jahr} ist {schaltjahr} Schaltjahr")
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>

txt = "kein"
jahr = int(input("Bitte Jahr eingeben: "))
if (jahr % 4 == 0) and (jahr % 100 != 0 or jahr % 400 == 0):
    txt = txt[1:]
print (f"Das Jahr {jahr} ist {txt} Schaltjahr")
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>

jahr = int(input("Bitte Jahr eingeben: "))

if jahr % 400 == 0:
    schaltjahr = True
elif jahr % 100 == 0:
    schaltjahr = False
elif jahr % 4 == 0:
    schaltjahr = True
else:
    schaltjahr = False

if schaltjahr:
    print(jahr, 'ist ein Schaltjahr.')
else:
    print(jahr, 'ist kein Schaltjahr.')
`````

### 7.3.3 |  Passwort-Validator

Erstelle ein Programm, das die Gültigkeit eines Passworts basierend auf mehreren Kriterien überprüft. Das Programm soll eine Zeichenkette (das Passwort) als Eingabe erhalten und überprüfen, ob das Passwort die folgenden Bedingungen erfüllt:

+ Es muss mindestens 8 Zeichen lang sein.
+ Es muss mindestens einen Großbuchstaben enthalten.
+ Es muss mindestens einen Kleinbuchstaben enthalten.
+ Es muss mindestens eine Zahl enthalten.
+ Es muss mindestens ein Sonderzeichen aus dieser Menge enthalten: @#$%&


Wenn das Passwort alle Bedingungen erfüllt, soll das Programm ausgeben: "Passwort ist gültig." Wenn das Passwort eine oder mehrere Bedingungen nicht erfüllt, soll das Programm ausgeben: "Passwort ist ungültig."



Schwierigkeit: ⭐⭐

`````python
#
# Bitte hier Ihren Code eingeben
#
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>

# Starte die ereignisgesteuerte Schleife
while True:
    # Fordere den Benutzer auf, ein Passwort einzugeben
    BenutzerPasswort = input("Bitte geben Sie ein Passwort ein: ")

    # Initialisiere die Kriterien für ein gültiges Passwort
    MinLaenge = 8
    HatGrossbuchstabe = False
    HatKleinbuchstabe = False
    HatZahl = False
    HatSonderzeichen = False
    Sonderzeichen = "_-"

    # Überprüfe die Länge des Passworts
    if len(BenutzerPasswort) >= MinLaenge:
        # Überprüfe jedes Zeichen im Passwort
        for zeichen in BenutzerPasswort:
            if zeichen.isupper():
                HatGrossbuchstabe = True
            if zeichen.islower():
                HatKleinbuchstabe = True
            if zeichen.isdigit():
                HatZahl = True
            if zeichen in Sonderzeichen:
                HatSonderzeichen = True

    # Überprüfe, ob alle Kriterien erfüllt sind
    if HatGrossbuchstabe and HatKleinbuchstabe and HatZahl and HatSonderzeichen:
        print("Passwort ist gültig.")
        break  # Beende die Schleife, wenn das Passwort gültig ist
    else:
        # Informiere den Benutzer, dass das Passwort ungültig ist und fordere zur erneuten Eingabe auf
        print("Passwort ist ungültig. Bitte versuchen Sie es erneut.")
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
def verschluesseln(text, verschiebung):
    verschluesselter_text = ""

    for char in text:
        # Nur Alphabete verschlüsseln
        if char.isalpha():
            verschoben = ord(char) + verschiebung

            if char.islower():
                if verschoben > ord('z'):
                    verschoben -= 26
            elif char.isupper():
                if verschoben > ord('Z'):
                    verschoben -= 26

            verschluesselter_text += chr(verschoben)
        else:
            verschluesselter_text += char

    return verschluesselter_text

def entschluesseln(text, verschiebung):
    return verschluesseln(text, -verschiebung)

# Eingabe von Text und Verschiebung vom Benutzer
text = input("Bitte geben Sie den zu verschlüsselnden Text ein: ")
verschiebung = int(input("Bitte geben Sie die Verschiebung ein (ganze Zahl): "))

verschluesselter_text = verschluesseln(text, verschiebung)
print(f"Verschlüsselter Text: {verschluesselter_text}")

entschluesselter_text = entschluesseln(verschluesselter_text, verschiebung)
print(f"Entschlüsselter Text: {entschluesselter_text}")
`````

## 10.5 |  Vigenère-Code

Eine sehr viel sicherere Verschlüsselungsmethode als der Cäsarcode ist der Vigenere-Code (s. wikipedia). Dabei wird ganz analog zum Cäsarcode verschlüsselt, aber es werden statt eines festen ks verschiedene ks verwendet, die aus einem Schlüsswelwort hervorgehen. Haben wir als Schlüsselwort z.B. 'key', dann verschlusseln wir den ersten Buchstaben des zu verschlüsselnden Worts mit dem Cäsarcode mit k = 10 (für 'k'), den zweiten Buchstaben des zu verschlüsselnden Worts mit dem Cäsarcode mit k = 4 (fur 'e'), den dritten Buchstaben mit k = 24 (für 'y'), den vierten Buchstaben wieder mit k = 10 (für 'k'), usw.

`````python
#@title
#@markdown   <p><font size="4" color='blue'>  Lösungshilfe - bei Bedarf lesen</font> </br></p>
#
# itertools.cycle ist eine Funktion aus dem itertools-Modul in Python.
# Sie gibt einen unendlichen Iterator zurück, der über die gegebenen Eingabewerte
# iteriert. Wenn das Ende der Eingabe erreicht ist, fängt sie wieder von vorne an.
# Das bedeutet, dass cycle die Eingabewerte endlos wiederholt.
# -----
from itertools import cycle
zahlen = cycle([1, 2, 3])
for i in range(10):
    print(next(zahlen))
# -----
import string
string.printable
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
import string
from itertools import cycle

def vigenere_encode(text, key):
    return vigenere_cypher(text, key)

def vigenere_decode(text, key):
    return vigenere_cypher(text, key, mode="decode")

def vigenere_cypher(text, key, mode="encode"):
    result = []
    key_cycle = cycle(key)
    base = (string.printable) * 2

    for t in text:
        if t in base:
            k = next(key_cycle)
            t_index = base.index(t)
            k_index = base.index(k)

            if mode == "decode":
                new_index = (t_index - k_index)
            else:
                new_index = (t_index + k_index)

            result.append(base[new_index])
        else:
            result.append(t)
            next(key_cycle)

    return ''.join(result)

if __name__ == "__main__":
    text = input("Bitte geben Sie den Text ein: ")
    key = input("Bitte geben Sie den Schlüssel ein: ")
    mode = input("Wählen Sie den Modus (encode/decode): ")

    if not key:
        print("Der Schlüssel darf nicht leer sein!")
    else:
        match mode:
            case "encode":
                processed_text = vigenere_encode(text, key)
                print(f"Verschlüsselter Text: {processed_text}")
            case "decode":
                processed_text = vigenere_decode(text, key)
                print(f"Entschlüsselter Text: {processed_text}")
            case _:
                print("Ungültiger Modus!")
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
import random

def computer_choice():
    return random.choice(['Schere', 'Stein', 'Papier'])

def player_choice():
    print("Wähle eine Option:")
    print("1. Schere")
    print("2. Stein")
    print("3. Papier")
    choice = input("Deine Wahl (1/2/3): ")
    while choice not in ['1', '2', '3']:
        print("Ungültige Eingabe. Bitte versuche es erneut.")
        choice = input("Deine Wahl (1/2/3): ")
    return ['Schere', 'Stein', 'Papier'][int(choice) - 1]

def round_winner(player, computer):
    if player == computer:
        return 'Unentschieden'
    elif (player == 'Schere' and computer == 'Papier') or \
         (player == 'Papier' and computer == 'Stein') or \
         (player == 'Stein' and computer == 'Schere'):
        return 'Spieler'
    else:
        return 'Computer'

def overall_winner(player_score, computer_score):
    if player_score > computer_score:
        return "Du bist der Gesamtsieger!"
    elif computer_score > player_score:
        return "Computer ist der Gesamtsieger!"
    else:
        return "Das Spiel endet Unentschieden!"

def main():
    print("Schere, Stein, Papier Spiel!")

    player_score = 0
    computer_score = 0

    for i in range(3):
        print(f"\nRunde {i + 1} von 3:")

        player = player_choice()
        computer = computer_choice()
        print(f"\nDu hast {player} gewählt.")
        print(f"Computer hat {computer} gewählt.")

        winner = round_winner(player, computer)
        print(winner, "gewinnt diese Runde!")

        if winner == 'Spieler':
            player_score += 1
        elif winner == 'Computer':
            computer_score += 1

    print("\n" + overall_winner(player_score, computer_score))

if __name__ == "__main__":
    main()
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag Klasse Magier</font> </br></p>
import random

class Magier:
    def __init__(self, name):
        self.name = name
        self.lebenspunkte = 100
        self.mana = 50

    def zaubern(self, gegner):
        if self.mana >= 10:
            if random.random() < 0.8:
                gegner.lebenspunkte -= 20
                print(f"{self.name} 🧙‍♂️ zaubert einen Spruch 🪄 auf {gegner.name} und verursacht 20 Schaden!")
            else:
                print(f"{self.name}'s Zauber hat verfehlt!")
            self.mana -= 10
        else:
            self.normaler_angriff(gegner)

    def normaler_angriff(self, gegner):
        if random.random() < 0.6:  # 60% Trefferchance für den normalen Angriff des Magiers
            gegner.lebenspunkte -= 5
            print(f"{self.name} greift {gegner.name} normal an und verursacht 5 Schaden!")
        else:
            print(f"{self.name}'s Angriff hat verfehlt!")

    def info(self):
        print(f"{self.name} - Lebenspunkte: {self.lebenspunkte}, Mana: {self.mana}")

magier = Magier("Merlin")

magier.info()   # Merlin - Lebenspunkte: 100, Mana: 5
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag Klasse Amazone</font> </br></p>
class Amazone:
    def __init__(self, name):
        self.name = name
        self.lebenspunkte = 120
        self.pfeile = 5

    def schiessen(self, gegner):
        if self.pfeile > 0:
            if random.random() < 0.7:
                gegner.lebenspunkte -= 15
                print(f"{self.name} 🧝🏻‍♀️ schießt einen Pfeil 🏹 auf {gegner.name} und verursacht 15 Schaden!")
            else:
                print(f"{self.name}'s Pfeil hat verfehlt!")
            self.pfeile -= 1
        else:
            self.normaler_angriff(gegner)

    def normaler_angriff(self, gegner):
        if random.random() < 0.65:  # 65% Trefferchance für den normalen Angriff der Amazone
            gegner.lebenspunkte -= 7
            print(f"{self.name} greift {gegner.name} normal an und verursacht 7 Schaden!")
        else:
            print(f"{self.name}'s Angriff hat verfehlt!")

    def info(self):
        print(f"{self.name} - Lebenspunkte: {self.lebenspunkte}, Pfeile: {self.pfeile}")

amazone = Amazone("Diana")

amazone.info()  # Diana - Lebenspunkte: 120, Pfeile: 5
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag Rollenspiel Kampf</font> </br></p>
def Kampf(magier, amazone):
    while magier.lebenspunkte > 0 and amazone.lebenspunkte > 0:
        magier.zaubern(amazone)
        if amazone.lebenspunkte <= 0:
            print(f"{amazone.name} wurde besiegt!")
            return

        amazone.schiessen(magier)
        if magier.lebenspunkte <= 0:
            print(f"{magier.name} wurde besiegt!")
            return

Kampf(magier, amazone)
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
class Buch:
    def __init__(self, titel, autor):
        self.titel = titel
        self.autor = autor
        self.ist_ausgeliehen = False

class Mitglied:
    def __init__(self, name):
        self.name = name
        self.ausgeliehene_buecher = []

    def buch_ausleihen(self, buch):
        if not buch.ist_ausgeliehen:
            buch.ist_ausgeliehen = True
            self.ausgeliehene_buecher.append(buch)
            print(f"{self.name} hat '{buch.titel}' ausgeliehen.")
        else:
            print(f"'{buch.titel}' ist bereits ausgeliehen.")

    def buch_zurueckgeben(self, buch):
        if buch in self.ausgeliehene_buecher:
            buch.ist_ausgeliehen = False
            self.ausgeliehene_buecher.remove(buch)
            print(f"{self.name} hat '{buch.titel}' zurückgegeben.")
        else:
            print(f"{self.name} hat '{buch.titel}' nicht ausgeliehen.")

# Beispiel für die Nutzung des Systems
buch1 = Buch("Der kleine Prinz", "Antoine de Saint-Exupéry")
mitglied1 = Mitglied("Anna")

mitglied1.buch_ausleihen(buch1)  # Anna leiht das Buch aus
mitglied1.buch_zurueckgeben(buch1)  # Anna gibt das Buch zurück
`````

`````python
#@title
#@markdown <p><font size="4" color='green'> Lösungsvorschlag</font> </br></p>
class Task:
    def __init__(self, description):
        """
        Initialisiert eine Aufgabe mit einer Beschreibung und als unvollendet markiert.
        """
        self.description = description
        self.completed = False

class ToDoList:
    def __init__(self):
        """
        Initialisiert eine leere ToDo-Liste, um Aufgaben zu verwalten.
        """
        self.tasks = []

    def add_task(self, task):
        """
        Fügt eine neue Aufgabe zur Liste hinzu.
        """
        self.tasks.append(task)

    def complete_task(self, task_description):
        """
        Markiert eine Aufgabe als erledigt, basierend auf der Beschreibung.
        """
        for task in self.tasks:
            if task.description == task_description:
                task.completed = True
                break  # Sobald die Aufgabe gefunden wurde, kann die Schleife abgebrochen werden

    def get_tasks_by_completion(self, completed):
        """
        Gibt eine Liste der Aufgaben zurück, die den angegebenen Abschlussstatus haben.
        """
        return [task.description for task in self.tasks if task.completed == completed]

if __name__ == "__main__":
    my_list = ToDoList()

    # Aufgaben hinzufügen
    task1_description = input("Gib eine Aufgabenbeschreibung ein: ")
    task1 = Task(task1_description)
    task2_description = input("Gib eine weitere Aufgabenbeschreibung ein: ")
    task2 = Task(task2_description)
    my_list.add_task(task1)
    my_list.add_task(task2)

    # Aufgabe als erledigt markieren
    completed_task = input("Gib den Namen der erledigten Aufgabe ein: ")
    my_list.complete_task(completed_task)

    # Abgeschlossene und unvollständige Aufgaben anzeigen
    completed_tasks = my_list.get_tasks_by_completion(True)
    incomplete_tasks = my_list.get_tasks_by_completion(False)
    print("Completed tasks:", completed_tasks)
    print("Incomplete tasks:", incomplete_tasks)
`````

`````python
#@title
#@markdown <p><font size="4" color='green'> Lösungsvorschlag - Fortsetzung</font> </br></p>
# Verbesserungen:
# Die Klassen Task und ToDoList wurden kommentiert, um ihren Zweck und ihre Funktionalitäten zu erklären.
# Die Methoden wurden klarer benannt, um ihre Funktionen besser widerzuspiegeln.
# Die Schleife in der Methode complete_task wird abgebrochen, sobald die Aufgabe gefunden und erledigt wurde.
# Die Methode get_tasks_by_completion wurde generalisiert, um abgeschlossene und unvollständige Aufgaben basierend auf einem Status abzurufen.
# Die Benutzeroberfläche wurde beibehalten, um Aufgaben hinzuzufügen, als erledigt zu markieren und die Listen der abgeschlossenen und unvollständigen Aufgaben anzuzeigen.
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsideen</font> </br></p>
class Mitglied:
    def __init__(self, name, mitgliedsnummer):
        self.name = name
        self.mitgliedsnummer = mitgliedsnummer
        self.gebuchte_kurse = []

    def kurs_buchen(self, kurs):
        if kurs in self.gebuchte_kurse:
            print(f"{self.name} ist bereits für den Kurs '{kurs.name}' angemeldet.")
        else:
            if kurs.teilnehmer_hinzufuegen(self):
                self.gebuchte_kurse.append(kurs)
                print(f"{self.name} hat sich erfolgreich für den Kurs '{kurs.name}' angemeldet.")
            else:
                print(f"Der Kurs '{kurs.name}' ist bereits voll.")

    def kurs_stornieren(self, kurs):
        if kurs in self.gebuchte_kurse:
            kurs.teilnehmer_entfernen(self)
            self.gebuchte_kurse.remove(kurs)
            print(f"{self.name} hat die Buchung für den Kurs '{kurs.name}' storniert.")
        else:
            print(f"{self.name} ist nicht für den Kurs '{kurs.name}' angemeldet.")

class Kurs:
    def __init__(self, name, max_teilnehmer):
        self.name = name
        self.max_teilnehmer = max_teilnehmer
        self.teilnehmer_liste = []

    def teilnehmer_hinzufuegen(self, mitglied):
        if len(self.teilnehmer_liste) < self.max_teilnehmer:
            self.teilnehmer_liste.append(mitglied)
            return True
        else:
            return False

    def teilnehmer_entfernen(self, mitglied):
        if mitglied in self.teilnehmer_liste:
            self.teilnehmer_liste.remove(mitglied)

class Fitnessstudio:
    def __init__(self, name):
        self.name = name
        self.mitglieder_liste = []
        self.kurs_liste = []

    def mitglied_hinzufuegen(self, mitglied):
        self.mitglieder_liste.append(mitglied)
        print(f"Mitglied {mitglied.name} wurde hinzugefügt.")

    def kurs_hinzufuegen(self, kurs):
        self.kurs_liste.append(kurs)
        print(f"Kurs '{kurs.name}' wurde hinzugefügt.")

    def verfuegbare_kurse_anzeigen(self):
        print("Verfügbare Kurse mit freien Plätzen:")
        for kurs in self.kurs_liste:
            freie_plaetze = kurs.max_teilnehmer - len(kurs.teilnehmer_liste)
            if freie_plaetze > 0:
                print(f"- {kurs.name}: {freie_plaetze} freie Plätze")

# Beispiel zur Verwendung des Systems
anna = Mitglied("Anna", 1)
ben = Mitglied("Ben", 2)

yoga = Kurs("Yoga", 2)
zumba = Kurs("Zumba", 3)

fitnessstudio = Fitnessstudio("Fit & Gesund")

# Mitglieder hinzufügen
fitnessstudio.mitglied_hinzufuegen(anna)
fitnessstudio.mitglied_hinzufuegen(ben)

# Kurse hinzufügen
fitnessstudio.kurs_hinzufuegen(yoga)
fitnessstudio.kurs_hinzufuegen(zumba)

# Verfügbare Kurse anzeigen
fitnessstudio.verfuegbare_kurse_anzeigen()

# Mitglieder buchen Kurse
anna.kurs_buchen(yoga)
ben.kurs_buchen(yoga)

# Verfügbare Kurse nach Buchung anzeigen
fitnessstudio.verfuegbare_kurse_anzeigen()

# Stornierung eines Kurses
anna.kurs_stornieren(yoga)

# Verfügbare Kurse nach Stornierung anzeigen
fitnessstudio.verfuegbare_kurse_anzeigen()
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
def ist_palindrom(s: str) -> bool:
    # Entfernen von Satzzeichen, Leerzeichen und Umwandlung in Kleinbuchstaben
    gereinigter_string = ''.join(e for e in s if e.isalnum()).lower()
    # Vergleich des Strings mit seiner Umkehrung
    return gereinigter_string == gereinigter_string[::-1]

# Test der Funktion
test_string = input("Palindrom? Wort oder Satz angeben: ")
if ist_palindrom(test_string):
    print(f"'{test_string}' ist ein Palindrom.")
else:
    print(f"'{test_string}' ist kein Palindrom.")
`````

Schreiben Sie ein Programm das einen Blockchain erstellt. Ein Blockchain ist eine Kette von Datenblöcken, bei denen jeder Block einen Verweis auf den vorherigen Block enthält (meist in Form eines Hashes).

Es soll diese Transaktionen fälschungssicher speichern, also so, dass sie nachträglich möglichst schwer zu manipulieren sind.

Der "Block" enthält jedweils die nachfolgenden Daten:

+ Index
+ alter hash (des vorigen Blocks)
+ Zeitstempel
+ Daten
+ neuer hash

Verwende die Hash-Funktion `sha256`.



```
import hashlib
.encode('utf-8')  # Methode konvertiert String in einen Byte-Array, wird von hashlib erwartet
hashlib.sha256()  # Funktion erzeugt ein Hash-Objekt
.hexdigest()      # Methode konvertiert das Hash-Objekts in einen hexadezimalen String
```


**Idee zum Vorgehen:**
1. Legen Sie zuerst die Blockchain als leere Liste an.
2. Erzeugen Sie den ersten Block, den "Genesis-Block". Verwenden Sie 'genesis' als vorangegangenen Hash.
3. Plazieren Sie einige zufällige Transaktionen/Daten im Block.
3. Fügen Sie den fertigen Block an die Blockchain an.
4. Erzeugen Sie den zweiten bis x. Block.

Schwierigkeit: ⭐⭐⭐⭐⭐

`````python
#
# Hier bitte Ihren Code eingeben
#
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
import hashlib
import time
from datetime import datetime
# --------------------------------------------------------
class Block:
    def __init__(self, index, previous_hash, timestamp, data, hash):
        self.index = index
        self.previous_hash = previous_hash
        self.timestamp = timestamp
        self.data = data
        self.hash = hash

    def __str__(self):
        return (f"Index: {self.index}\n"
                f"Previous Hash: {self.previous_hash}\n"
                f"Timestamp: {self.timestamp}\n"
                f"Data: {self.data}\n"
                f"Hash: {self.hash}\n")
# --------------------------------------------------------
# Ermittle Hash-Objekt
def calculate_hash(index, previous_hash, timestamp, data):
    return hashlib.sha256(f'{index}{previous_hash}{timestamp}{data}'.encode('utf-8')).hexdigest()
# Erstelle Genesis-Block
def create_genesis_block():
    timestamp = datetime.now().strftime('%Y-%m-%d-%H-%M-%S-%f')
    return Block(0, '0', timestamp, 'Genesis Block', calculate_hash(0, '0', timestamp, 'Genesis Block'))
# Erstelle neuen Folgeblock
def create_new_block(prev_block, data):
    index = prev_block.index + 1
    timestamp = datetime.now().strftime('%Y-%m-%d-%H-%M-%S-%f')
    hash = calculate_hash(index, prev_block.hash, timestamp, data)
    return Block(index, prev_block.hash, timestamp, data, hash)
# --------------------------------------------------------
# Erstelle die Blockchain und füge den Genesis-Block hinzu
blockchain = [create_genesis_block()]
previous_block = blockchain[0]
# --------------------------------------------------------
# Füge 10 weitere Blöcke zur Blockchain hinzu
num_of_blocks_to_add = 10
# --------------------------------------------------------
for i in range(1, num_of_blocks_to_add + 1):
    new_data = f"Block #{i} data"
    new_block = create_new_block(previous_block, new_data)
    blockchain.append(new_block)
    previous_block = new_block
    print(f"Block #{new_block.index} has been added to the blockchain!")
    print(f"Hash: {new_block.hash}\n")
# --------------------------------------------------------
# Beispiele
print(blockchain[0])  # Genesis-Block
print(blockchain[5])  # 5. Block
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
#
# Version 1 - exklusive doc-strings
# Erstellt mit ChatGPT
#
# reading the file
with open(file_name) as fp:
    contents=fp.readlines()

# initialize two counter to check mismatch between "(" and ")"
open_bracket_counter=0
close_bracket_counter=0

# whenever an element deleted from the list length of the list will be decreased
decreasing_counter=0

for number in range(len(contents)):

    # checking if the line contains "#" or not
    if "#" in contents[number-decreasing_counter]:

        # delete the line if startswith "#"
        if contents[number-decreasing_counter].startswith("#"):
            contents.remove(contents[number-decreasing_counter])
            decreasing_counter+=1

        # delete the character after the "#"
        else:
            newline=""
            for character in contents[number-decreasing_counter]:
                if character=="(":
                    open_bracket_counter+=1
                    newline+=character
                elif character==")":
                    close_bracket_counter+=1
                    newline+=character
                elif character=="#" and open_bracket_counter==close_bracket_counter:
                    break
                else:
                    newline+=character
            contents.remove(contents[number-decreasing_counter])
            contents.insert(number-decreasing_counter,newline)

new_file_name = "new_" + file_name
# writing into a new file
with open(new_file_name,"w") as fp:
    fp.writelines(contents)
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Lösungsvorschlag</font> </br></p>
#
# Version 2 - inklusive doc-strings
# Erstellt mit ChatGPT
#
import io, tokenize, re
def remove_comments_and_docstrings(source):
    io_obj = io.StringIO(source)
    out = ""
    prev_toktype = tokenize.INDENT
    last_lineno = -1
    last_col = 0
    for tok in tokenize.generate_tokens(io_obj.readline):
        token_type = tok[0]
        token_string = tok[1]
        start_line, start_col = tok[2]
        end_line, end_col = tok[3]
        ltext = tok[4]
        if start_line > last_lineno:
            last_col = 0
        if start_col > last_col:
            out += (" " * (start_col - last_col))
        if token_type == tokenize.COMMENT:
            pass
        elif token_type == tokenize.STRING:
            if prev_toktype != tokenize.INDENT:
                if prev_toktype != tokenize.NEWLINE:
                    if start_col > 0:
                        out += token_string
        else:
            out += token_string
        prev_toktype = token_type
        last_col = end_col
        last_lineno = end_line
    out = '\n'.join(l for l in out.splitlines() if l.strip())
    return out


# reading the file
with open(file_name) as f_in:
  new_file_name = "new_" + file_name
  with open(new_file_name,"w") as f_out:
    f_out.writelines(remove_comments_and_docstrings(f_in.read()))
`````

CadQuery ist eine skriptbasierte 3D-CAD-Modellierungspaket, die in Python geschrieben ist. Es ermöglicht den Benutzern, 3D-Modelle mit Python-Scripts zu erstellen, anstatt herkömmliche GUI-basierte Ansätze zu verwenden. Das Paket verwendet eine parametrische Modellierungsweise, was bedeutet, dass Sie Modelle erstellen können, indem Sie eine Serie von Modellierungsoperationen definieren, die auf vorherige Schritte aufbauen.

Einige Hauptmerkmale und Vorteile von CadQuery sind:

1. Python-basiert: Benutzer, die bereits mit Python vertraut sind, finden CadQuery zugänglich und können ihre vorhandenen Python-Kenntnisse nutzen.

2. Parametrisch: Modelle in CadQuery sind vollständig parametrisch, so dass Sie leicht Änderungen vornehmen und verschiedene Variationen eines Modells generieren können.

3. Selektoren: CadQuery beitet eine leistungsstarke Methode, um Teile eines 3D-Modells zu "selektieren" und Operationen darauf anzuwenden. Dies macht das Skripten komplexer Modelle einfacher und intuitiver.

4. Open Source: CadQuery ist Open Source und kann frei verwendet und modifiziert werden.

5. Erweiterbar: Da es in Python geschrieben ist, können Benutzer CadQuery leicht erweitern oder es in größere Python-Programme oder -Workflows integrieren.

Eine STL-Datei enthält Informationen über die Oberfläche eines dreidimensionalen Objekts, die in Form eines Netzes aus Dreiecken beschrieben wird.

`````python
#
# Code durch ChatGPT erstellt
#
def einkommensteuer_berechnen(zu_versteuerndes_einkommen, tarif):
    if tarif == "splitting":
        zu_versteuerndes_einkommen /= 2

    # Berechnung der Einkommensteuer gemäß den Grenzen und Formeln
    if zu_versteuerndes_einkommen <= 10347:
        steuer = 0
    elif zu_versteuerndes_einkommen <= 62978:
        z = (zu_versteuerndes_einkommen - 10347) / 10000
        steuer = (995.21 * z + 1400) * z
    elif zu_versteuerndes_einkommen <= 300919:
        y = (zu_versteuerndes_einkommen - 62978) / 10000
        steuer = (216.16 * y + 2397) * y + 950
    else:
        steuer = 0.45 * zu_versteuerndes_einkommen - 17386

    if tarif == "splitting":
        steuer *= 2  # Steuer wird verdoppelt, da das Einkommen halbiert wurde

    return round(steuer, 2)  # Rundet die Steuer auf zwei Dezimalstellen
`````
