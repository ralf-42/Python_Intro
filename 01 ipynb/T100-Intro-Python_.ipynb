`````python
#
# Bitte hier Ihren Code eingeben
#
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  L√∂sungsvorschlag</font> </br></p>
jahr = int(input("Jahr: "))
schaltjahr = "kein"
if jahr % 4 == 0:
    schaltjahr = "ein"
    if jahr % 100 == 0:
        schaltjahr = "kein"
    if jahr % 400 == 0:
        schaltjahr = "ein"

print(f"Das Jahr {jahr} ist {schaltjahr} Schaltjahr")
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  L√∂sungsvorschlag</font> </br></p>
def verschluesseln(text, verschiebung):
    verschluesselter_text = ""

    for char in text:
        # Nur Alphabete verschl√ºsseln
        if char.isalpha():
            verschoben = ord(char) + verschiebung

            if char.islower():
                if verschoben > ord('z'):
                    verschoben -= 26
            elif char.isupper():
                if verschoben > ord('Z'):
                    verschoben -= 26

            verschluesselter_text += chr(verschoben)
        else:
            verschluesselter_text += char

    return verschluesselter_text

def entschluesseln(text, verschiebung):
    return verschluesseln(text, -verschiebung)

# Eingabe von Text und Verschiebung vom Benutzer
text = input("Bitte geben Sie den zu verschl√ºsselnden Text ein: ")
verschiebung = int(input("Bitte geben Sie die Verschiebung ein (ganze Zahl): "))

verschluesselter_text = verschluesseln(text, verschiebung)
print(f"Verschl√ºsselter Text: {verschluesselter_text}")

entschluesselter_text = entschluesseln(verschluesselter_text, verschiebung)
print(f"Entschl√ºsselter Text: {entschluesselter_text}")
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  L√∂sungsvorschlag</font> </br></p>
import string
from itertools import cycle

def vigenere_encode(text, key):
    return vigenere_cypher(text, key)

def vigenere_decode(text, key):
    return vigenere_cypher(text, key, mode="decode")

def vigenere_cypher(text, key, mode="encode"):
    result = []
    key_cycle = cycle(key)
    base = (string.printable) * 2

    for t in text:
        if t in base:
            k = next(key_cycle)
            t_index = base.index(t)
            k_index = base.index(k)

            if mode == "decode":
                new_index = (t_index - k_index)
            else:
                new_index = (t_index + k_index)

            result.append(base[new_index])
        else:
            result.append(t)
            next(key_cycle)

    return ''.join(result)

if __name__ == "__main__":
    text = input("Bitte geben Sie den Text ein: ")
    key = input("Bitte geben Sie den Schl√ºssel ein: ")
    mode = input("W√§hlen Sie den Modus (encode/decode): ")

    if not key:
        print("Der Schl√ºssel darf nicht leer sein!")
    else:
        match mode:
            case "encode":
                processed_text = vigenere_encode(text, key)
                print(f"Verschl√ºsselter Text: {processed_text}")
            case "decode":
                processed_text = vigenere_decode(text, key)
                print(f"Entschl√ºsselter Text: {processed_text}")
            case _:
                print("Ung√ºltiger Modus!")
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  L√∂sungsvorschlag Klasse Magier</font> </br></p>
import random

class Magier:
    def __init__(self, name):
        self.name = name
        self.lebenspunkte = 100
        self.mana = 50

    def zaubern(self, gegner):
        if self.mana >= 10:
            if random.random() < 0.8:
                gegner.lebenspunkte -= 20
                print(f"{self.name} üßô‚Äç‚ôÇÔ∏è zaubert einen Spruch ü™Ñ auf {gegner.name} und verursacht 20 Schaden!")
            else:
                print(f"{self.name}'s Zauber hat verfehlt!")
            self.mana -= 10
        else:
            self.normaler_angriff(gegner)

    def normaler_angriff(self, gegner):
        if random.random() < 0.6:  # 60% Trefferchance f√ºr den normalen Angriff des Magiers
            gegner.lebenspunkte -= 5
            print(f"{self.name} greift {gegner.name} normal an und verursacht 5 Schaden!")
        else:
            print(f"{self.name}'s Angriff hat verfehlt!")

    def info(self):
        print(f"{self.name} - Lebenspunkte: {self.lebenspunkte}, Mana: {self.mana}")

magier = Magier("Merlin")

magier.info()   # Merlin - Lebenspunkte: 100, Mana: 5
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  L√∂sungsvorschlag Klasse Amazone</font> </br></p>
class Amazone:
    def __init__(self, name):
        self.name = name
        self.lebenspunkte = 120
        self.pfeile = 5

    def schiessen(self, gegner):
        if self.pfeile > 0:
            if random.random() < 0.7:
                gegner.lebenspunkte -= 15
                print(f"{self.name} üßùüèª‚Äç‚ôÄÔ∏è schie√üt einen Pfeil üèπ auf {gegner.name} und verursacht 15 Schaden!")
            else:
                print(f"{self.name}'s Pfeil hat verfehlt!")
            self.pfeile -= 1
        else:
            self.normaler_angriff(gegner)

    def normaler_angriff(self, gegner):
        if random.random() < 0.65:  # 65% Trefferchance f√ºr den normalen Angriff der Amazone
            gegner.lebenspunkte -= 7
            print(f"{self.name} greift {gegner.name} normal an und verursacht 7 Schaden!")
        else:
            print(f"{self.name}'s Angriff hat verfehlt!")

    def info(self):
        print(f"{self.name} - Lebenspunkte: {self.lebenspunkte}, Pfeile: {self.pfeile}")

amazone = Amazone("Diana")

amazone.info()  # Diana - Lebenspunkte: 120, Pfeile: 5
`````

 + buch_zurueckgeben(buch): Diese Methode soll es einem Mitglied erm√∂glichen, ein Buch zur√ºckzugeben. Der Status des Buches wird auf False gesetzt, und das Buch wird aus der Liste der ausgeliehenen B√ºcher entfernt.

`````python
#@title
#@markdown   <p><font size="4" color='green'>  L√∂sungsvorschlag</font> </br></p>
class Buch:
    def __init__(self, titel, autor):
        self.titel = titel
        self.autor = autor
        self.ist_ausgeliehen = False

class Mitglied:
    def __init__(self, name):
        self.name = name
        self.ausgeliehene_buecher = []

    def buch_ausleihen(self, buch):
        if not buch.ist_ausgeliehen:
            buch.ist_ausgeliehen = True
            self.ausgeliehene_buecher.append(buch)
            print(f"{self.name} hat '{buch.titel}' ausgeliehen.")
        else:
            print(f"'{buch.titel}' ist bereits ausgeliehen.")

    def buch_zurueckgeben(self, buch):
        if buch in self.ausgeliehene_buecher:
            buch.ist_ausgeliehen = False
            self.ausgeliehene_buecher.remove(buch)
            print(f"{self.name} hat '{buch.titel}' zur√ºckgegeben.")
        else:
            print(f"{self.name} hat '{buch.titel}' nicht ausgeliehen.")

# Beispiel f√ºr die Nutzung des Systems
buch1 = Buch("Der kleine Prinz", "Antoine de Saint-Exup√©ry")
mitglied1 = Mitglied("Anna")

mitglied1.buch_ausleihen(buch1)  # Anna leiht das Buch aus
mitglied1.buch_zurueckgeben(buch1)  # Anna gibt das Buch zur√ºck
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  L√∂sungsvorschlag</font> </br></p>
class Buch:
    def __init__(self, titel, autor):
        self.titel = titel
        self.autor = autor
        self.ist_ausgeliehen = False

class Mitglied:
    def __init__(self, name, maximal_ausleihen=3):
        self.name = name
        self.ausgeliehene_buecher = []
        self.maximal_ausleihen = maximal_ausleihen  # Maximale Ausleihanzahl

    def buch_ausleihen(self, buch):
        if len(self.ausgeliehene_buecher) < self.maximal_ausleihen:
            if not buch.ist_ausgeliehen:
                buch.ist_ausgeliehen = True
                self.ausgeliehene_buecher.append(buch)
                print(f"{self.name} hat '{buch.titel}' ausgeliehen.")
            else:
                print(f"'{buch.titel}' ist bereits ausgeliehen.")
        else:
            print(f"{self.name} kann keine weiteren B√ºcher ausleihen, da das Limit erreicht ist.")

    def buch_zurueckgeben(self, buch):
        if buch in self.ausgeliehene_buecher:
            buch.ist_ausgeliehen = False
            self.ausgeliehene_buecher.remove(buch)
            print(f"{self.name} hat '{buch.titel}' zur√ºckgegeben.")
        else:
            print(f"{self.name} hat '{buch.titel}' nicht ausgeliehen.")

    def anzahl_ausgeliehene_buecher(self):
        return len(self.ausgeliehene_buecher)

class Bibliothek:
    def __init__(self):
        self.buecher = []

    def buch_hinzufuegen(self, buch):
        self.buecher.append(buch)
        print(f"'{buch.titel}' wurde zur Bibliothek hinzugef√ºgt.")

    def verfuegbare_buecher_anzeigen(self):
        verfuegbare_buecher = [buch for buch in self.buecher if not buch.ist_ausgeliehen]
        if verfuegbare_buecher:
            print("Verf√ºgbare B√ºcher:")
            for buch in verfuegbare_buecher:
                print(f"- {buch.titel} von {buch.autor}")
        else:
            print("Keine verf√ºgbaren B√ºcher.")

# Beispiel zur Verwendung
buch1 = Buch("Der kleine Prinz", "Antoine de Saint-Exup√©ry")
buch2 = Buch("1984", "George Orwell")
buch3 = Buch("Der Prozess", "Franz Kafka")
buch4 = Buch("Die Verwandlung", "Franz Kafka")

anna = Mitglied("Anna")
bibliothek = Bibliothek()

bibliothek.buch_hinzufuegen(buch1)
bibliothek.buch_hinzufuegen(buch2)
bibliothek.buch_hinzufuegen(buch3)
bibliothek.buch_hinzufuegen(buch4)

# Anzeige der verf√ºgbaren B√ºcher in der Bibliothek
bibliothek.verfuegbare_buecher_anzeigen()

# Anna leiht einige B√ºcher aus
anna.buch_ausleihen(buch1)
anna.buch_ausleihen(buch2)
anna.buch_ausleihen(buch3)

# Versuch, ein weiteres Buch auszuleihen (Limit √ºberschritten)
anna.buch_ausleihen(buch4)

# Anna gibt ein Buch zur√ºck und leiht ein anderes aus
anna.buch_zurueckgeben(buch1)
anna.buch_ausleihen(buch4)

# Anzeige der erneut verf√ºgbaren B√ºcher in der Bibliothek
bibliothek.verfuegbare_buecher_anzeigen()
`````

In diesem Abschlussprojekt sollen die Teilnehmer ein Verwaltungssystem f√ºr ein Fitnessstudio entwickeln. Das System soll es erm√∂glichen, Mitglieder zu verwalten, Kurse anzubieten und Buchungen f√ºr Kurse zu organisieren. Das Projekt ist so gestaltet, dass es die wichtigsten Konzepte der objektorientierten Programmierung (OOP) in Python abdeckt.

**Anforderungen:**
+ Klasse Mitglied
 + name: Der Name des Mitglieds.
mitgliedsnummer: Eine eindeutige Nummer f√ºr jedes Mitglied.
 + gebuchte_kurse: Eine Liste der Kurse, die das Mitglied gebucht hat.                  
 + kurs_buchen(kurs): Bucht das Mitglied f√ºr einen Kurs, wenn noch Pl√§tze verf√ºgbar sind.
 + kurs_stornieren(kurs): Storniert die Buchung eines Kurses.
+ Klasse Kurs:
 + name: Der Name des Kurses (z.B. "Yoga", "Zumba").
  + max_teilnehmer: Die maximale Anzahl an Teilnehmern, die an dem Kurs teilnehmen k√∂nnen.
  + teilnehmer_liste: Eine Liste der Mitglieder, die den Kurs gebucht haben.
  + teilnehmer_hinzufuegen(mitglied): F√ºgt ein Mitglied zur Teilnehmerliste hinzu, wenn der Kurs noch nicht voll ist.
  + teilnehmer_entfernen(mitglied): Entfernt ein Mitglied aus der Teilnehmerliste.
+ Klasse Fitnessstudio:
  + name: Der Name des Fitnessstudios.
  + mitglieder_liste: Eine Liste aller registrierten Mitglieder.
  + kurs_liste: Eine Liste aller angebotenen Kurse.
  + mitglied_hinzufuegen(mitglied): F√ºgt ein neues Mitglied zur Mitgliederliste hinzu.
  + kurs_hinzufuegen(kurs): F√ºgt einen neuen Kurs zur Kursliste hinzu.
  + verfuegbare_kurse_anzeigen(): Zeigt eine Liste aller Kurse mit freien Pl√§tzen an.

Zus√§tzliche Anforderungen:
Jedes Mitglied darf sich nur einmal pro Kurs anmelden.
Es soll m√∂glich sein, alle Mitglieder eines bestimmten Kurses anzuzeigen.
Es soll m√∂glich sein, alle Kurse anzuzeigen, an denen ein bestimmtes Mitglied teilnimmt.

`````python
#
# Platz f√ºr Ihren Code
#
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  L√∂sungsvorschlag</font> </br></p>
class Mitglied:
    def __init__(self, name, mitgliedsnummer):
        self.name = name
        self.mitgliedsnummer = mitgliedsnummer
        self.gebuchte_kurse = []

    def kurs_buchen(self, kurs):
        if kurs in self.gebuchte_kurse:
            print(f"{self.name} ist bereits f√ºr den Kurs '{kurs.name}' angemeldet.")
        else:
            if kurs.teilnehmer_hinzufuegen(self):
                self.gebuchte_kurse.append(kurs)
                print(f"{self.name} hat sich erfolgreich f√ºr den Kurs '{kurs.name}' angemeldet.")
            else:
                print(f"Der Kurs '{kurs.name}' ist bereits voll.")

    def kurs_stornieren(self, kurs):
        if kurs in self.gebuchte_kurse:
            kurs.teilnehmer_entfernen(self)
            self.gebuchte_kurse.remove(kurs)
            print(f"{self.name} hat die Buchung f√ºr den Kurs '{kurs.name}' storniert.")
        else:
            print(f"{self.name} ist nicht f√ºr den Kurs '{kurs.name}' angemeldet.")

class Kurs:
    def __init__(self, name, max_teilnehmer):
        self.name = name
        self.max_teilnehmer = max_teilnehmer
        self.teilnehmer_liste = []

    def teilnehmer_hinzufuegen(self, mitglied):
        if len(self.teilnehmer_liste) < self.max_teilnehmer:
            self.teilnehmer_liste.append(mitglied)
            return True
        else:
            return False

    def teilnehmer_entfernen(self, mitglied):
        if mitglied in self.teilnehmer_liste:
            self.teilnehmer_liste.remove(mitglied)

class Fitnessstudio:
    def __init__(self, name):
        self.name = name
        self.mitglieder_liste = []
        self.kurs_liste = []

    def mitglied_hinzufuegen(self, mitglied):
        self.mitglieder_liste.append(mitglied)
        print(f"Mitglied {mitglied.name} wurde hinzugef√ºgt.")

    def kurs_hinzufuegen(self, kurs):
        self.kurs_liste.append(kurs)
        print(f"Kurs '{kurs.name}' wurde hinzugef√ºgt.")

    def verfuegbare_kurse_anzeigen(self):
        print("Verf√ºgbare Kurse mit freien Pl√§tzen:")
        for kurs in self.kurs_liste:
            freie_plaetze = kurs.max_teilnehmer - len(kurs.teilnehmer_liste)
            if freie_plaetze > 0:
                print(f"- {kurs.name}: {freie_plaetze} freie Pl√§tze")

# Beispiel zur Verwendung des Systems
anna = Mitglied("Anna", 1)
ben = Mitglied("Ben", 2)

yoga = Kurs("Yoga", 2)
zumba = Kurs("Zumba", 3)

fitnessstudio = Fitnessstudio("Fit & Gesund")

# Mitglieder hinzuf√ºgen
fitnessstudio.mitglied_hinzufuegen(anna)
fitnessstudio.mitglied_hinzufuegen(ben)

# Kurse hinzuf√ºgen
fitnessstudio.kurs_hinzufuegen(yoga)
fitnessstudio.kurs_hinzufuegen(zumba)

# Verf√ºgbare Kurse anzeigen
fitnessstudio.verfuegbare_kurse_anzeigen()

# Mitglieder buchen Kurse
anna.kurs_buchen(yoga)
ben.kurs_buchen(yoga)

# Verf√ºgbare Kurse nach Buchung anzeigen
fitnessstudio.verfuegbare_kurse_anzeigen()

# Stornierung eines Kurses
anna.kurs_stornieren(yoga)

# Verf√ºgbare Kurse nach Stornierung anzeigen
fitnessstudio.verfuegbare_kurse_anzeigen()
`````

Schreiben Sie ein Programm das einen Blockchain erstellt. Ein Blockchain ist eine Kette von Datenbl√∂cken, bei denen jeder Block einen Verweis auf den vorherigen Block enth√§lt (meist in Form eines Hashes).

Es soll diese Transaktionen f√§lschungssicher speichern, also so, dass sie nachtr√§glich m√∂glichst schwer zu manipulieren sind.

Der "Block" enth√§lt jedweils die nachfolgenden Daten:

+ Index
+ alter hash (des vorigen Blocks)
+ Zeitstempel
+ Daten
+ neuer hash

Verwende die Hash-Funktion `sha256`.



```
import hashlib
.encode('utf-8')  # Methode konvertiert String in einen Byte-Array, wird von hashlib erwartet
hashlib.sha256()  # Funktion erzeugt ein Hash-Objekt
.hexdigest()      # Methode konvertiert das Hash-Objekts in einen hexadezimalen String
```


**Idee zum Vorgehen:**
1. Legen Sie zuerst die Blockchain als leere Liste an.
2. Erzeugen Sie den ersten Block, den "Genesis-Block". Verwenden Sie 'genesis' als vorangegangenen Hash.
3. Plazieren Sie einige zuf√§llige Transaktionen/Daten im Block.
3. F√ºgen Sie den fertigen Block an die Blockchain an.
4. Erzeugen Sie den zweiten bis x. Block.

Schwierigkeit: ‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê

`````python
#
# Hier bitte Ihren Code eingeben
#
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  L√∂sungsvorschlag</font> </br></p>
import hashlib
import time
from datetime import datetime
# --------------------------------------------------------
class Block:
    def __init__(self, index, previous_hash, timestamp, data, hash):
        self.index = index
        self.previous_hash = previous_hash
        self.timestamp = timestamp
        self.data = data
        self.hash = hash

    def __str__(self):
        return (f"Index: {self.index}\n"
                f"Previous Hash: {self.previous_hash}\n"
                f"Timestamp: {self.timestamp}\n"
                f"Data: {self.data}\n"
                f"Hash: {self.hash}\n")
# --------------------------------------------------------
# Ermittle Hash-Objekt
def calculate_hash(index, previous_hash, timestamp, data):
    return hashlib.sha256(f'{index}{previous_hash}{timestamp}{data}'.encode('utf-8')).hexdigest()
# Erstelle Genesis-Block
def create_genesis_block():
    timestamp = datetime.now().strftime('%Y-%m-%d-%H-%M-%S-%f')
    return Block(0, '0', timestamp, 'Genesis Block', calculate_hash(0, '0', timestamp, 'Genesis Block'))
# Erstelle neuen Folgeblock
def create_new_block(prev_block, data):
    index = prev_block.index + 1
    timestamp = datetime.now().strftime('%Y-%m-%d-%H-%M-%S-%f')
    hash = calculate_hash(index, prev_block.hash, timestamp, data)
    return Block(index, prev_block.hash, timestamp, data, hash)
# --------------------------------------------------------
# Erstelle die Blockchain und f√ºge den Genesis-Block hinzu
blockchain = [create_genesis_block()]
previous_block = blockchain[0]
# --------------------------------------------------------
# F√ºge 10 weitere Bl√∂cke zur Blockchain hinzu
num_of_blocks_to_add = 10
# --------------------------------------------------------
for i in range(1, num_of_blocks_to_add + 1):
    new_data = f"Block #{i} data"
    new_block = create_new_block(previous_block, new_data)
    blockchain.append(new_block)
    previous_block = new_block
    print(f"Block #{new_block.index} has been added to the blockchain!")
    print(f"Hash: {new_block.hash}\n")
# --------------------------------------------------------
# Beispiele
print(blockchain[0])  # Genesis-Block
print(blockchain[5])  # 5. Block
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  L√∂sungsvorschlag</font> </br></p>
#
# Version 1 - exklusive doc-strings
# Erstellt mit ChatGPT
#
# reading the file
with open(file_name) as fp:
    contents=fp.readlines()

# initialize two counter to check mismatch between "(" and ")"
open_bracket_counter=0
close_bracket_counter=0

# whenever an element deleted from the list length of the list will be decreased
decreasing_counter=0

for number in range(len(contents)):

    # checking if the line contains "#" or not
    if "#" in contents[number-decreasing_counter]:

        # delete the line if startswith "#"
        if contents[number-decreasing_counter].startswith("#"):
            contents.remove(contents[number-decreasing_counter])
            decreasing_counter+=1

        # delete the character after the "#"
        else:
            newline=""
            for character in contents[number-decreasing_counter]:
                if character=="(":
                    open_bracket_counter+=1
                    newline+=character
                elif character==")":
                    close_bracket_counter+=1
                    newline+=character
                elif character=="#" and open_bracket_counter==close_bracket_counter:
                    break
                else:
                    newline+=character
            contents.remove(contents[number-decreasing_counter])
            contents.insert(number-decreasing_counter,newline)

new_file_name = "new_" + file_name
# writing into a new file
with open(new_file_name,"w") as fp:
    fp.writelines(contents)
`````

`````python
#@title
#@markdown   <p><font size="4" color='green'>  L√∂sungsvorschlag</font> </br></p>
#
# Version 2 - inklusive doc-strings
# Erstellt mit ChatGPT
#
import io, tokenize, re
def remove_comments_and_docstrings(source):
    io_obj = io.StringIO(source)
    out = ""
    prev_toktype = tokenize.INDENT
    last_lineno = -1
    last_col = 0
    for tok in tokenize.generate_tokens(io_obj.readline):
        token_type = tok[0]
        token_string = tok[1]
        start_line, start_col = tok[2]
        end_line, end_col = tok[3]
        ltext = tok[4]
        if start_line > last_lineno:
            last_col = 0
        if start_col > last_col:
            out += (" " * (start_col - last_col))
        if token_type == tokenize.COMMENT:
            pass
        elif token_type == tokenize.STRING:
            if prev_toktype != tokenize.INDENT:
                if prev_toktype != tokenize.NEWLINE:
                    if start_col > 0:
                        out += token_string
        else:
            out += token_string
        prev_toktype = token_type
        last_col = end_col
        last_lineno = end_line
    out = '\n'.join(l for l in out.splitlines() if l.strip())
    return out


# reading the file
with open(file_name) as f_in:
  new_file_name = "new_" + file_name
  with open(new_file_name,"w") as f_out:
    f_out.writelines(remove_comments_and_docstrings(f_in.read()))
`````

Schwierigkeit: ‚≠ê‚≠ê‚≠ê‚≠ê

`````python
#@title
#@markdown   <p><font size="4" color='green'>  L√∂sungsvorschlag</font> </br></p>
import csv
import requests

from google.colab import drive
drive.mount('/content/drive')

# Pfad zur CSV-Datei
csv_path = '/content/drive/MyDrive/Python_Intro/02 data/urls.csv'

# URLs aus der CSV-Datei lesen
with open(csv_path, 'r') as file:
    reader = csv.reader(file)
    websites = [row[0] for row in reader]

# Ihr API-Schl√ºssel von screenshotlayer
ACCESS_KEY = 'key1234567890'

# Basis-URL f√ºr die screenshotlayer API
BASE_URL = "http://api.screenshotlayer.com/api/capture"

for i, url in enumerate(websites, 1):
    # Parameter f√ºr die API
    params = {
        'access_key': ACCESS_KEY,
        'url': url,
        'viewport': '1440x900',  # Sie k√∂nnen die Gr√∂√üe des Viewports nach Bedarf anpassen
        'fullpage': '1'  # Dies erstellt einen Screenshot der gesamten Webseite
    }

    response = requests.get(BASE_URL, params=params)

    with open(f'screenshot_{i}.png', 'wb') as file:
        file.write(response.content)

    print(f"Screenshot f√ºr {url} erstellt.")
`````

CadQuery ist eine skriptbasierte 3D-CAD-Modellierungspaket, die in Python geschrieben ist. Einige Hauptmerkmale und Vorteile von CadQuery sind:

`````python
#
# Code durch ChatGPT erstellt
#
def einkommensteuer_berechnen(zu_versteuerndes_einkommen, tarif):
    if tarif == "splitting":
        zu_versteuerndes_einkommen /= 2

    # Berechnung der Einkommensteuer gem√§√ü den Grenzen und Formeln
    if zu_versteuerndes_einkommen <= 10347:
        steuer = 0
    elif zu_versteuerndes_einkommen <= 62978:
        z = (zu_versteuerndes_einkommen - 10347) / 10000
        steuer = (995.21 * z + 1400) * z
    elif zu_versteuerndes_einkommen <= 300919:
        y = (zu_versteuerndes_einkommen - 62978) / 10000
        steuer = (216.16 * y + 2397) * y + 950
    else:
        steuer = 0.45 * zu_versteuerndes_einkommen - 17386

    if tarif == "splitting":
        steuer *= 2  # Steuer wird verdoppelt, da das Einkommen halbiert wurde

    return round(steuer, 2)  # Rundet die Steuer auf zwei Dezimalstellen
`````

`````python
# Beispiel f√ºr die Funktionsnutzung
einkommen_steuer_grundtarif = einkommensteuer_berechnen(50000, "grundtarif")
einkommen_steuer_splitting = einkommensteuer_berechnen(50000, "splitting")

einkommen_steuer_grundtarif, einkommen_steuer_splitting
`````
