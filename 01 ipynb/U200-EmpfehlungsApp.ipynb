```python
import gradio as gr
```

```python
buecher = {
    "Fantasy": "Der Herr der Ringe von J.R.R. Tolkien",
    "Sci-Fi": "Dune von Frank Herbert",
    "Krimi": "Sherlock Holmes von Arthur Conan Doyle",
    "Romantik": "Stolz und Vorurteil von Jane Austen"
}

def BuchEmpfehlen(Genre):
    return buecher.get(Genre, "Keine Empfehlung verfgbar")
```

```python
iface = gr.Interface(fn=BuchEmpfehlen,
    inputs=gr.Dropdown(list(buecher.keys())),
    outputs=gr.Text(label="Empfehlung:"),
    title="Buchempfehlungs-App",
    description="Whlen Sie ein Genre aus, um eine Buchempfehlung zu erhalten.",
    allow_flagging="never")
iface.launch()
```
