```python
# Lesen API Key
from google.colab import userdata
KEY = userdata.get('openai_key')
```

```python
# Erstellen eines Clients
import openai
from openai import OpenAI
client = openai.OpenAI(api_key=KEY)
```

```python
# Erstellen eines Chats
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
    {"role": "system", "content": "Du bist ein Python-Coding Assistent."},
    {"role": "user", "content": "Schreibe ein Code Snippet zur Kalkulation der Fibonacci-Zahlen. Kommentiere den Code."}
    ],
    temperature=0.7,
    max_tokens=300,
)
```

```python
type(response)
```

```python
# Request & Response
response
```

```python
for i in response:
    print(i)
```

```python
# Konvertieren in ein Dictionary für besseres Handling
result = response.model_dump()
result
```

```python
print(result['model'])
print("-"*20)
print(result['object'])
print("-"*20)
```
