## 5.2 |  Indizieren von Listen
---

**Listenindex 0**

Der Listenindex in Python beginnt bei 0 aus historischen und technischen Grnden, die mit der Speicheradressierung zusammenhngen.

Grnde fr das 0-basierte Indexieren:
+ Speicheradressierung: In Arrays entspricht der Index dem Offset zur Startadresse. Bei 0-basierter Indexierung liegt das erste Element an der Startadresse, was Berechnungen effizienter macht, da kein zustzlicher Offset bentigt wird.

+ Geschichte: Frhe Programmiersprachen wie C nutzen 0-basierte Indexierung, und Python hat dieses Konzept bernommen, um konsistent zu bleiben.

+ Mathematische Logik: In vielen mathematischen und algorithmischen Konzepten beginnt das Zhlen bei 0, was die Anwendung in der Programmierung vereinfacht.

**[Start:Ende:Schrittweite]**

+ Definiert einen Bereich (auch "Slice" genannt) innerhalb einer Sequenz (Liste oder String).
+ Start: Gibt den Index an, bei dem der Bereich beginnt (**inklusive**).
+ Ende: Gibt den Index an, bei dem der Bereich endet (**exklusive**).
+ Schrittweite: Bestimmt, wie viele Elemente beim Durchlaufen der Liste bersprungen werden
+ Kein Startindex: Slice beginnt am Anfang.
+ Kein Endindex: Slice geht bis zum Ende.
+ Keine Schrittweite: Alle Elemente
+ Kein Start- und kein Endindex: Liefert die gesamte Sequenz.

**Plus/Minus**

+ Plus: Eine positive Schrittweite bedeutet, dass man von **links nach rechts** durch die Sequenz iteriert.
- Minus: Eine negative Schrittweite bedeutet, dass man von **rechts nach links** durch die Sequenz iteriert.

**Index inklusive/exklusive**

In Python sind Indizes bei der Listindizierung so konzipiert, dass der Startindex inklusiv und der Endindex exklusiv ist. Diese Konvention bietet einige Vorteile:


+ **Einfache Berechnung der Lnge**: Wenn der Endindex exklusiv ist, ist die Lnge des Teils einer Liste oder eines Arrays einfach als `stop - start` zu berechnen. Zum Beispiel ergibt der Slice `list[1:4]` drei Elemente, und die Lnge ist `4 - 1 = 3`.

+ **Natrliches Teilen**: Das exklusive Ende macht das Aufteilen der Daten in Abschnitte einfacher und intuitiver. Wenn Sie eine Liste in zwei teilen mchten, knnen Sie einfach den mittleren Index als Endpunkt fr den ersten Abschnitt und als Startpunkt fr den zweiten Abschnitt verwenden, ohne einen Offset hinzuzufgen oder abzuziehen. Zum Beispiel, `list[:n]` und `list[n:]` teilt die Liste genau am Index `n`.

+ **Vermeidung von berlappungen und Lcken**: Da der Endindex exklusiv ist, gibt es keine berlappungen oder Lcken zwischen aufeinanderfolgenden Slices. Das bedeutet, dass beim Hintereinanderausfhren von Slices wie `list[:n]` und `list[n:]` jedes Element der Liste genau einmal erscheint.

+ **Einfachheit bei negativen Indizes**: Die Konvention hilft auch bei der Verwendung von negativen Indizes. Zum Beispiel gibt `list[-3:]` die letzten drei Elemente der Liste zurck, und Sie mssen nicht die Gesamtlnge der Liste berechnen, um den Endindex zu bestimmen.

+ **Konsistenz mit `range` und Slicing**: Die Syntax fr Slicing (`start:stop`) ist konsistent mit der `range()`-Funktion. Beispielsweise gibt `range(0, 3)` und `list[0:3]` beide den Bereich von 0 bis 2 zurck. Das erleichtert das Verstndnis und die Verwendung, da beide Mechanismen auf hnliche Weise arbeiten.

Diese Konventionen tragen dazu bei, Fehler zu vermeiden und die Lesbarkeit des Codes zu verbessern, da die Absichten des Codierenden klarer werden. Es macht die Indizierung intuitiver und die Programmierung in Python effizienter.

## 7.3 |  Exkurs: range()
---

Die Funktion range() in Python erzeugt eine Sequenz von Ganzzahlen (Integer-Werten).


`range(start, stop, step)`

Parameter:
+ start (optional): Der Startwert der Sequenz. Die Sequenz beginnt bei diesem Wert. Wenn kein Startwert angegeben wird, wird standardmig bei 0 begonnen. Der Startwert ist inklusiv, d.h., er wird in der Sequenz eingeschlossen.

+ stop (erforderlich): Der Endwert der Sequenz. Die Sequenz endet vor diesem Wert, d.h., der Endwert ist exklusiv und wird nicht eingeschlossen.

+ step (optional): Der Schrittwert, der die Differenz zwischen aufeinanderfolgenden Zahlen der Sequenz angibt. Der Standardwert ist 1. Ein negativer Schrittwert ermglicht das Iterieren in absteigender Reihenfolge, die in einer for-Schleife oder an anderen Stellen, an denen eine Sequenz bentigt wird, verwendet werden kann. range() ist eine eingebaute Funktion und gibt ein range-Objekt zurck, das ein Iterable ist.



Die Entscheidung, dass `start` bei der Funktion range in Python inklusiv und `stop` exklusiv ist, hat mehrere praktische Grnde, die auf Effizienz und Konsistenz abzielen:
+ Konsistenz mit Listenindizes: Python verwendet nullbasierte Indizes fr Listen und andere sequenzielle Datentypen. Wenn man eine Liste von Elementen durchluft, ist es blich, range(len(liste)) zu verwenden, um alle Indizes der Liste zu erhalten. Da die Obergrenze exklusiv ist, wird der Index des letzten Elements korrekt bercksichtigt, ohne dass eine Anpassung erforderlich ist.
+ Mathematische Notation: Die Notation [start:end) (inklusive Start, exklusive Ende) ist in der Mathematik weit verbreitet und wird oft bei Intervallen verwendet. Diese Notation macht es einfach, Bereiche zu kombinieren oder zu teilen, ohne dass sich Bereiche berschneiden oder Lcken entstehen3.
+ Vermeidung von Off-by-One-Fehlern: Durch die exklusive Obergrenze wird das Risiko von Off-by-One-Fehlern reduziert, da man sich keine Gedanken darber machen muss, den Endpunkt explizit zu reduzieren oder anzupassen.
+ Flexibilitt bei der Bereichserstellung: Mit einer exklusiven Obergrenze kann man leicht Teilbereiche erstellen oder Bereiche aneinanderreihen. Zum Beispiel kann man zwei Bereiche range(a, b) und range(b, c) nahtlos aneinanderfgen, ohne dass eine Anpassung notwendig ist.

Diese Designentscheidung erleichtert viele Programmieraufgaben und sorgt fr eine konsistente Handhabung von Bereichen in Python.

`````python
#@title
#@markdown   <p><font size="4" color='green'>  Code fr Grafik</font> </br></p>
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Definition der Mengen
A = {"Apfel", "Banane", "Kirsche", "Birne"}
B = {"Feige", "Aprikose", "Kirsche", "Birne", "Kiwi", "Orange"}

# Funktion um alle Kreise des Venn-Diagramms mit einem Rahmen zu versehen
def add_all_patch_borders_to_venn(venn_diagram, ax):
    for region in venn_diagram.id2idx.keys():
        try:
            path = venn_diagram.get_patch_by_id(region).get_path()
            patch = plt.PathPatch(path, edgecolor='black', lw=1)
            ax.add_patch(patch)
        except:
            pass

# Erstellung eines Plots mit 4 Subplots nebeneinander
fig, axs = plt.subplots(1, 4, figsize=(20, 5))

# Vereinigungsmenge
v = venn2(subsets=(len(A - B), len(B - A), len(A & B)), set_labels=('A', 'B'), ax=axs[0])
add_all_patch_borders_to_venn(v, axs[0])
axs[0].set_title('Vereinigungsmenge (A  B)')

# Schnittmenge
v = venn2(subsets=(len(A - B), len(B - A), len(A & B)), set_labels=('A', 'B'), ax=axs[1])
add_all_patch_borders_to_venn(v, axs[1])
v.get_label_by_id('10').set_text('')
v.get_label_by_id('01').set_text('')
axs[1].set_title('Schnittmenge (A  B)')

# Differenzmenge
v = venn2(subsets=(len(A - B), len(B - A), len(A & B)), set_labels=('A', 'B'), ax=axs[2])
add_all_patch_borders_to_venn(v, axs[2])
v.get_label_by_id('01').set_text('')
v.get_label_by_id('11').set_text('')
axs[2].set_title('Differenzmenge (A - B)')

# Symmetrische Differenzmenge
v = venn2(subsets=(len(A - B), len(B - A), len(A & B)), set_labels=('A', 'B'), ax=axs[3])
add_all_patch_borders_to_venn(v, axs[3])
v.get_label_by_id('11').set_text('')
axs[3].set_title('Symmetrische Differenzmenge (A  B)')

# Anzeige des Plots
plt.tight_layout()
plt.show()
`````

# 11 |  Higher-Order-Functions (HOFs)

Higher-Order Functions sind Funktionen, die andere Funktionen als Argumente annehmen oder Funktionen als Ergebnis zurckgeben. map und filter sind klassische Beispiele dafr, da beide eine Funktion als eines ihrer Argumente nehmen. Beide Funktionen geben einen Iteratoren zurck. Ein Iterator in Python ist ein Objekt, das eine Sequenz von Werten nacheinander durchlaufen kann. Werte Werte werden nach und nach bereitgestellt. Es erfolgt eine sog. *Lazy Evaluation*.

**Lazy Evaluation**, oder **trges Auswerten**, bedeutet, dass der Wert eines Ausdrucks nur berechnet wird, **wenn er bentigt wird**. Dies kann die Effizienz verbessern, insbesondere bei der Verarbeitung groer Datenstze oder bei der Erstellung von potenziell unendlichen Datenstrukturen. ```map``` und ```filter``` in Python geben einen Iteratoren zurck, die diesem Prinzip folgen, da Elemente einzeln und nur auf Anfrage bearbeitet werden.

Die Ergebnisse eines Iterators lassen sich z.B. mit list(), zhlergesteuerte Schleife, next() oder Verwendung von '*' Operator anzeigen.

| Eigenschaft                  | Eager Evaluation                                   | Lazy Evaluation                           |
|------------------------------|---------------------------------------------------|-------------------------------------------|
| Wann werden die Elemente generiert? | Sofort (alle auf einmal)                           | On demand (bei Bedarf)                     |
| Speicherverbrauch             | Hher (alle Elemente werden im Voraus gespeichert) | Gering (nur das aktuell bentigte Element) |
| Beispiele in Python           | Listen                                    | range(), map(), filter()                   |

Technisch wre es mglich gewesen, OOP in Python ohne die explizite Verwendung von self zu gestalten, wie es in manchen anderen Sprachen der Fall ist.

Allerdings wrde das:

+ Weniger Flexibilitt bieten: Durch die explizite Verwendung von self kann eine Methode auch von anderen Instanzen oder **ohne Objektbezug** aufgerufen werden, was zustzliche Mglichkeiten schafft.
+ Die Philosophie von Python verletzen: Python legt groen Wert auf expliziten und leicht verstndlichen Code. Ein verstecktes oder implizites self wrde diese Philosophie untergraben.

**Fazit:**
Obwohl es theoretisch mglich gewesen wre, OOP in Python ohne self zu gestalten, entspricht die explizite Verwendung von self der Philosophie von Python und trgt zur Klarheit und Flexibilitt des Codes bei.

`````python
class Person():
    def __init__(self, name):
        self.name = name

    def sage_hallo(self):
        print("Hallo, mein Name ist ", self.name)

p1 = Person("Alice")
p2 = Person("Bob")

p1.sage_hallo()
`````

**Sollte man das tun?**

Technisch mglich, aber nicht empfohlen. Hier sind einige Grnde, warum die Verwendung von self vorzuziehen ist:

+ Konvention und Lesbarkeit:
`self` ist der allgemein akzeptierte Standard in der Python-Community. Jeder Python-Programmierer versteht sofort, dass self die Instanz der Klasse innerhalb einer Methode darstellt. Wenn man stattdessen `instanz` oder einen anderen Namen verwendest, knnte das den Lesefluss fr andere Entwickler beeintrchtigen.
+ Wartbarkeit:
Wenn man `self` verwendet, bleibt der Code konsistent mit dem, was in der gesamten Python-Community blich ist. Dies erleichtert die Wartung des Codes durch andere Entwickler in der Zukunft.
+ Verwirrung vermeiden:

`````python
%%writefile modul.py
#
# --- erste Funktion
#
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]

    fib_sequence = [0, 1]
    for i in range(2, n):
        next_fib = fib_sequence[-1] + fib_sequence[-2]
        fib_sequence.append(next_fib)
    return fib_sequence
#
# --- zweite Funktion
#
def ist_primzahl(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
`````

## 15.1 |     csv/txt Dateien (Python)
---

Umfangreicher Ein-/Ausgaben mssen ber Dateien realisiert werden. Die Funktion open() liefert eine Referenz auf eine Datei. Zudem muss die Zugriffsart definiert erden, z.B. "w" fr schreibenden und "r" fr lesenden Zugriff. Mit dem Mode "a", knnen neue Daten an eine bestehende Datei angehngt werden. Nach Benutzung der Datei muss dieses mit der Funktion close() geschlossen werden.

Die Hauptmodi und ihre Bedeutungen:

+ r: ffnet die Datei zum Lesen. Dies ist der Standardmodus, wenn kein Modus angegeben ist. Die Datei muss existieren, sonst wird eine Ausnahme ausgelst.

+ w: ffnet die Datei zum Schreiben. Wenn die Datei bereits existiert, wird ihr Inhalt gelscht, bevor mit dem Schreiben begonnen wird. Wenn die Datei nicht existiert, wird sie erstellt.

+ a: ffnet die Datei zum Anhngen. Wenn die Datei existiert, wird der neue Inhalt an das Ende der Datei angehngt, ohne den bestehenden Inhalt zu lschen. Wenn die Datei nicht existiert, wird sie erstellt.

Mit `with`: Der `with`-Befehl in Python wird verwendet, um das automatische ffnen und Schlieen von Dateien durchzufhren.

`````python
#
# ------ ffne die Datei im Modus 'r+' (Lesen und Schreiben), Datei wird nach Ausfhrung "with" (Kontextmanager, siehe unten) geschlossen
#
with open(dateiname, 'r+') as datei:
    # Lies den Inhalt der Datei
    Inhalt = datei.read()
    print('Aktueller Inhalt:\n', "-" * 40, "\n", Inhalt)

    # Gehe zum Ende der Datei - 0: Offset, 2. Parameter: 0 (Anfang der Datei), 1 (aktuelle Position) und 2 (Ende der Datei)
    datei.seek(0,2)

    # Schreibe weitere Datenstze in die Datei
    for i in range(10):
        datei.write(f"{generate_fake_person()}\n")

    # Optional: Gehe zum Anfang der Datei, um den neuen Inhalt zu lesen
    datei.seek(0)
    NeuerInhalt = datei.read()
    print('Aktueller Inhalt:\n', "-"*40, "\n", NeuerInhalt)
`````

`````python
# Variante 5: Fehlerverfolgung
#
import traceback
import sys

try:
    programm()
except Exception:
    # exc_info() liefert ein Tupel mit (Typ, Wert, Traceback-Objekt)
    exc_type, exc_value, exc_traceback = sys.exc_info()

    # Traceback-Informationen extrahieren
    filename, line_number, function_name, text = traceback.extract_tb(exc_traceback)[-1]

    print(f"Ein Fehler ist in der Datei '{filename}' aufgetreten.")
    print(f"Fehler in Zeile: {line_number}")
    print(f"In der Funktion: '{function_name}'")
    print(f"Ausgefhrter Code: '{text}'")
    print(f"Fehlertyp: {exc_type.__name__}")
    print(f"Fehlermeldung: {exc_value}")
`````

## 20.2 |  Vernderlich vs Unvernderlich
---


In Python gibt es zwei Arten von Variablen: mutable (vernderliche) und immutable (unvernderliche).


+ Mutable (vernderlich): Objekte, deren Inhalt gendert werden kann, ohne dass eine neue Referenz (Speicheradresse) erstellt wird.
+ Immutable (unvernderlich): Objekte, deren Inhalt nach ihrer Erstellung nicht mehr gendert werden kann. Jede nderung erzeugt ein neues Objekt (ggf. mit gleichem Variablennamen) mit einer neuen Referenz (Speicheradresse).

| Datentyp        | Immutable/Mutable   | Beispiel       |
|-----------------|---------------------|----------------|
| `int`           | Immutable           | `x = 1`        |
| `float`         | Immutable           | `x = 1.0`      |
| `str`           | Immutable           | `x = "Hello"`  |
| `tuple`         | Immutable           | `x = (1, 2)`   |
| `list`          | Mutable             | `x = [1, 2]`   |
| `set`           | Mutable             | `x = {1, 2}`   |
| `dict`          | Mutable             | `x = {'a': 1}` |



**Warum ist das wichtig?**

+ Speicherverwaltung: Bei unvernderlichen Objekten kann Python den Speicher effizienter nutzen, indem es dasselbe Objekt an mehreren Stellen verwendet, da sicher ist, dass es nicht verndert wird.
+ Fehlervermeidung: Wenn man wei, welche Objekte vernderlich sind, kann man ungewollte nderungen in Funktionen vermeiden, die Parameter modifizieren.
+ Performance: Bei vernderlichen Objekten kann der Speicher an Ort und Stelle gendert werden, was in manchen Situationen effizienter ist.

`````python
# --- Beispiel immutable
a = 3
b = a  # b verweist auf Objekt a mit Inhalt 3

print("a:", a, id(a))
print("b:", b, id(b))
`````

`````python
a = 4  # a verweist nun auf eine neues Objekt mit neuem Inhalt 4, b verweist weiterhin auf 3

print("a:", a, id(a))
print("b:", b, id(b))
`````

`````python
# --- Beispiel mutable
a = [3]
b = a        # b verweist jetzt auch auf dieselbe Liste wie a

print("a:", a, id(a))
print("b:", b, id(b))
`````

`````python
a[0] = 4    # Eine nderung an der Liste a wirkt auch bei b; die Verweise ndern sich nicht

print("a:", a, id(a))
print("b:", b, id(b))
`````

### 21.1.1 |  Generator







In der Informatik ist ein Generator eine Routine, mit der das Iterationsverhalten einer Schleife gesteuert werden kann. Alle Generatoren sind auch Iteratoren. Man kann sich einen Generator also vorstellen wie eine Funktion, die mit jedem Aufruf nur das nchste Element einer Liste zurckgibt. Kann der Generator keinen weiteren Rckgabewert liefern, wird ein StopIteration-Error ausgelst.

Der Vorteil von Generatoren gegenber Listen liegt darin, dass auch bei sehr groen Datenmengen kein groer Speicherbedarf ntig ist, da immer nur das jeweils aktuell zurckgegebene Objekt im Speicher existiert. Beispielsweise handelt es sich auch bei File-Objekten um Generatoren, die beim Aufruf von ``next(object)`` jeweils die nchste Zeile der geffneten ausgeben. Auf diese Weise kann der Inhalt einer (beliebig) groen Datei beispielsweise mittels einer for-Schleife abgearbeitet werden, ohne dass der Inhalt der Datei auf einmal eingelesen und als Liste gespeichert werden muss.

### 21.1.2 |  Decorator



Ein Dekorator ist im Wesentlichen eine Funktion, die eine andere Funktion als Argument nimmt und eine neue Funktion zurckgibt. Diese neue Funktion enthlt die **ursprngliche** Funktionalitt sowie die **hinzugefgte** Funktionalitt des Dekorators. Die grundlegende Syntax eines Dekorators in Python verwendet das `@`-Symbol, gefolgt vom Namen des Dekorators, direkt ber der zu dekorierenden Funktion.

**Vorteile von Dekoratoren:**

+ Reduzierung von Boilerplate-Code: Dekoratoren helfen dabei, wiederkehrende Muster und redundanten Code zu vermeiden, indem sie hufig bentigte Funktionalitten kapseln und wiederverwendbar machen.
+ Trennung von Anliegen: Sie ermglichen eine klare Trennung von Kernfunktionalitt und zustzlichen Aspekten wie Logging, Caching oder Authentifizierung.
+ Verbesserte Lesbarkeit und Wartbarkeit: Durch die Verwendung von Dekoratoren wird der Code bersichtlicher und leichter zu warten, da die zustzliche Funktionalitt klar und explizit an einer Stelle definiert ist.

`````python
from datetime import datetime
import time

def mein_dekorator(funktion):
    def einbettung(*args, **kwargs):
        now = datetime.now()
        print(f"----- Vor Aufruf Funktion {funktion.__name__} um {now.strftime('%H:%M:%S')}")
        ergebnis = funktion(*args, **kwargs)
        now = datetime.now()
        print(f"----- Nach Aufruf Funktion {funktion.__name__} um {now.strftime('%H:%M:%S')}")
        return ergebnis
    return einbettung
`````

`````python
@mein_dekorator
def add(a, b):
    time.sleep(2)
    print(f"Ergebnis add: {a+b}")
# ---
@mein_dekorator
def sub(a, b):
    time.sleep(2)
    print(f"Ergebnis sub: {a-b}")
`````

`````python
#
# Beispiel: Dekorator-Funktion, die die Ausfhrungszeit einer Funktion misst
#
import time

def timing_decorator(func):
    # Wrapper-Funktion, die die eigentliche Funktion umschliet
    def wrapper(*args, **kwargs):
        start_time = time.time()        # Startzeit aufzeichnen
        result = func(*args, **kwargs)  # Die dekorierte Funktion mit allen Argumenten aufrufen
        end_time = time.time()          # Endzeit aufzeichnen
        # Ausgabe der Ausfhrungszeit der Funktion
        print(f"Funktion {func.__name__} ausgefhrt in: {end_time - start_time:.7f} Sekunden.")
        return result                   # Ergebnis der Funktion zurckgeben

    return wrapper                      # Wrapper-Funktion zurckgeben
`````

`````python
# Anwendung des Dekorators auf die Addiere-Funktion
@timing_decorator
def Addiere(*args):
    return sum(args)

# Die dekorierte Addiere-Funktion aufrufen und das Ergebnis ausgeben
print(Addiere(1, 2, 3, 4, 5, 6, 7))
`````

`````python
# Anwendung des Dekorators auf die Fibonacci-Funktion
@timing_decorator
def fibonacci(n):
    # Initialisiere die ersten beiden Terme
    a, b = 0, 1
    # Initialisiere die Liste mit dem ersten Term
    sequence = [a]
    # Fhre die Iteration n-1 mal durch, da der erste Term bereits in der Liste ist
    for i in range(n-1):
        # Aktualisiere a und b auf die nchsten zwei Terme
        a, b = b, a + b
        # Fge den neuen Term der Liste hinzu
        sequence.append(a)
    return sequence

# Die dekorierte Fibonacci-Funktion aufrufen und das Ergebnis ausgeben
print(fibonacci(100)[:10], " ...")
`````

Die Fibonacci-Folge ist die unendliche Folge natrlicher Zahlen, die (ursprnglich) mit zweimal der Zahl 1 beginnt oder (hufig, in moderner Schreibweise) zustzlich mit einer fhrenden Zahl 0 versehen ist. Im Anschluss ergibt jeweils die Summe zweier aufeinanderfolgender Zahlen die unmittelbar danach folgende Zahl.

`````python
# Beispiel: iterative Funktion
def fib_i(n):
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return a

# Beispiel: rekursive Funktion
def fib_r(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib_r(n-1) + fib_r(n-2)

# Beispiel: rekursive Funktion mit Memo-Dict
memo = {0:0, 1:1}
def fib_m(n):

    if not n in memo:
        memo[n] = fib_m(n-1) + fib_m(n-2)
    return memo[n]
`````

`````python
import time

n = 30
fib_i_start = time.time(); fib_i(n); fib_i_dauer = time.time() - fib_i_start
fib_r_start = time.time(); fib_r(n); fib_r_dauer = time.time() - fib_r_start
fib_m_start = time.time(); fib_m(n); fib_m_dauer = time.time() - fib_m_start

print(f"iterativ : {fib_i_dauer:.6f} Sek")
print(f"rekursiv : {fib_r_dauer:.6f} Sek")
print(f"rekursiv+: {fib_m_dauer:.6f} Sek")
`````

### 21.3.2 |  Threading



Gut fr **I/O-intensive Aufgaben**. In Python jedoch durch den Global Interpreter Lock (GIL) in Bezug auf CPU-intensive Aufgaben eingeschrnkt.


`````python
import threading
import time

def rechenintensive_aufgabe(name):
    print(f"{name} startet")
    time.sleep(1)
    print(f"{name} ist fertig")

start_time = time.time()

thread1 = threading.Thread(target=rechenintensive_aufgabe, args=("Aufgabe 1",))
thread2 = threading.Thread(target=rechenintensive_aufgabe, args=("Aufgabe 2",))

thread1.start()
thread2.start()

thread1.join()
thread2.join()

print(f"Threading Ausfhrungsdauer: {(time.time() - start_time):.5f} Sekunden")
`````

### 21.3.3 |  Multiprocessing



Umgeht den GIL und eignet sich gut fr **CPU-intensive Aufgaben**, hat aber einen hheren Overhead fr Prozesserstellung und -kommunikation.


`````python
from multiprocessing import Process
import time

def rechenintensive_aufgabe(name):
    print(f"{name} startet")
    time.sleep(1)  # Simuliert eine rechenintensive Aufgabe
    print(f"{name} ist fertig")

start_time = time.time()

# Erstelle Prozesse fr parallele Ausfhrung
prozesse = [
    Process(target=rechenintensive_aufgabe, args=("Aufgabe 1",)),
    Process(target=rechenintensive_aufgabe, args=("Aufgabe 2",))
]

# Starte alle Prozesse
for p in prozesse:
    p.start()

# Warte, bis alle Prozesse abgeschlossen sind
for p in prozesse:
    p.join()

print(f"Parallelverarbeitung Ausfhrungsdauer: {time.time() - start_time} Sekunden")
`````

## 22.1 |  Trigger auf Attribute

`````python
class Troll():
    def __init__(self, punkte):
        self._lebenspunkte = punkte

    def check_lebenspunkte(self):
        if self._lebenspunkte < 0:  # Angenommener Schwellenwert
            print(f"Stand Lebenspunkte: {self._lebenspunkte}! Der Troll ist tot.")

    @property                       # Holen des Wertes (Getter)
    def punkte(self):
        print(f"1. Getter: {self._lebenspunkte}")
        return self._lebenspunkte

    @punkte.setter                  # Behandeln des Wertes (Setter)
    def punkte(self, neu_punkte):
        print(f"2. Setter: {neu_punkte}")
        self._lebenspunkte = neu_punkte
        self.check_lebenspunkte()
`````

Ein ```Dekorator``` ist eine Zusatz, der einem Attribut oder einer Funktion neue Fhigkeiten oder Verhaltensweisen hinzufgt.

Der ```@property```-Dekorator erlaubt es, den Wert eines Attributs zu bekommen, um den Wert zu berprfen oder zu modifizieren.

Der ```@lebenspunkte.setter```-Dekorator erlaubt es, dass eine Methode aufgerufen wird, der es erlaubt, den Wert eines Attributs zu setzen und/oder dabei eine zustzliche Logik ausfhrt, wie z.B. den Wert zu validieren oder zu modifizieren, bevor er tatschlich gesetzt wird.

Der ```@property```-Dekorator macht lebenspunkte zu einer berechneten Eigenschaft, die beim Lesen den Wert von _lebenspunkte zurckgibt. Der ```@lebenspunkte.setter```-Dekorator ermglicht es dann, eine Methode zu definieren, die aufgerufen wird, wenn der Wert von lebenspunkte gendert wird. Diese Methode erlaubt es, zustzliche Logik auszufhren, wie in diesem Fall die berprfung der Lebenspunkte.

`````python
# Beispielverwendung
Grindel = Troll(25)
`````

`````python
# Trigger auslsen
Grindel.punkte -= 20
`````

**Variante 2:**

In dem genderten Code wird ein neues *privates Attribut* _ist_tot eingefhrt, das dazu dient, den Zustand des Trolls zu speichern, ob er tot ist oder nicht. Die ```check_lebenspunkte```-Methode setzt die lebenspunkte auf 0, wenn sie negativ werden, und ndert den Zustand auf tot, wenn dies noch nicht geschehen ist. Der ```Setter``` berprft diesen Zustand und erlaubt keine nderungen an den lebenspunkten, wenn der Troll bereits tot ist.
Durch diese nderungen wird das lebenspunkte-Attribut schreibgeschtzt, sobald der Troll "tot" ist, d.h. seine Lebenspunkte den Schwellenwert von 0 erreicht oder unterschritten haben

**Info:**
Ein privates Attribut in einer Klasse ist ein Attribut, das nicht direkt von auerhalb der Klasse zugegriffen oder gendert werden *soll*. In Python gibt es *keine* strenge Zugriffskontrolle, die einen Zugriff verhindet, aber es gibt Konventionen, um private Attribute zu kennzeichnen.
In Python wird ein Attribut als privat betrachtet, wenn sein Name mit einem Unterstrich (_) beginnt.

`````python
class Troll():
    def __init__(self, punkte):
        self._lebenspunkte = punkte
        self._ist_tot = False  # Neues Attribut, um den Zustand zu speichern

    def check_lebenspunkte(self):
        if self._lebenspunkte < 0:
            self._lebenspunkte = 0  # Setzt die Lebenspunkte auf 0, falls unterschritten
            if not self._ist_tot:
                print(f"Stand Lebenspunkte: {self._lebenspunkte}! Der Troll ist tot.")
                self._ist_tot = True  # Setzt den Zustand auf tot

    @property
    def punkte(self):
        print(f"1. Getter: {self._lebenspunkte}")
        return self._lebenspunkte

    @punkte.setter
    def punkte(self, neu_punkte):
        print(f"2. Setter: {neu_punkte}")
        if not self._ist_tot:  # Erlaubt nderungen nur, wenn der Troll nicht tot ist
            self._lebenspunkte = neu_punkte
            self.check_lebenspunkte()
        else:
            print("nderungen nicht erlaubt, der Troll ist bereits tot.")
`````

## 22.3 |  Abstrakte Klassen
---


Abstrakte Klassen dienen dazu, unvollstndige Klassen zu definieren, die als Basis fr andere Klassen dienen, die bestimmte Methoden implementieren mssen.

In Python werden abstrakte Klassen mit dem Modul abc (Abstract Base Classes) implementiert. Dies ermglicht es, Klassen zu definieren, die nicht direkt instanziiert werden knnen, und Methoden, die in abgeleiteten Klassen implementiert werden mssen.

Eine abstrakte Klasse Auto definiert die Struktur fr abgeleitete Klassen, indem sie sicherstellt, dass bestimmte Methoden in jeder abgeleiteten Klasse implementiert werden.

`````python
from abc import ABC, abstractmethod

class Auto(ABC):
    def __init__(self, marke):
        self.Marke = marke

    @abstractmethod
    def Start(self):
        pass

    @abstractmethod
    def Stop(self):
        pass

class ElektroAuto(Auto):
    def Start(self):
        print(f"{self.Marke} Elektroauto startet leise.")

    def Stop(self):
        print(f"{self.Marke} Elektroauto hlt an.")

class VerbrennungsAuto(Auto):
    def Start(self):
        print(f"{self.Marke} Verbrennungsauto startet mit Motorgerusch.")

    def Stop(self):
        print(f"{self.Marke} Verbrennungsauto hlt an.")
`````

`````python
# Nutzung Tesla
meinElektroAuto = ElektroAuto("Tesla")
meinElektroAuto.Start()
meinElektroAuto.Stop()
`````

`````python
# Nutzung Volkswagen
meinVerbrennungsAuto = VerbrennungsAuto("Volkswagen")
meinVerbrennungsAuto.Start()
meinVerbrennungsAuto.Stop()
`````

## 22.4 |  Komposition und Aggregation
---

Beides sind Konzepte, die verwendet werden, um komplexe Typen durch Kombination von Objekten einfacherer Typen zu bilden. Bei der Komposition ist das Leben der enthaltenen Objekte von dem des Containerobjekts abhngig, whrend bei der Aggregation die enthaltenen Objekte unabhngig vom Containerobjekt existieren knnen.


### 22.4.1 |  Komposition
---

Bei der Komposition ist ein Objekt Eigentmer eines anderen Objekts. Wenn das Eigentmerobjekt zerstrt wird, wird auch das Eigentumsobjekt zerstrt. Im Automobilkontext knnte das Auto aus mehreren Komponenten bestehen, wie z.B. einem Motor. Wenn das Auto zerstrt wird, existiert der Motor nicht mehr unabhngig.

`````python
class Motor():
    def __init__(self, typ):
        self.Typ = typ

    def ZeigeDetails(self):
        print(f"Motor Typ: {self.Typ}")

class Auto():
    def __init__(self, marke, motorTyp):
        self.Marke = marke
        self.Motor = Motor(motorTyp)  # Komposition

    def ZeigeDetails(self):
        print(f"Auto Marke: {self.Marke}")
        self.Motor.ZeigeDetails()
`````

### 22.4.2 |  Aggregation
---


Bei der Aggregation hat ein Objekt ein oder mehrere andere Objekte, aber diese Objekte knnen ohne das Eigentmerobjekt existieren. Zum Beispiel knnte ein Auto mehrere Fahrer haben, aber die Fahrer existieren unabhngig vom Auto.

`````python
class Fahrer():
    def __init__(self, name):
        self.Name = name

    def ZeigeDetails(self):
        print(f"Fahrer Name: {self.Name}")

class Auto():
    def __init__(self, marke):
        self.Marke = marke
        self.Fahrer = []

    def FahrerHinzufuegen(self, fahrer):
        self.Fahrer.append(fahrer)

    def ZeigeDetails(self):
        print(f"Auto Marke: {self.Marke}")
        for fahrer in self.Fahrer:
            fahrer.ZeigeDetails()
`````

`````python
# Beispiel Nutzung
fahrer1 = Fahrer("Max")
fahrer2 = Fahrer("Anna")
meinAuto = Auto("Volkswagen")
meinAuto.FahrerHinzufuegen(fahrer1)
meinAuto.FahrerHinzufuegen(fahrer2)
`````

Das Kompositum-Muster ermglicht die einheitliche Behandlung von Einzelobjekten und Zusammensetzungen von Objekten. Hier modellieren wir Autoteile, die sowohl einzeln als auch als Teil eines greren Ganzen betrachtet werden knnen.

`````python
class AutoTeil():
    def ZeigeDetails(self):
        pass

class Einzelteil(AutoTeil):
    def __init__(self, name):
        self.Name = name

    def ZeigeDetails(self):
        print(f"Einzelteil: {self.Name}")

class Zusammensetzung(AutoTeil):
    def __init__(self, name):
        self.Name = name
        self.Teile = []

    def Hinzufuegen(self, teil):
        self.Teile.append(teil)

    def ZeigeDetails(self):
        print(f"Zusammensetzung: {self.Name}")
        for teil in self.Teile:
            teil.ZeigeDetails()
`````

Mehrfachvererbung ist ein Merkmal einiger objektorientierter Programmiersprachen, bei dem **eine Klasse** direkt von **mehr als einer Basisklasse** erben kann. Das bedeutet, dass eine Unterklasse die Methoden und Attribute von mehreren Oberklassen erben und nutzen kann. Mehrfachvererbung ermglicht es, sehr flexible und mchtige Hierarchien zu erstellen, bringt aber auch Komplexitt und potenzielle Probleme mit sich, wie z.B. das *Diamant-Problem*, bei dem dieselbe Basisklasse ber verschiedene Pfade mehrfach geerbt wird, was zu Unklarheiten darber fhren kann, von welcher Basisklasse Methoden und Attribute geerbt werden.

### 22.6.1 |  Vererbung
---


`````python
class Auto():
  def __init__(self, hersteller, modell, leistung): # Methode zur Erstellung des Blaupause / Bauplan der Klasse
    self.hersteller = hersteller
    self.modell = modell
    self.leistung = leistung

  def __str__(self):
    return f"Hersteller: {self.hersteller}, Modell: {self.modell}, Leistung: {self.leistung}"
`````

`````python
class Antrieb():
  def __init__(self, kraftstoff, antriebsart): # Methode zur Erstellung des Blaupause / Bauplan der Klasse
    self.kraftstoff = kraftstoff
    self.antriebsart = antriebsart

  def __str__(self):
    return f"Kraftstoff: {self.kraftstoff}, Antrieb: {self.antriebsart}"
`````

`````python
class Kompakt(Auto, Antrieb):
  def __init__(self, hersteller, modell, leistung, kraftstoff, antriebsart):
    Auto.__init__(self, hersteller, modell, leistung)
    Antrieb.__init__(self, kraftstoff, antriebsart)

  def __str__(self):
    return Auto.__str__(self) + ", " + Antrieb.__str__(self)
`````

### 22.6.2 |  Method Resolution Order
---


Die Methodenauflsungsreihenfolge (MRO) gehrt zu den fortgeschrittenen Konzepten der objektorientierten Programmierung (OOP), insbesondere im Zusammenhang mit der Mehrfachvererbung. MRO bestimmt die Reihenfolge, in der Basisklassen durchsucht werden, wenn Methoden oder Attribute in einer Klassenhierarchie aufgerufen oder abgefragt werden. Dies wird besonders wichtig, wenn mehrere Basisklassen dieselben Methoden oder Attribute definieren.

Python verwendet einen Algorithmus namens C3 Linearisierung, um die MRO zu bestimmen. Du kannst die MRO einer Klasse mit der Methode __mro__ oder der Funktion mro() eines Klassenobjekts anzeigen.

`````python
class Fahrzeug():
    def Start(self):
        print("Fahrzeug startet")

class Auto(Fahrzeug):
    def Start(self):
        print("Auto startet")

class ElektroFahrzeug(Fahrzeug):
    def Start(self):
        print("Elektrofahrzeug startet")

class Tesla(Auto, ElektroFahrzeug):
    pass

class Nissan(ElektroFahrzeug, Auto):
    pass
`````

In diesem Beispiel haben wir zwei Basisklassen **Auto** und **ElektroFahrzeug**, die beide von **Fahrzeug** erben und ihre eigene Start-Methode definieren. Die Klassen *Tesla* und *Nissan* erben von Auto und ElektroFahrzeug in **unterschiedlicher Reihenfolge**. Die Ausgabe des Scripts zeigt, wie die MRO die Auflsung der Start-Methode beeinflusst.

`````python
# MRO fr Tesla anzeigen
print("Tesla MRO:", [klasse.__name__ for klasse in Tesla.__mro__])
`````

`````python
# MRO fr Nissan anzeigen
print("Nissan MRO:", [klasse.__name__ for klasse in Nissan.__mro__])
`````

`````python
# Instanzen erstellen
tesla = Tesla()
nissan = Nissan()
`````

`````python
# Startmethode aufrufen, um zu sehen, welche Methode ausgefhrt wird
tesla.Start()
nissan.Start()
`````

### 22.7.2 |  Metaprogrammierung
---

Metaprogrammierung bezieht sich auf das Schreiben von Code, der anderen Code manipuliert oder generiert. Ein einfaches Beispiel ist die dynamische Erstellung einer Klasse zur Laufzeit.

In diesem Beispiel wird die ```type```-Funktion verwendet, um eine neue Klasse Auto dynamisch zu erstellen. Die Klasse hat einen Konstruktor, der Marke und Modell als Attribute akzeptiert, und zwei Methoden: Start und Stop. Diese Technik ermglicht es, zur Laufzeit flexibel neue Klassen zu definieren, ohne sie explizit im Code vorab definieren zu mssen.

`````python
def auto_start(self):
    print(f"{self.Marke} {self.Modell} startet.")

def auto_stop(self):
    print(f"{self.Marke} {self.Modell} hlt an.")
`````

`````python
# Dynamische Erstellung einer Klasse
AutoKlasse = type("Auto", (object,),
                  {"__init__": lambda self, marke, modell: setattr(self, "Marke", marke) or setattr(self, "Modell", modell), "Start": auto_start, "Stop": auto_stop})
`````

`````python
# Objekt der dynamisch erstellten Klasse
meinAuto = AutoKlasse("Volkswagen", "Golf")
meinAuto.Start()
meinAuto.Stop()
`````

`````python
from babel.dates import format_date
from datetime import datetime

# Erstelle ein Datum
d = datetime.today()

# Standard-Datumsformatierung
print(format_date(d, locale='de_DE'))

# Langformat
print(format_date(d, format='long', locale='de_DE'))

# Vollformat
print(format_date(d, format='full', locale='de_DE'))
`````

`````python
from babel.dates import format_time
from datetime import time

# Erstelle eine Zeit
t = datetime.today()

# Standard-Zeitformatierung
print(format_time(t, locale='de_DE'))

# Langformat (mit Sekunden)
print(format_time(t, format='long', locale='de_DE'))
`````

`````python
from babel.dates import format_datetime
from datetime import datetime

# Erstelle ein datetime-Objekt
dt = datetime.today()

# Standard-Formatierung
print(format_datetime(dt, locale='de_DE'))

# Kurzformat
print(format_datetime(dt, format='short', locale='de_DE'))

# Langformat
print(format_datetime(dt, format='long', locale='de_DE'))
`````

`````python
from babel.dates import format_datetime
import pytz
from datetime import datetime

# Aktuelles Datum und Uhrzeit in der "Europe/Berlin"-Zeitzone
berlin_tz = pytz.timezone('Europe/Berlin')
dt_berlin = datetime.now(berlin_tz)

# Aktuelles Datum und Uhrzeit in der "America/New_York"-Zeitzone
ny_tz = pytz.timezone('America/New_York')
dt_ny = datetime.now(ny_tz)

# Formatierung mit automatischer Zeitzonen-Bercksichtigung
print(format_datetime(dt_berlin, format='full', locale='de_DE'))

print(format_datetime(dt_ny, format='full', locale='en_US'))
`````

Arrow ist eine populre Python-Bibliothek, die das Arbeiten mit Datums- und Zeitstempeln erleichtert. Sie bietet viele Formatierungsoptionen, hnlich wie das eingebaute datetime-Modul, bietet jedoch zustzliche Funktionalitt und ist benutzerfreundlicher.

Hier sind einige der gngigsten Zeitformatierungs-Token, die Sie mit Arrow verwenden knnen:

- YYYY: Vierstelliges Jahr (z.B. "2023")
- YY: Zweistelliges Jahr (z.B. "23")
- MM: Monat als Zahl mit fhrender Null (z.B. "07" fr Juli)
- MMM: Abgekrzter Monatsname (z.B. "Jul")
- MMMM: Vollstndiger Monatsname (z.B. "July")
- DD: Tag des Monats mit fhrender Null
- ddd: Abgekrzter Wochentag (z.B. "Sun")
- dddd: Vollstndiger Wochentag (z.B. "Sunday")
- HH: Stunden im 24-Stunden-Format mit fhrender- Null
- hh: Stunden im 12-Stunden-Format mit fhrender Null
- mm: Minuten mit fhrender Null
- ss: Sekunden mit fhrender Null
- A: AM/PM
- ZZZ: Zeitzonenname (z.B. "UTC", "EST")
- ZZ: Zeitzone im Format "+HH:MM" oder "-HH:MM"
