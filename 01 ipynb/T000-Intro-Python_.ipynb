# <b>1 <font color='red'>|</font>  Rabattrechner
---



Entwickeln Sie einen Rabattrechner für einen Online-Shop. Der Kunde gibt den ursprünglichen Preis ein und erhält je nach Höhe unterschiedliche Rabatte:
- Bei Beträgen über 50 Euro: 5% Rabatt
- Bei Beträgen über 100 Euro: 10% Rabatt
- Bei Beträgen über 200 Euro: 20% Rabatt
Das Programm soll den ursprünglichen Preis, den Rabatt in Prozent und Euro sowie den finalen Preis ausgeben.



# <b>2 <font color='red'>|</font>  Einkaufsliste
---



Entwickeln Sie ein Programm zur Verwaltung einer Einkaufsliste. Der Benutzer soll Produkte zur Liste hinzufügen können. Bei jedem Produkt soll auch die Menge angegeben werden. Am Ende soll die komplette Liste angezeigt werden.



# <b>3 <font color='red'>|</font>  Temperaturumrechnung
---



Schreiben Sie ein Programm, das Temperaturen zwischen Celsius und Fahrenheit umrechnet. Der Benutzer soll wählen können, in welche Richtung die Umrechnung erfolgen soll. Die Formel lautet: °F = (°C × 9/5) + 32



# <b>4 <font color='red'>|</font>  Notendurchschnitt
---



Entwickeln Sie ein Programm zur Berechnung des Notendurchschnitts. Der Benutzer soll beliebig viele Noten (1-6) eingeben können. Das Programm soll den Durchschnitt berechnen und zusätzlich ausgeben, wie viele Noten eingetragen wurden.



# <b>5 <font color='red'>|</font>  Palindrom-Prüfer
---



Schreiben Sie ein Programm, das prüft, ob ein eingegebenes Wort ein Palindrom ist (vorwärts und rückwärts gelesen das gleiche Wort ergibt). Beispiele für Palindrome sind "OTTO" oder "ANNA".



# <b>6 <font color='red'>|</font>  Würfelspiel
---



Programmieren Sie ein Würfelspiel für zwei Spieler. Jeder Spieler würfelt dreimal (Zufallszahl 1-6). Die Würfe werden addiert. Der Spieler mit der höheren Summe gewinnt. Bei Gleichstand gibt es eine Wiederholung.



# <b>7 <font color='red'>|</font>  Textanalyse
---



Erstellen Sie ein Programm, das einen Text analysiert und folgende Informationen ausgibt:
- Anzahl der Zeichen (mit und ohne Leerzeichen)
- Anzahl der Wörter
- Anzahl der Sätze (durch Punkt getrennt)
- Das längste Wort im Text



# <b>8 <font color='red'>|</font>  Einkaufsrechner
---



Entwickeln Sie einen Einkaufsrechner. Der Benutzer gibt Produkte mit ihren Preisen ein. Das Programm soll:
- Die Gesamtsumme berechnen
- Den teuersten und günstigsten Artikel anzeigen
- Die durchschnittlichen Kosten pro Artikel ausgeben



# <b>9 <font color='red'>|</font>  Zahlenstatistik
---



Schreiben Sie ein Programm, das eine Liste von Zahlen analysiert. Der Benutzer gibt nacheinander Zahlen ein. Das Programm soll folgende Statistiken erstellen:
- Kleinste und größte Zahl
- Durchschnitt aller Zahlen
- Summe aller Zahlen
- Anzahl der positiven und negativen Zahlen
